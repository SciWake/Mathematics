<h1 style="color:black" align="center">Правила сложения и умножения</h1>

Пусть у нас есть два множества объектов $A$ и $B$:

$\large A = \left \{a_1,...,a_n \right \}$

$\large B = \left \{b_1,...,b_m \right \}$

<h1 style="color:#008B8B">Основные правила</h1>

### Правило сложениия

Количество способов вытащить либо объект из множества $A$, либо объект из множества $B$, то, естественно, это количество способов равняется сумме количеств объектов в наших множествах: $n + m$.

### Правило умножения

Количество способов выбрать, выбрать сперва ровно один объект из множества $A$, а вслед за ним ровно один объект из множества $B$, равно: $n \cdot m$.

Сначала выбрать объект $a_1$, потом вслед за ним поставить объект $b_1$. Дальше можно сперва выбрать объект $a_1$, но вслед за ним поставить объект $b_2$:

$$\large \begin{matrix}
a_1 b_1 & a_1 b_2 & a_1 b_3 & ... & a_1 b_m\\
a_2 b_1 & a_2 b_2 & a_2 b_3 & ... & a_2 b_m\\
& & ... \\
a_n b_1 & a_n b_2 & a_n b_3 & ... & a_n b_m\\
\end{matrix}$$

__Следствие из правила умножения:__

Пусть у нас есть множества $A_1, \ldots, A_k$ каждое из которых состоит из некоторого числа объектов. 

Давайте зафиксируем величину их мощности. Мощность — это просто количество элементов множеств. Итак, мощность $A_i$, количество элементов в этом множестве — давайте просто обозначим $|A_i| = n_i$. Где $n$ с индексом $i$ — это просто количество объектов, которые входят в данное множество.

Тогда число способов извлечь сперва, один объект из множества $A_1$, поставить вслед за ним, один объект из множества $A_2$, расположить вслед за этими двумя еще один, ровно один объект из множества $A_3$  и так далее... И всю эту цепочку последовательно выбираемых объектов завершить ровно одним объектом из множества $A_k$. Все это число способов равно произведению чисел $n_1$ (количество объектов в $A_1$), это было количество объектов в $A_1$, $n_2$, это было по определению количество объектов $A_2$, и так далее, вплоть до $n_k$. Это количество объектов в $A_k$:

$$\large n_1 \cdot n_2 \cdot \ldots \cdot n_k$$

<h1 style="color:#008B8B">Задачи на основные правила</h1>

<h3 style="color:#2E8B57">Задача № 1</h3>

Найти количество автомобильных номеров в рамках одного региона в рамках страны.
Пример номера `K884MX`, `X486BB`, `K268OB`.

<h4 align='center'>Решение:</h4>

Виды множеств в данной задаче:
* $A_1, A_5, A_6$ - Множество букв;

* $A_2, A_3, A_4$ - Множество цифр.

Количество элементов в множестве:
* $n_1 = n_5 = n_6$ - Количество букв, которых $12$ штук внутри множества;
* $n_2 = n_3 = n_4$ - Количество цифр, которых $10$ штук внутри множества.

Следовательно, пользуясь правилом умножения получаем:

$12^3 \cdot 10^3 = 1728000$

__Ответ:__ Количество автомобильных номеров в рамках одного региона $1728000$.

<h3 style="color:#2E8B57">Задача № 2</h3>

Есть компания из $6$ человек, среди которых $2$ девушки. В ложе всего $6$ мест. Сколько существует рассадок этой компании, если девушки не сидят с краю?

<h4 align='center'>Решение:</h4>

Значит, есть $4$ человека, которые могут сидеть с краю. Выбираем из $4$ человек двух, которые займут крайние места: $4 \cdot 3$ - количество способов рассадить парней по краям. Вслед за ними осталось рассадить остальных $4$ людей: всего таких способов $4!$

Найдём количество рассадок:

$4 \cdot 3 \cdot 4! = 288$

__Ответ:__ Количество способов рассадить $6$ человек в ложе равно: $288$

<h3 style="color:#2E8B57">Задача № 3</h3>

Есть пароль, и он может состоять из цифр и из маленьких букв латинского алфавита.

<h4 align='center'>Решение:</h4>

`1. Сколько таких паролей длины 8?`

Цифр $10$ штук, букв латинского алфавита $26$ букв. Получаем всего $36$ символов из которых можно составить пароль.

На первую позицию пароля мы можем поставить любой из $36$ символов, на вторую позицию аналогично и так $8$ раз.

__Ответ:__ $36 \cdot 36 \cdot \ldots \cdot 36 = 36^8$

`2. Сколько таких паролей длины 8, в которых присутствует хотя бы одна цифра?`

Посчитаем, сколько есть паролей, в которых нет ни одной цифры:

Если мы не используем цифры, то это просто означает, что на каждую позицию в нашем распоряжении имеется не $36$ символов, а только $26$, ведь ни одну из цифр ни на одну из позиций мы ставить в этом случае не собираемся, когда говорим о количестве паролей, не содержащих цифру. 

Число паролей без цифр $26^8$, а число всех паролей это $36^8$, значит, число паролей, которые содержат хотя бы одну цифру, это разность общего числа паролей и числа паролей, которые напротив цифры не содержатся. То есть вот эта величина, которая ищется в пункте "1" — это есть просто $36$ в восьмой степени − $26$ в восьмой степени. 

__Ответ:__ $36^8 - 26^8$

`3. Сколько таких паролей длины 8, если в них присутствуют совпадающие символы?`

Возьмём общее количество паролей, которое равняется $36^8$ степени. И из этого общего количества вычтем количество тех паролей, в которых, напротив, все символы различны. 

На первую позицию выбрать символ можно $36$ способами, на вторую позицию осталось только $35$ символов и так надо выбрать $8$ позиций: $36 \cdot 35 \cdot ... \cdot 29$

Из общего количества паролей вычтем, количество паролей, где все символы различны:

__Ответ:__ $36^8 - 36 \cdot 35 \cdot \ldots \cdot 29$

<h1 style="color:black" align="center">Принцип Дирихле</h1>

**Формулировка принципа Дирихле**

Есть $n$ ящиков и $n+1$ кролик. Как ни рассаживай кроликов по ящикам найдется ящик, в котором окажутся, как минимум, $2$ кролика.

<h1 style="color:#008B8B">Задачи на принцип Дирихле</h1>

<h3 style="color:#2E8B57">Задача № 1</h3>

Есть монеты стоимостью $1$ рубль, $2$ рубля, $5$ рублей и $10$ рублей. Спрашивается: какое минимальное количество таких монет необходимо и достаточно взять, чтобы с гарантией среди этих взятых монет, нашлось $6$ монет одного достоинства?

<h4 align='center'>Решение:</h4>

Если взять $20$ монет, тогда может быть такая ситуация, что мы достали:

$5$ Монет достоинством $1$;<br> 
$5$ Монет достоинством $2$;<br>
$5$ Монет достоинством $5$;<br>
$5$ Монет достоинством $10$.

Количество этих монет равно $20$, если достать ещё одну монету, тогда среди этих $4$ монет, обязательно найдётся $6$ монет одного достоинства.

__Ответ:__ $4 \cdot 5 + 1 = 21$

<h3 style="color:#2E8B57">Задача № 2</h3>

Есть $101$ пуговица одного из $11$ цветов. Утверждается, что найдется $11$ пуговиц одного (все пуговицы одинаковые), либо найдется $11$ пуговиц разного цвета (Каждая пуговица имеет свой уникальный цвет).

<h4 align='center'>Решение:</h4>

Давайте действовать методом от противного. Предположим, что не верно ни первое, ни второе утверждение, что этой альтернативы нет. Это значит, что верно отрицание первого и отрицание второго. 

`1. Предположим, что нет 11 пуговиц одного цвета нет;`

`2. Предположим, что нет 11 пуговиц разного цвета.`

Из второго предположения конечно же следует, что задействовано не более чем $10$ цветов.

Первое предположение говорит, что нет $11$ пуговиц одного цвета. Иначе говоря пуговиц каждого конкретного цвета не больше $10$, если нету $11$ пуговиц какого-либо одного цвета, то пуговиц каждого конкретного цвета не больше $10$.

Пуговиц каждого цвета не больше $10$, значит и первого тоже. Дальше есть пуговицы второго цвета и их тоже не больше $10$. И так далее. А номер последнего цвета, сам не превосходит $10$. 

$1$ цвет пуговиц и в него покрашено $\leq 10$

$2$ цвет пуговиц и в него покрашено $\leq 10$

$\large \ldots \ldots \ldots \ldots \ldots \ldots \ldots \ldots$

$n \leq 10$ цвет пуговиц и в него покрашено $\leq 10$


Значит, сколько у нас суммарно получается всего пуговиц в нашем распоряжении? Не больше $10$, не больше $10$, не больше $10$, и так $n$ раз. Всего получается пуговиц не больше, чем $10$ умножить на $n$. При этом $n$ не превосходит $10$. Мы получаем не больше $100$. И это дает нам искомое противоречие, ведь мы-то знаем, что у нас $101$ пуговица. Все, противоречие мы предположили неправильно, и значит действительно наша альтернатива была справедлива. Либо найдется $11$ пуговиц одного цвета, либо найдется такая вот радужная раскрасочка, когда все пуговицы разных цветов.

<h3 style="color:#2E8B57">Задача № 1</h3>

Автомобильные номера штата Калифорнии состоят из одной цифры, трёх больших букв латинского алфавита и ещё трёх цифр (например, 5PPP064). Сколько всего имеется номеров такого типа?

<h4 align='center'>Решение:</h4>

В качестве первой цифры номера можно взять любую из $9$ цифр $(\{1,\ldots,9\})$, в качестве каждой из трех букв можно взять любую из $26$ букв латинского алфавита, и ещё три цифры, для каждой из которой есть $10$ вариантов выбора $(\{0,\ldots,9\})$. Итого, по правилу умножения получаем: $9 \cdot 10^3 \cdot 26^3 = 158184000$

__Ответ:__ $158184000$

<h3 style="color:#2E8B57">Задача № 2</h3>

Путешественнику нужно добраться из города $A$ в город $F$ двигаясь каждый раз вправо или вверх (см. карту). Сколькими способами это можно сделать?

![Картинка к задаче 2](img/Task_2.png)

<h4 align='center'>Решение:</h4>

Используя правила сложения и умножения, получаем, что количество способов попасть из $A$ в $F$ равно сумме количества способов пройти по маршрутам $A \rightarrow D \rightarrow E \rightarrow F $, $A \rightarrow B \rightarrow E \rightarrow F$ и $A \rightarrow B \rightarrow С \rightarrow F$. В итоге получаем: $2\cdot 2\cdot 2 + 3\cdot 1\cdot 2+ 3\cdot 3\cdot 1 = 8 + 6 + 9 = 23$.

__Ответ:__ $23$

<h3 style="color:#2E8B57">Задача № 3</h3>

В мешке $50$ шаров, отличающихся только цветом: $8$ красных, $9$ синих, $9$ желтых, остальные – поровну черные и белые. Какое наименьшее число шаров надо вынуть из мешка, не видя их, чтобы среди них было не менее $7$ шаров одного цвета?

<h4 align='center'>Решение:</h4>

Сначала заметим, что черных и белых шаров по $\frac{50-8-9-9}{2} = 12 > 7$, то есть шаров белого и чёрного цвета по $12$ штук.

Если взять всего $30$ шаров, то может получиться так, что мы взяли только по 6 шаров каждого цвета. Значит, нужно брать больше $30$ шаров. Попробуем вытянуть $31$ шар. Тогда по принципу Дирихле, если роль "клеток" играют цвета, а роль "кроликов" вытянутые шары, получаем, что какие бы мы ни вытащили шары, среди них обязательно найдется по крайней мере $7$ шаров одного цвета.

__Ответ:__ $31 = 5 \cdot 6 + 1$

<h3 style="color:#2E8B57">Задача № 4</h3>

$15$ футбольных команд (в каждой по $11$ человек) летят из Москвы в Санкт-Петербург на соревнования. Какое минимальное количество мест может быть в самолете, чтобы гарантированно нашлась команда, долетевшая в полном составе?

<h4 align='center'>Решение:</h4>

Если в самолете будет всего $150$ мест, то можно будет посадить в него по $10$ человек от каждой из $15$ команд, тогда самолет будет заполнен целиком, но в нем не будет ни одной команды в полном составе. Значит, нужен с самолет больше, чем со $150$ местами. Рассмотрим самолет, в котором $151$ место. Тогда по принципу Дирихле, если роль "клеток" играют команды, а роль "кроликов" футболисты, получаем, что как бы ни произошла рассадка футболистов, среди пассажиров самолета обязательно найдется хотя бы одна команда, летевшая в полном составе 

__Ответ:__ $151 = 15 \cdot 10 + 1$

<h3 style="color:#2E8B57">Задача № 5</h3>

Сколько чисел от $1$ до $9999$ (включая $1$ и $9999$) не имеют в своей десятичной записи одинаковых подряд идущих цифр? (к примеру, не подходят $1488$, $2259$, $3233$)

<h4 align='center'>Решение:</h4>

Первой цифрой числа, не имеющего в своей записи подряд идущих цифр, может быть любая цифра, кроме $0$, то есть всего $9$ вариантов. Второй цифрой этого числа может быть любая цифра, кроме той, которая стоит на первом месте, то есть $9$ вариантов, третьей - любая, кроме стоящей на втором месте, то есть опять $9$ вариантов, и т.д. Тогда таких четырехзначных чисел всего $9^4$, аналогично, трехзначных - $9^3$, двузначных - $9^2$ и однозначных - $9$. Всего получаем $9^4 + 9^3 + 9^2 + 9 = 7380$

__Ответ:__ $7380$

<h3 style="color:#2E8B57">Задача № 6</h3>

У вас есть $4$ ящика и $15$ кроликов. Отметьте верные утверждения:

- [x] При любой рассадке кроликов по ящикам найдётся ящик, в котором будет сидеть по крайней мере 3 кролика.


- [x] При любой рассадке кроликов по ящикам найдётся ящик, в котором будет сидеть по крайней мере 4 кролика.


- [ ] При любой рассадке кроликов по ящикам найдётся ящик, в котором будет сидеть по крайней мере 5 кроликов.


- [ ] При любой рассадке кроликов по ящикам найдётся ящик, в котором будет сидеть по крайней мере 6 кроликов.

<h4 align='center'>Решение:</h4>

При такой рассадке найдётся ящик, в котором сидит $\geq [\frac{15}{4}]$ кроликов, то есть $4$ кролика. Для $5$ и $6$ кроликов есть контрпример: $4, 4, 4, 3$

<h3 style="color:#2E8B57">Задача № 7</h3>

Имеется $4$ банки желтой краски, $5$ банок синей краски и $7$ банок красной краски, все банки разного размера. Отметьте верные утверждения:

- [X] Банку краски можно выбрать 16 способами.


- [ ] По одной банке каждой краски (1 желтую, 1 синюю, 1 красную) можно выбрать 16 способами.


- [X] По одной банке каждой краски (1 желтую, 1 синюю, 1 красную) можно выбрать 140 способами.


- [ ] Банку краски можно выбрать 140 способами.

<h4 align='center'>Решение:</h4>

Одну банку краски по правилу сложения можно выбрать $4 + 5 + 7 = 16$ способами. Если же выбирать по одной банке каждой краски, то по правилу умножения получаем $4 \cdot 5 \cdot 7 = 140$ способов.
