<h2 style="color:black" align="center">Правила сложения и умножения</h2>

Пусть у нас есть два множества объектов $A$ и $B$:

$A = \left \{a_1,...,a_n \right \}$, $B = \left \{b_1,...,b_m \right \}$

<h3 style="color:#008B8B">Правило сложения</h3>

Количество способов вытащить либо объект из множества $A$, либо объект из множества $B$, то, естественно, это количество способов равняется сумме количеств объектов в наших множествах: $n + m$.


<h3 style="color:#008B8B">Правило умножения</h3>

Количество способов выбрать, выбрать сперва ровно один объект из множества $A$, а вслед за ним ровно один объект из множества $B$, равно: $n \cdot m$.

Сначала выбрать объект $a1$, потом вслед за ним поставить объект $b1$. Дальше можно сперва выбрать объект $a1$, но вслед за ним поставить объект $b2$:

$$\begin{matrix}
a_1 b_1 & a_1 b_2 & a_1 b_3 & ... & a_1 b_m\\
a_2 b_1 & a_2 b_2 & a_2 b_3 & ... & a_2 b_m\\
& & ... \\
a_n b_1 & a_n b_2 & a_n b_3 & ... & a_n b_m\\
\end{matrix}$$

__Следствие из правила умножения:__

Пусть у нас есть множества $A_1$ ... $A_k$ каждое из которых состоит из некоторого числа объектов. 

Давайте зафиксируем величину их мощности. Мощность — это просто количество элементов множеств. Итак, мощность $A_i$, количество элементов в этом множестве — давайте просто обозначим $|A_i| = n_i$. $n$ с индексом $i$ — это просто количество объектов, которые входят в данное множество.

Тогда число способов извлечь сперва, один объект из множества $A_1$, поставить вслед за ним, один объект из множества $A_2$, расположить вслед за этими двумя еще один, ровно один объект из множества $A_3$  и так далее... И всю эту цепочку последовательно выбираемых объектов завершить ровно одним объектом из множества $A_k$. Все это число способов равно произведению чисел $n_1$, это было количество объектов в $A_1$, $n_2$, это было по определению количество объектов $A_2$, и так далее, вплоть до $n_k$. Это количество объектов в $A_k$:

$$n_1 \cdot n_2 \cdot ... \cdot n_k$$

<h3 style="color:#2E8B57">Задача № 1</h3>

Найти количество автомобильных номеров в рамках одного региона в рамках страны.
Пример номера `K884MX`, `X486BB`, `K268OB`.

<h4 align='center'>Решение:</h4>

Виды множеств в данной задаче:
* $A_1, A_5, A_6$ - Множество букв;

* $A_2, A_3, A_4$ - Множество цифр.

Количество элементов в множестве:
* $n_1 = n_5 = n_6$ - Количество букв, которых $12$ штук внутри множества;
* $n_2 = n_3 = n_4$ - Количество цифр, которых $10$ штук внутри множества.

Следовательно, пользуясь правилом умножения получаем:

$12^3 \cdot 10^3 = 1728000$

__Ответ:__ Количество автомобильных номеров в рамках одного региона $1728000$.

<h3 style="color:#2E8B57">Задача № 2</h3>

Есть компания из $6$ человек, среди которых $2$ девушки. В ложе всего $6$ мест. Сколько существует рассадок этой компании, если девушки не сидят с краю?

<h4 align='center'>Решение:</h4>

Значит, есть $4$ человека, которые могут сидеть с краю. Выбираем из $4$ человек двух, которые займут крайние места: $4 \cdot 3$ - количество способов рассадить парней по краям. Вслед за ними осталось рассадить остальных $4$ людей: всего таких способов $4!$

Найдём количество рассадок:

$4 \cdot 3 \cdot 4! = 288$

__Ответ:__ Количество способов рассадить $6$ человек в ложе равно: $288$

<h3 style="color:#2E8B57">Задача № 3</h3>

Есть пароль, и он может состоять из цифр и из маленьких букв латинского алфавита.

<h4 align='center'>Решение:</h4>

`1. Сколько таких паролей длины $8$?`

Цифр $10$ штук, букв латинского алфавита $26$ букв. Получаем всего $36$ символов из которых можно составить пароль.

На первую позицию пароля мы можем поставить любой из $36$ символов, на вторую позицию аналогично и так $8$ раз.

__Ответ:__ $36 \cdot 36 \cdot ... \cdot 36 = 36^8$

`2. Сколько таких паролей длины $8$, в которых присутствует хотя бы одна цифра?`

Посчитаем, сколько есть паролей, в которых нет ни одной цифры:

Если мы не используем цифры, то это просто означает, что на каждую позицию в нашем распоряжении имеется не $36$ символов, а только $26$, ведь ни одну из цифр ни на одну из позиций мы ставить в этом случае не собираемся, когда говорим о количестве паролей, не содержащих цифру. 

Число паролей без цифр $26^8$, а число всех паролей это $36^8$, значит, число паролей, которые содержат хотя бы одну цифру, это разность общего числа паролей и числа паролей, которые напротив цифры не содержатся. То есть вот эта величина, которая ищется в пункте "1" — это есть просто $36$ в восьмой степени − $26$ в восьмой степени. 

__Ответ:__ $36^8 - 26^8$

`3. Сколько таких паролей длины $8$, если в них присутствуют совпадающие символы?`

Возьмём общее количество паролей, которое равняется $36^8$ степени. И из этого общего количества вычтем количество тех паролей, в которых, напротив, все символы различны. 

На первую позицию выбрать символ можно $36$ способами, на вторую позицию осталось только $35$ символов и так надо выбрать 8 позиций: $36 \cdot 35 \cdot ... \cdot 29$

Из общего количества паролей вычтем, количество паролей, где все символы различны:

__Ответ:__ $36^8 - 36 \cdot 35 \cdot ... \cdot 29$

<h3 style="color:#008B8B">Формулировка принципа Дирихле</h3>

Есть $n$ ящиков и $n+1$ кролик. Как ни рассаживай кроликов по ящикам, давайте я это так и напишу, как ни рассаживай кроликов по ящикам, найдется ящик, в котором окажутся, как минимум, $2$ кролика.

<h3 style="color:#2E8B57">Задача № 1</h3>

Есть монеты стоимостью $1$ рубль, $2$ рубля, $5$ рублей и $10$ рублей. Спрашивается: какое минимальное количество таких монет необходимо и достаточно взять, чтобы с гарантией среди этих взятых монет, нашлось $6$ монет одного достоинства?

<h4 align='center'>Решение:</h4>

Если взять $20$ монет, тогда может быть такая ситуация, что мы достали:

$5$ Монет достоинством $1$;<br> 
$5$ Монет достоинством $2$;<br>
$5$ Монет достоинством $5$;<br>
$5$ Монет достоинством $10$.

Количество этих монет равно $20$, если достать ещё одну монету, тогда среди этих $4$ монет, обязательно найдётся $6$ монет одного достоинства.

__Ответ:__ $4 \cdot 5 + 1 = 21$


<h3 style="color:#2E8B57">Задача № 1</h3>

Есть $101$ пуговица одного из $11$ цветов. Утверждается, что найдется $11$ пуговиц одного (все пуговицы одинаковые), либо найдется $11$ пуговиц разного цвета (Каждая пуговица имеет свой уникальный цвет).

<h4 align='center'>Решение:</h4>

Давайте действовать методом от противного. Предположим, что не верно ни первое, ни второе утверждение, что этой альтернативы нет. Это значит, что верно отрицание первого и отрицание второго. 

1. Предположим, что нет $11$ пуговиц одного цвета нет;
2. Предположим, что нет $11$ пуговиц разного цвета.

Из второго предположения конечно же следует, что задействовано не более чем $10$ цветов.

Первое предположение говорит, что нет $11$ пуговиц одного цвета. Иначе говоря пуговиц каждого конкретного цвета не больше $10$, если нету $11$ пуговиц какого-либо одного цвета, то пуговиц каждого конкретного цвета не больше $10$.

Пуговиц каждого цвета не больше $10$, значит и первого тоже. Дальше есть пуговицы второго цвета и их тоже не больше $10$. И так далее. А номер последнего цвета, сам не превосходит $10$. 

$1$ цвет пуговиц и в него покрашено $\leq 10$

$2$ цвет пуговиц и в него покрашено $\leq 10$

.....................................

$n \leq 10$ цвет пуговиц и в него покрашено $\leq 10$


Значит, сколько у нас суммарно получается всего пуговиц в нашем распоряжении? Не больше $10$, не больше $10$, не больше $10$, и так $n$ раз. Всего получается пуговиц не больше, чем $10$ умножить на $n$. При этом n, как я уже сказал, не превосходит $10$. Мы получаем не больше $100$. И это дает нам искомое противоречие, ведь мы-то знаем, что у нас $101$ пуговица. Все, противоречие мы предположили неправильно, и значит действительно наша альтернатива была справедлива. Либо найдется $11$ пуговиц одного цвета, либо найдется такая вот радужная раскрасочка, когда все пуговицы разных цветов.
