<h2 style="color:black">Правила сложения и умножения</h2>

Пусть у нас есть два множества объектов $A$ и $B$:

$A = \left \{a_1,...,a_n \right \}$, $B = \left \{b_1,...,b_m \right \}$

<h3 style="color:#008B8B">Правило сложения</h3>

Количество способов вытащить либо объект из множества $A$, либо объект из множества $B$, то, естественно, это количество способов равняется сумме количеств объектов в наших множествах: $n + m$.


<h3 style="color:#008B8B">Правило умножения</h3>

Количество способов выбрать, выбрать сперва ровно один объект из множества $А$, а вслед за ним ровно один объект из множества $В$, равно: $n \cdot m$.

Сначала выбрать объект $a1$, потом вслед за ним поставить объект $b1$. Дальше можно сперва выбрать объект $a1$, но вслед за ним поставить объект $b2$:

$$\begin{matrix}
a_1 b_1 & a_1 b_2 & a_1 b_3 & ... & a_1 b_m\\
a_2 b_1 & a_2 b_2 & a_2 b_3 & ... & a_2 b_m\\
& & ... \\
a_n b_1 & a_n b_2 & a_n b_3 & ... & a_n b_m\\
\end{matrix}$$

__Следствие из правила умножения:__

Пусть у нас есть множества $А_1$ ... $A_k$ каждое из которых состоит из некоторого числа объектов. 

Давайте зафиксируем величину их мощности. Мощность — это просто количество элементов множеств. Итак, мощность $A_i$, количество элементов в этом множестве — давайте просто обозначим $|A_i| = n_i$. $n$ с индексом $i$ — это просто количество объектов, которые входят в данное множество.

Тогда число способов извлечь сперва, один объект из множества $А_1$, поставить вслед за ним, один объект из множества $А_2$, расположить вслед за этими двумя еще один, ровно один объект из множества $А_3$  и так далее... И всю эту цепочку последовательно выбираемых объектов завершить ровно одним объектом из множества $А_k$. Все это число способов равно произведению чисел $n_1$, это было количество объектов в $А_1$, $n_2$, это было по определению количество объектов $А_2$, и так далее, вплоть до $n_k$. Это количество объектов в Аk:

$$n_1 \cdot n_2 \cdot ... \cdot n_k$$

<h3 style="color:#2E8B57">Задача № 1</h3>

Найти количество автомобильных номеров в рамках одного региона в рамках страны.
Пример номера `K884MX`, `X486BB`, `K268OB`.

<h4 align='center'>Решение:</h4>

Виды множеств в данной задаче:
* $A_1, A_5, A_6$ - Множество букв;

* $A_2, A_3, A_4$ - Множество цифр.

Количество элементов в множестве:
* $n_1 = n_5 = n_6$ - Количество букв, которых 12 штук внутри множества;
* $n_2 = n_3 = n_4$ - Количество цифр, которых 10 штук внутри множества.

Следовательно, пользуясь правилом умножения получаем:

$12^3 \cdot 10^3 = 1728000$

__Ответ:__ Количество автомобильных номеров в рамках одного региона $1728000$.
