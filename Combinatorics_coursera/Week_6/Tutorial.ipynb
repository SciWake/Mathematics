<h1 align="center">Формула включений и исключений</h1>

<h1 style="color:#008B8B">Пример 1</h1>

**Условие**

В аудитории находится $30$ человек. Есть следующие языки: Английский, Французский, Немецкий. Предположим, что $20$ человек владеют Английским языком, $5$ человек владеют французским языком и $5$ человек владеют Немецким языком.

1. Одновременно А, Ф язык знают $2$ человека (они могут знать и Н язык, мы только оговорили то, что они точно знают два языка);

2. А, Н - $2$ человека;

3. Н, Ф - $1$ человек;

4. А, Н, Ф - $1$ человек.

Сколько человек в аудитории не знают ни один из представленных языков?

**Решение**

Всего $30$ человек, из которых мы вычитаем $20$ человек, которые точно знаю А язые:

$\large 30 - 20$

Дальше необходимо вычесть людей, которые знают Ф язык и Н язык:

$\large 30 - 20 - 5 - 5$

Но среди людей, которые знали А были люди которые знали Ф язык и наоборот, среди людей, которые знали Ф были люди которые знали А язык. Поэтому исключая $20$ и $5$, мы исключили пересечение двух множеств (исключили ещё раз тех, кто одновременно значет А и Ф язык), необходимо добавить тех кто знает А и Ф:

$\large 30 - 20 - 5 - 5 + 2$

Аналогичным образом вычитая $20$ и $5$ мы лишний раз исключили людей, которые одновременно знают А и Ф язык:

$\large 30 - 20 - 5 - 5 + 2 + 2$

Двухкратное вычитание $5$  означает, что мы один раз вычитаем людей, которые знают Ф и людей, которые знают Н. Но среди этих людей, есть человек, который знает Н и Ф одновременно, поэтому его необходимо добавить:

$\large 30 - 20 - 5 - 5 + 2 + 2 + 1$

Когда мы производили включение людей, лишний раз включили одного человека, который одновременно знает все три языка, его необходимо вычесть:

$\large 30 - 20 - 5 - 5 + 2 + 2 + 1 - 1 = 4$

<h1 style="color:#008B8B">Общая постановка задачи</h1>

**Определение свойств для объектов:**

Имеется $a_1, \ldots, a_N$ объектов. Есть перечень свойств, которые мы проверяем относительно этих объектов $\alpha_1, \ldots, \alpha_n$.

1) $N(\alpha_1)$ - количество объектов среди $a_1, \ldots, a_N$, которые обладают как минимум свойством $\alpha_1$.

Аналогичным образом определим остальные свойства $N(\alpha_2), \ldots, N(\alpha_n)$ - количество объектов, которые обладают как минимум свойством $\alpha_2, \ldots, \alpha_n$;

2) Можно определить парные свойства для объектов: $N(\alpha_1, \alpha_2), N(\alpha_1, \alpha_3), \ldots, N(\alpha_{n-1}, \alpha_n)$ - берём два произвольных свойства из множества свойств, где количество свойств $n$ штук. И количество таких пар будет $C_n^2$;

3) Можно определить тройные свойства для объектов: $N(\alpha_1, \alpha_2, \alpha_3), \ldots, N(\alpha_{n-2}, \alpha_{n-1} \alpha_n)$ - берём три произвольных свойства из множества свойств мощностью $n$. И число таких комбинаций будет $C_n^3$;

n) И так далее до: $N(\alpha_1, \ldots, \alpha_n)$ - количество объектов которые обладают всеми свойствами разом.

**Формулировка задачи:**

Необходимо найти $N(\alpha_1^{'}, \ldots, \alpha_n^{'})$. Подразумевая под штрихом отрицание свойства, значит, $\alpha_1^{'}$ и так далее - это означает, что объект не обладает данным свойством.

Для вычисления этой величины, необходимо применить формулу включений и исключений.

<h1 style="color:#008B8B">Теорема 1</h1>

$\large N(\alpha_1^{'}, \ldots, \alpha_n^{'}) = N - N(\alpha_1) - \ldots - N(\alpha_n) + N(\alpha_1, \alpha_2) + \ldots + N(\alpha_{n-1}, \alpha_{n}) - \ldots + (-1)^n N(\alpha_1, \ldots, \alpha_n)$

<h1 align="center">Задачи на формулу включений и исключений</h1>

<h3 style="color:#2E8B57">Задача № 1</h3>

В России $1112$ городов. Есть два путешествинника Алексей и Даниил, где Алексей посетил половину из этих городов, а Даниил посетил тоже половину городов. Известно, что в $\frac{1}{4}$ городов побывал как Алексей, так и Даниил. Сколько городов России не посетили ни один из них?

<h4 align='center'>Решение:</h4>

Объектами являются города России $N = 1112$. Свойств $n=2$, $\alpha_1$ состоит в том, что город посещал Алексей, $\alpha_2$ состоит в том, что город посещал Даниил.

$\large N(\alpha_1) = \frac{1112}{2}$

$\large N(\alpha_2) = \frac{1112}{2}$

$\large N(\alpha_1, \alpha_2) = \frac{1112}{4}$

$\large N(\alpha_1^{'}, \alpha_2^{'}) = N - N(\alpha_1) - N(\alpha_2) + N(\alpha_1, \alpha_2) = 278$
