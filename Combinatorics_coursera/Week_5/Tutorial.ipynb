<h1 align="center">Полиноминальный коэффициент</h1>

<h1 style="color:#008B8B">Перестановки букв в слове "Комбинаторика"</h1>

### Введение

Возьмём слово `л, я, г, у, ш, к, а`, где количество букв $7$ и все они различные, тогда количество слов которое можно составить из этих букв $7!$.

Рассмотрим слово, где буквы могут повторяться `к, о, к`. В данном примере мы можем получить только три различных слова, но количество перестановок составит $3! = 6$. Соответственно, количество слов, которые можно составить когда буквы повторяются будет вычисляться по другой формуле.

### Комбинаторика

`к, о, м, б, и, н, а, т, о, р, и, к, а`

Мы имеем дело с кучей различных совпадений, то есть если мы переставляем между собой буквы `к`, то мы получаем одно и то же слово, это нам надо учитывать.

**Запишем параметры слова:**

Количество букв в слове $13$. Запишем встречаемость букв `к` - 2, `о` - 2, `м` - 1, `б` - 1, `и` - 1, `н` - 1, `а` - 2, `т` - 1, `р` - 1

**Как посчитать количество способов расставить буквы так, чтобы каждый раз получалось новое слово?**

У нас есть всего $13$ позиций, на которые мы можем расставлять буквы. Во-первых, мы можем поставить на какие-то две позиции из $13$-ти имеющихся букву `к`. Таких способов $C_{13}^2$. Мы зафиксировали какие-то две позиции, на которые поставлены буквы `к`. У нас осталось незадействованных `11` позиций, продолжим данную операцию. Выберим из этих оставшихся $11$ позиций две позиции для буквы `o` - $C_{11}^2$ и так далее...

Пользуемся, правилом умножения, потому что мы последовательно выбираем друг за другом сначала позиции для букв `к`, потом позиции для букв `о`, потом позиции для букв `м` и так далее, то есть мы это делаем одно за другим. Стало быть, применимо правило умножения. И искомое количество принимает следующий вид:

$\large C_{13}^2 \cdot C_{11}^2 \cdot C_{9}^1 \cdot C_{8}^1 \cdot C_{7}^2 \cdot C_{5}^1 \cdot C_{4}^2 \cdot C_{2}^1 \cdot C_{1}^1 = $

**Упрощение ответа:**

Но зачастую принято этот ответ упрощать до некоторого очень приятного вида. Расписываем просто каждую $C$ согласно формуле с факториалами:

$= \large \frac{13!}{2! \cdot 11!} \cdot \frac{11!}{2! \cdot 9!} \cdot \frac{9!}{1! \cdot 8!} \cdot \frac{8!}{1! \cdot 7!} \cdot \frac{7!}{2! \cdot 5!} \cdot \frac{5!}{1! \cdot 4!} \cdot \frac{4!}{2! \cdot 2!} \cdot \frac{2!}{1! \cdot 1!} \cdot \frac{1!}{1! \cdot 0!} = $

Дальше факториалы, которые расположены по диагонали сокращаются:

$= \large \frac{13!}{2! \cdot 2! \cdot 1! \cdot 2! \cdot 1! \cdot 2! \cdot 1! \cdot 1!}$

Почему эта запись хорошая? Дело в том, что $13$ — это исходное количество букв, которые имелись в нашем распоряжении. А в знаменателе стоит произведение факториалов, где каждый факториал - это то, сколько раз встречается буква в данном слове.

<h1 style="color:#008B8B">Доказательство формулы для полиномиального коэффициента</h1>

**Определение объектов:**

$\large n$ - количество объектов;

$\large n_1$ - количество объектов первого типа;

$\large n_2$ - количество объектов второго типа;

$\large n_k$ - количество объектов $k$-го типа.

Сумма чисел $n_1, \ldots, n_k$ должна равняться общему количеству объектов:

$\large n = n_1 + \ldots + n_k$

**Постановка вопроса:**

И общий вопрос состоит в том, каково количество способов переставить в различном порядке объекты так, чтобы каждый раз получалась новая последовательность. Давайте это количество обозначим вот так:

$\large P(n_1, \ldots, n_k)$ - количество способов составить последовательность из $n$ объектов.

### Теорема для формулы для полиномиального коэффициента

$\large P(n_1, \ldots, n_k) = \frac{n!}{n_1! \cdot \ldots \cdot n_k!}$ 

<h3 align='center'>Доказательство - Общий случай:</h3>

**Выбор объектов:**

Есть всего $n$ позиций из которых можно составлять последовательность. Тогда:

$\large С_{n}^{n_1}$ - количество способов выбрать позиции для объектов первого типа;

После того как мы выбрали $n_1$ позицию для объектов первого типа, в нашем распоряжении, конечно же осталось только $n − n_1$ позиция и  на эти $n − n_1$ позиции мы можем расставлять объекты вторго, третьего, четвёртого типа и так далее.

$\large С_{n - n_1}^{n_2}$ - количество способов выбрать позиции для объектов второго типа;

Дальше позиции становятся еще на $n_2$ меньше: $C_{n − n_1 − n_2}$ и из них мы выбираем $n_3$.

$\large C_{n − n_1 − n_2}^{n_3}$ - количество способов выбрать позиции для объектов третьего типа;

В последней $C$, мы должны расставлять объекты $k$-того типа. Это значит, что на предыдущем шаге мы уже задействовали объекты $k − 1$ типа. 

$\large C_{n − n_1 − n_2 - \ldots - n_{k - 1}}$

И на объекты $k − 1$ типа мы использовали $n_{k - 1}$ позицию. Потому что ровно столько объектов $k − 1$ типа имелось в нашем распоряжении и все их мы должны были задействовать. Значит $n_1, n_2, n_3, \ldots, n_{k−1}$ количество объектов мы вычели из $n$. А выбрать нам нужно позиции для объектов $k$-того типа. То есть $n$ с индексом $k$. 

$\large C_{n − n_1 − n_2 - \ldots - n_{k - 1}}^{n_k}$ -  количество способов выбрать позиции для объектов $k$-того типа.

**Последний выбор:**

Вспомним, что 

$\large n = n_1 + \ldots + n_k$

Перенесём все слогаемые из правой части в левую, кроме последнего:

$\large n - n_1 - \ldots - n_{k-1} = n_k$, тогда 

$\large C_{n − n_1 − n_2 - \ldots - n_{k - 1}}^{n_k} = С_{n_k}^{n_k} = 1$ 

Это не удивительно, потому что после того как мы выбрали позиции для всех типов объектов, кроме последнего, для последнего уже остается один способ для расположения объектов, других нет.

**Итоговое количество:**

Тогда итоговое количество выражается по правилу умножения, как произведение всех $C$:

$\large P(n_1, \ldots, n_k) = C_{n}^{n_1} \cdot C_{n - n_1}^{n_2} \cdot \ldots \cdot C_{n − n_1 - \ldots - n_{k - 1}}^{n_k} =$

Далее расписываем по формуле с факториалами:

$\large = \frac{n!}{n! \cdot (n - n_1)!} \cdot \frac{(n - n_1)!}{n_2! \cdot (n - n_1 - n_2)!} \cdot \ldots \cdot \frac{(n - n_1 - \ldots - n_{k-1})!}{n_k! \cdot (n - n_1 - \ldots - n_k)!} =$ 

Заметим, что у последнего выражения в знаменателе получаем $(n - n_1 - \ldots - n_k)! = 0!$

Дальше всё сокращается и мы получаем:

$\large = \frac{n!}{n_1! \cdot n_2! \cdot \ldots \cdot n_k!}$

И теорема доказана!

<h1 align="center">Задачи на полиномиальные коэффициенты</h1>

<h3 style="color:#2E8B57">Задача № 1</h3>

Имеется 5 букв `а`, 4 буквы `б` и 5 букв `в`. Сколько существует различных неповторяющихся двенадцатибуквенных слов, которые можно составить?

<h4 align='center'>Решение:</h4>

$P(5, 4, 3) = \frac{12!}{5!4!3!}$

<h3 style="color:#2E8B57">Задача № 2</h3>

Имеется $15$ различных шаров и $5$ различных ящиков. Сколько существует различных способов разложить шары по ящикам, чтобы в каждом ящике было ровно $3$ шара?

<h4 align='center'>Решение:</h4>

У нас $15$ различных шаров. И из этих $15$-ти шаров мы должны выбрать $3$ в первый ящик, $3$ во второй ящик, $3$ в третий, $3$ в четвертый и $3$ в пятый. 

Выбираем из $15$-ти $3$, естественно, это можно сделать:

$\large C_{15}^3$

Выбрали $3$ шара, осталось $12$ шаров в нашем распоряжении, из которых надо снова выбрать $3$ шара с целью положить их уже во второй ящик:

$\large C_{12}^3$

Дальше запишем все отсальное по правилу умножения:

$\large C_{15}^3  \cdot \large C_{12}^3 \cdot \large C_{9}^3 \cdot \large C_{6}^3 \cdot \large C_{3}^3 =$

Дальше всё можно записать в виде формулы и факториалы сокразаются, мы получаем:

$= \large \frac{15!}{3! \cdot 3! \cdot 3! \cdot 3! \cdot 3!} = P(3, 3, 3, 3, 3)$

То есть мы в этой задаче о расположении шаров в ящиках пришли к полиномиальному коэффициенту.

<h3 style="color:#2E8B57">Задача № 3</h3>

Имеется $11$ цветков разного цвета. Спрашивается, сколькими способами $11$ цветков можно раздать трем девочкам, так чтобы одной любой из трех девочек достались $5$ цветко, а двум другим по $3$? 

В данной задаче совмещаются полиномиальные коэффициенты и биномиальные.

<h4 align='center'>Решение:</h4>

**Пять цветков:**

Во-первых, нужно выбрать одну любую из трех девочку, которая получит $5$ цветков, это можно сделать:

$\large C_3^1$

После того как мы выбрали эту девочку, вслед за этим, — это правило умножения, — мы должны распределить цветы. Мы уже знаем, что этой выбранной нами девочке мы должны дать $5$ цветков. А цветков всего $11$, тогда:

$\large C_3^1 \cdot C_{11}^5$

**Три цветка второй девочке:**

После того как $5$ цветков выбраны, остается $6$ цветков, из которых $3$ мы должны отдать второй девочке:

$\large C_3^1 \cdot C_{11}^5 \cdot C_6^3$

**Три цветка третьей девочке:**

Количество способов отдать последние $3$ цветка последней девочке:

$\large C_3^1 \cdot C_{11}^5 \cdot C_6^3 \cdot C_3^3 = $

Перепишем в виде формулы:

$\large = 3 \cdot \frac{11!}{5! \cdot 6!} \cdot \frac{6!}{3! \cdot 3!} \cdot \frac{3!}{3! \cdot 0!} = 3 \cdot \frac{11!}{5! \cdot 3! \cdot 3!}$

Итоговый ответ будет:

$\large 3 \cdot P(5, 3, 3)$

<h1 align="center">Полиноминальная формула</h1>

<h1 style="color:#008B8B">Связь полиномиальных и биномиальных коэффициентов</h1>

Возьмете $P$ всего лишь от двух типов объектов:

$\large P(k, n - k)$

Сверху в числителе стоит сумма количества объектов с факториалом, то есть $n!$, а в знаменателе у получается произведение количества объектов с факториалом:

$\large P(k, n - k) = \frac{n!}{k!(n-k)!} = C_n^k$

В данном частном случае, когда типов объектов всего $2$ эта буковка $P$ совпадает с буковкой $C$. То есть, оказывается, что это просто биномиальный коэффициент.

**Полиномиальные коэффициенты:**

Дело в том, что числа $P(n_1, \ldots, n_k)$, служат очень естественным обобщением именно для биномиальных коэффициентов.

Даже вот не для чисел сочетания, а для выражения $C_n^K$, когда оно называется биномиальным коэффициентом.

$P(n_1, \ldots, n_k)$ - Полиномиальные коэффициенты.

**Вспомним что такое бином**

Бином — это вот выражение: $(x + y)^n$. Почему оно называется «бином»? Потому что здесь два слагаемых. Вот эта приставочка «би» в слове «ном» – она указывает на то, что слагаемых два. И коэффициенты биномиальные, потому что они находятся при разложении бинома, такого выражения из двух слагаемых (двухчлен). 

**Что такое полином?**

«Поли» по-гречески это «много». То есть вместо двух слагаемых,  нам предстоит рассматривать много слагаемых. Не двучлены, а многочлены. 

$(x_1 + \ldots + x_k)^n$

Причём $k$ — это количество типов объектов. В данном случае это количество слагаемых. 

Действительно при возведении в $n$-ную степень такого выражения получается некая сумма, в которой коэффициентами служат количество типов объектов $P(n_1,\ldots, n_k)$. Именно поэтому  они и называются полиномиальными коэффициентами. И поэтому они обозначены буквой $P$, от слова «полином», полиномиальный коэффициент.

Почему коэффициентами в этом разложении служат именно наши числа $P(n_1,\ldots, n_k)$ мы рассмотрим дальше.

### Пример 1

Пример в случае $k=3$ и $n=2$:

$\large (x_1 + x_2 + x_3)^2 = (x_1 + x_2 + x_3) \cdot (x_1 + x_2 + x_3) = $

Дальше остаётся расскрыть скобки

$\large = x_1^2 + x_1 x_2 + x_1 x_3 + x_2 x_1 + x_2^2 + x_2 x_3 + x_3 x_1 + x_3 x_2 + x_3^2$

Приводим подобные 

$\large = x_1^2 + x_2^2 + x_3^2 + 2 x_1 x_2 + 2 x_1 x_3 + 2 x_2 x_3$

**Интерпретация коэффициентов:**

Но причём здесь полиномиальные коэффициенты? Возьмём первое слогаемое $x_1^2$ у него коэффициент равен одному или в терминах полинома $P(2, 0, 0) = \frac{2!}{2!0!0!} = 1$. Следующий коэффициент при $x_2^2$ считается следующим образом $P(0, 2, 0)$, при $x_3^2$ коэффицинт считается как $P(0, 0, 2)$.

А что обстоить с коэффициентом два при $2 x_1 x_2$, его можно получить следующим образом $P(1, 1, 0) = \frac{2!}{1!1!0!} = 2$. Слдущее слогаемое $2 x_1 x_3$ и коэффициент можно найти как $P(1, 1, 0)$

**Но почему так?**

Можем заметить любопытную закнономерность, у первого выражения $x_1^2$ коэффициент при нём считается как P(2 - степень x_1, 0 - степень x_2, 0 - степень x_3), соответственно, $x_1$ и $x_2$ у нас нет, это тоже самое что возвести их в нулевую степень. Аналогично с другими коэффициентами.
