<h3 style="color:#2E8B57">Задача № 1</h3>

Имеется $10$ сосудов. Среди них ровно $5$ стаканов, мы считаем эти стааны неразличимыми и $5$ различны чашек. Cколько есть способов выбрать из $10$ сосудов $5$ штук.

<h4 align='center'>Решение:</h4>

Давайте перечислим все случаи:

1) Сначала можно взять 5 стаканов - Это можно сделать 1 способом $C_5^0$

2) Можно выбрать 4 стакана можно выбрать 1 способом, так как они не различимы. Осталось выбрать ещё одну чашку, чтобы получить 5 ёмкостей. Чашку можно выбрать 5 способомами $C_5^1$

3) 3 стакана + 2 чашки = $C_5^2$

4) 2 стакана + 3 чашкм = $C_5^3$

4) 2 стакана + 3 чашкм = $C_5^3$

И так далее... 

После нам необходимо всё сложить и мы получим:

$\large C_5^0 + C_5^1 + \ldots + C_5^5 = 2^5 = 32$

<h3 style="color:#2E8B57">Задача № 2</h3>

У нас есть множество, состоящее из шести элементов $\{а_1, а_2, а_3, а_4, а_5, а_6\}$. Спрашивается, сколькими способами можно выбрать из этого множества некоторое подмножество, которое состояло бы из четного числа элементов?

Пустое подмножество - это тоже подмножество из чётного числа элементов.

<h4 align='center'>Решение:</h4>

$\large C_6^0 + C_6^2 + C_6^4 + C_6^6 = 2^5 = 32$

Запишем это более коротко для большего количества элементов. Что значит выбрать подмножество отсюда, состоящее из скольких-то, пусть из четного числа элементов? 

**1) Возьмем элемент $а_1$ и либо положим его в строящееся множество, либо не положим.**

**2) Возьмем элемент $а_2$ и либо положим его в строящееся множество, либо не положим.** 

**И так далее, внимание, не вплоть, до элемента с индексом $6$, а вплоть до элемента с индексом $5$.**

**5) Возьмем элемент $а_5$ и либо положим его в строящееся множество, либо не положим.**

В зависимости от того, какие именно элементы мы включим в него, а что не включим, оно может получиться как четной мощности, то есть как содержащим четное число объектов, так и содержащим нечетное число объектов. Например, мы вообще ни один объект могли не включить. А могло быть по-другому. Мы положили $a_1$, а дальше ничего не клали. Получилось, множество нечетной мощности. 

Если получилась чётная мощность, то последний элемент $a_6$ не кладём, иначе $a_6$ кладём.

И таких ситуаций, в нашем случае $2^5$ степени.

<h3 style="color:#2E8B57">Задача № 3 на применение тождеств для вычисления суммы биномиальных коэффициентов</h3>

Давайте сосчитаем такую сумму чисел сочетания, или биномиальных коэффициентов:

$\large C_n^1 + 2C_n^2 + 3C_n^3 + \ldots + nC_n^n = x$

<h4 align='center'>Решение:</h4>

Вспомним первое тождество, так как наша сумма имеет симметрию 

$\large C_n^k = C_n^{n-k}$

Давайте обозначим интересующую нас величину $x$ и каждую $C$-шку, которая присутствует в этом x, заменим согласно указанному тождеству. 

$x = C_n^{n-1} + 2C_n^{n-2} + 3C_n^{n-3} + \ldots + nC_n^{0}$

Ну вот переписали просто согласно указанному тождеству.  А теперь давайте возьмем и прибавим к первой записи $x$ вторую запись $x$. То есть сложим первое равенство и второе равенство. 

В верхней строчке есть ли где-нибудь $C$-шка, которая бралась бы по $0$? Конечно же нет. Ну значит, единственная $C$-шка, которая берется по $0$, коль скоро мы складываем нижнюю сумму с верхней суммой. Ну, давайте ее напишем: 

$\large 2x = nC_n^0$

Теперь давайте посмотрим: хорошо, а $C_n^1$? Вверху $C_n^1$ присутсвует с коэффициентом $1$, а внизу $(n-1)C_n^1$. Если их сложить, то получим:

$\large 2x = nC_n^0 + nC_n^1$

Теперь давайте посмотрим: хорошо, а $C_n^2$? Вверху $2C_n^2$ присутсвует с коэффициентом $2$, а внизу $(n-2)C_n^2$. Если их сложить, то получим:

$\large 2x = nC_n^0 + nC_n^1 + nC_n^2$

И так далее до $C_n^n$, в нижнем слогаемом нет таких, но зато вверху имеется слогаемое $C_n^n$ с коэффициентом $n$:

$\large 2x = nC_n^0 + nC_n^1 + nC_n^2 + \ldots + nC_n^n = n (C_n^0 + \ldots + C_n^n)$

Но мы знаем, что $(C_n^0 + \ldots + C_n^n)$ равняется $2^n$, итого получаем:

$\large 2x = n (C_n^0 + \ldots + C_n^n) = n2^n$

Тогда $x$ это:

$\large x = n2^{n-1}$

<h3 style="color:#2E8B57">Задача № 4 на сумму чисел треугольника Паскаля, расположенных на одной диагонали</h3>

Обозначим буквой $C$ произвольное число треугольника Паскаля. Сосчитаем сумму чисел в треугольнике Паскаля, которые стоят на прямой, справа вверх идущей от вот этого числа $C$, включая число $C$. 

Утверждается, – что сумма, которая получится в итоге, равна числу, которое, наоборот, стоит снизу вправо от числа $C$.

<h4 align='center'>Решение:</h4>

Для наглаядности можно нарисровать треугольник поскаля в виде биномиальных коэффициентов. Где у $C$-шки нижний коэффицент - это номер её строчки, а верхний - позиция элемента в этой строчке.

То есть вот это число $C$ – это, на самом деле, просто какое-то $C_n^k$. И когда мы двигаеся по данной диагонали, то просто меняем номер строки, не меняя верхнего коэфициента, который является позицией в этой строке:

$\large C_n^k + C{n-1}^k + \ldots + C_k^k$

Так вот утверждается, что вся эта сумма равна числу, которое стоит вправо вниз. И как оно выражается? Понятно, что если $C$-шка была на n-ной строчка треугольника Паскаля, то в следующей строке номер будет $(n + 1)$. То есть это $C_n + 1$, и осталось понять по кому? Ну и понятно, что это будет $C_n + 1^{k+1}$ так мы сместились на одну диагональ вниз.

$\large C_n^k + C{n-1}^k + \ldots + C_k^k = C_{n+1}^{k + 1}$

**Доказательство**

$\large C_{n+1}^{k + 1}$ 

Если применить стандартное тождество оно в лекции за номером $2$, это тождество, которое касается, собственно, треугольника Паскаля. Это получится так:

$\large C_{n+1}^{k + 1} = C_n^{k+1} + C_n^k$ 

Теперь $C_n^k$ запишем в начале, а к $C_n^{k+1}$ опять применим тождество номер $2$:

$\large C_n^{k+1} + C_n^k = C_n^k + C_{n - 1}^k + C_{n - 1}^{k + 1}$ 

И теперь развернём $C_{n - 1}^{k + 1}$:

$\large C_n^k + C_{n - 1}^k + C_{n - 1}^{k + 1} = C_n^k + C_{n - 1}^k + C_{n - 2}^k + C_{n - 2}^{k + 1}$ 

Ну и видно, что мы так каждый раз будем понижать $n$ на очередную единичку, пока не дойдём до $C_k^K$:

$\large C_n^k + C_{n-1}^k + \ldots + C_k^k$

<h3 style="color:#2E8B57">Задача № 1</h3>

На дереве висит $10$ разных яблок. Сколькими способами можно сорвать нечётное количество яблок?

<h4 align='center'>Решение:</h4>

Из задачи Наборы из чётного числа символов мы знаем, что чётное количество яблок можно сорвать $2^{10-1} = 2^9 = 512$ способами. Так как общее количество способов сорвать яблоки равно $2^{10}$, то нечётное количество яблок можно сорвать также $512$ способами.

__Ответ:__ $512$

<h3 style="color:#2E8B57">Задача № 2</h3>

Сумма $C_{10}^1+C_{10}^2+\ldots+C_{10}^{10}$ равна?

<h4 align='center'>Решение:</h4>

Мы знаем, что $C_{n}^0 + C_n^1+ \ldots +C_n^n = 2^n$. Подставляя $n=10$, получаем, что $C_{10}^0 + C_{10}^1+C_{10}^2+\ldots+C_{10}^{10} = 2^{10}=1024$. Наша сумма получается из данной вычитанием $C_{10}^0 = 1$. Следовательно, ответ равен $1023$.

__Ответ:__ $1023$

<h3 style="color:#2E8B57">Задача № 3</h3>

Коэффициент при $x^7$ в разложении $(1+x)^{11}$ равен?

<h4 align='center'>Решение:</h4>

По формуле бинома Ньютона коэффициент при $x^7$ равен $C_{11}^7 = C_{11}^4$.

__Ответ:__ $C_{11}^4$

<h3 style="color:#2E8B57">Задача № 4</h3>

В наборе из $12$ сосудов имеется $5$ неразличимых стаканов и $7$ различных чашек. Сколькими способами можно выбрать $6$ сосудов из $12$?

<h4 align='center'>Решение:</h4>

Для каждого фиксированного $k$ существует только один способ выбрать $k$ неразличимых стаканов. Отсюда искомое количество способов равно количеству способов выбрать от $1$ до $6$ чашек. Искомая сумма равна 

$\large C_{7}^1+ \ldots + C_{7}^6 =C_{7}^0+ C_{7}^1+ \ldots + C_{7}^6 + C_7^7 - (C_7^0+C_7^7) = 2^{7} - 2 =128-2=126$

__Ответ:__ $126$

<h3 style="color:#2E8B57">Задача № 5</h3>

Сумма $C_{n+m-1}^m+C_{n+m-2}^m+\ldots + C_{m}^m$ при всех $m \ge 1$, $n \ge 1n$ равна:

<h4 align='center'>Решение:</h4>

Эта в сумма в точности равна сумме чисел в треугольнике Паскаля, расположенных на одной диагонали, начиная с числа $C_{n+m-1}^m$ и выше. Эта задача разобрана на видео, и ответ - число, стоящее под $C_{n+m-1}^m$ справа, то есть $C_{n+m}^{m+1} = C_{n+m}^{n-1}$.

__Ответ:__ $C_{n+m}^{n-1}$
