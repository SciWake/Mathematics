<h1 align="center">Сочетания с повторениями и без</h1>

<h1 style="color:#008B8B">Теорема 1. Сочетания без повторений</h1>

$\large C_n^k = \frac{A_n^k}{k!} = \frac{n!}{k!(n-k)!}$

Исходя из формулы, $k$-сочетаний без повторений в $k$ факториал раз меньше, чем $k$-размещений без повторений.

## Доказательство:

Почему сочетаний без повторений, ровно в $k$-факториал раз меньше, чем размещений без повторений?

**Размещения без повторений:**

Мы вытащили какую-то последовательность объектов, в которой учитывается порядок встречаемости объектов. Такую последовательность можно вытащить столькими способами, сколько у нас всего есть размещений без повторений. И это количество есть $A_n^K$. 

Вытащили последовательность: `л, я, г. у, ш, к, а`. Всего $7$ букв. В этом примере $k$ равняется $7$. А могли вытащить последовательность `г, у, л, я, ш, к, а`. C точки зрения размещений без повторений − это, конечно, другая последовательность, потому что буквы в ней встречаются в ином порядке. Действительно, получается просто другое слово. 

**Сочетания без повторений:**

Если нас не интересует порядок букв, порядок объектов в последовательности, которую вы извлекаем, а интересует лишь кучка, лишь пригоршня, состоящая из вот этих вот $7$, в данном случае, букв, то первый и второй набор букв - это тоже самое сочетание без повторений. То есть это разные размещения, но это одно и тоже сочетание.

То есть, каждое из них является одним и тем же сочетанием. Но естественно сочетаний получается во столько раз меньше, сколько вот этих вот слов можно записать переставляя буквы местами. Но, а сколько можно таких слов записать, например, вот в этом конкретном случае? Если у нас всего $7$ букв и мы их можем между собой переставлять как угодно, в итоге всякий раз, получая одну и ту же кучку из букв?

Есть $7$ факториал способов переставить буквы слова "лягушка". Ну понятно, значит в $7$ факториал раз больше у нас получается отдельных слов - размещений без повторений, нежели есть сочетаний без повторений, каждое из которых это просто кучка букв любого слова.

## Доказательство - Общий случай:

У нас есть множество, состоящее из $n$ объектов $\{a_1, \ldots , a_n \}$. Из этого множества мы извлекает всевозможные $k$ сочетания без повторений. Берем какое-нибудь из этих $k$ сочетаний и обозначим его в виде круга с номером в центре. 

Для формальности, можно около этого круга написать $\{a_{i_1}, a_{i_2}, \ldots, a_{i_k} \}$. Подразумевая, что $k$ сочетание без повторений состоит из объектов с какими-то различными индексами $i_1, \ldots, i_k$, и эти объекты, все изначально принадлежащие исходному множеству $a_1, \ldots, a_n$.

<img src="img/Teorema_1.svg" width="1200" height="1000">

**Сколько всего сочетаний?**

Сколько всего различных $k$ сочетаний без повторений? Ну допустим мы даже не знаем формулу, но мы же точно знаем обозначение их $С$ из $n$ по $k$, просто по определению. 

**Количество перестановок в одном сочетании:**

Зафиксируем какое-нибудь сочетание (круг). В этом сочетании есть объекты $a$ с какими-то индексами, и начнем эти объекты между собой переставлять. То есть будем записывать эти последовательности в разном порядке. Таких перестановок на множестве из этих $k$ различных объектов - $k$ факториал штук. Значит, у каждого сочетания, есть ровно $k!$ перестановок.

**Общее количество размещений**

А сколько всего размещений? В первом сочетании $k$ факториал размещений, во втором $k$ факториал, в третьем $k$ факториал, в $C_n^k$-той $k$ факториал размещений. А суммарно сколько? Ну суммарно мы получим просто все возможные размещения, в точности $A_n^k$ – количество всех возможных размещений без повторений из $n$ объектов. Но сколько всего этих $k$ факториалов? А столько, сколько у нас всего сочетаний, то есть $C_n^k$ штук. 

Если $C_n^k$ – количество сочетаний, умножить на $k$ факториал, то есть сложить вот эти все $k$ факториалы по всем $C_n^k$ сочетаним, то у нас получится в аккурат $A_n^k$ – количество всех возможных размещений:

$\large C_n^k \cdot k = A_n^k$

Мы просто каждое размещение определили так: взяли некоторое $k$ сочетание и по нему построили все возможные размещения, которые из него получаются. Естественно, мы исчерпали таким образом множество всех возможных размещений. Ну и в итоге мы получаем ровно ту самую формулу, которая и была заявлена в формулировке теоремы. Ну, а это соответственно будет:

$\large C_n^k = \frac{A_n^k}{k!} = \frac{n!}{k!(n-k)!}$

И теорема, тем самым, полностью доказана. 

Взяли каждое сочетание, посмотрели все размещения, которые ему соответствуют. Сочетаний суммарно $C_n^k$, размещений стало быть в $k$ факториал раз больше. А сколько всего размещений? $A_n^k$. Ну значит, $C_n^k$ — это $A_n^k$ поделить на $k$ факториал.

Теорема полностью доказана.

<h1 align="center">Задачи на cочетания без повторений</h1>

<h3 style="color:#2E8B57">Задача № 1</h3>

В медецинской лаборатории имеется $10$ мышей (все они разные). Для эксперементка надо выбрать $5$ мышей. Сколько есть способов выбрать $5$ мышей для эксперемента?

<h4 align='center'>Решение:</h4>

Поскольку в вопросе ничего не сказано о деталях про длину хвоста, красноту носа и глаз, и так далее, то предполагается, что нам нужно вытащить из клетки, где сидят $10$ мышей, пяток из них пригоршней, — такой сачок туда запустить и этим сачком $5$ мышей изловить.

$\large C_{10}^5 = 252$

<h3 style="color:#2E8B57">Задача № 2</h3>

Есть двое друзей, которые коллекционируют юбилейные монеты. У первого друга $9$ юбилейных монет, у второго $7$. Считаем, что все $16$ монет, которые в сумме имеются у товарищей, они разные. Ну раз они разные, тогда имеет смысл вопрос об обмене: товарищи могут попробовать поменяться друг с дружкой монетами. Спрашивается, сколькими способами можно обменять какие-то $2$ монеты первого из друзей на какие-то $2$ монеты второго? 

<h4 align='center'>Решение:</h4>

То есть вопрос состоит только в том, сколькими способами можно выбрать $2$ монеты из девяти и, соответственно, сколькими способами можно выбрать $2$ монеты из семи.

$\large C_{9}^2 \cdot C_{7}^2$

<h3 style="color:#2E8B57">Задача № 3</h3>

Есть $28$ костей домино. Спрашивается: «Сколькими способами можно выдать из этих $28$ костей $7$ некоему игроку таким образом, чтобы среди этих $7$ костей присутствовал хотя бы один дубль?» Дубль — это $0$ — $0$, $1$ — $1$ и так далее. $6$ — $6$. 

<h4 align='center'>Решение:</h4>

Проще, посчитать общее количество способов, выдать какому-либо игроку $7$ костей, после чего из этого общего количества вычесть те ситуации, когда игроку ни один дубль не выдан.

Давайте для начала действительно ответим на вопрос «А сколько есть различных способов из этих $28$ костей выбрать $7$?». Дальше порядок следования костей внутри вот этой выдачи — он для нас ведь никакого значения не имеет. Мы опять же просто пригоршнями зачерпываем $7$ костей из общего котла, в котором варятся $28$ косточек. И получаем естественно сочетание без повторений.

$\large C_{28}^7 - C_{21}^7$

<h3 style="color:#2E8B57">Задача № 4</h3>

В тренировочной группе $5$ новичков и $3$ опытных альпиниста. В поход идут $6$ человек, причем предполагается, что среди них обязательно должно быть хотя бы двое опытных. Cколькими способами можно составить такую группу для похода?
 
<h4 align='center'>Решение:</h4>

Что значит в поход идут $6$ человек, среди которых не менее двух опытных? Это значит, что либо ровно двое опытных, либо ровно трое опытных, больше никаких вариантов быть не может. Если мы расссмотрим сначала случай, когда в поход идут двое опытных, а затем, когда трое опытных, то посчитав для каждого из этих случаев количество способов составить соответствующую группу, мы должны будем, естественно, сложить эти два количества.

Два опытных:

$\large C_{3}^2 \cdot C_{5}^4$

Три опытных:

$\large C_{3}^3 \cdot C_{5}^3$

Число способов составить такую команду

$\large C_{3}^2 \cdot C_{5}^4 + C_{3}^3 \cdot C_{5}^3$

<h2 style="color:#008B8B">Теорема 1. Число сочетаний с повторениями</h2>

### $\bar{C}_n^k = C_{n + k - 1}^k$

Оказывается, что число сочетаний без повторений, оно помогает посчитать число сочетаний с повторениями.

<h3 style="color:#2E8B57" align="right">Задача на применение сочетания с повторениями</h3>

В магазине есть $4$ сорта пирожных и нам нужно купить $20$ штук.

**Решение:**

Нам только важно, сколько из этих $20$ первого сорта, сколько из этих $20$ второго сорта, сколько третьего, сколько четвертого, а порядок, повторяю, никакого значения не имеет, потому что речь не идет о том, в каком порядке их съесть. Нам нужно из этих четырех объектов составить кучку, в которой будет всего $20$ объектов. 

$\bar{C}_{4}^{20} = C_{23}^{20} = \frac{23!}{20!3!} = \frac{23 \cdot 22 \cdot 21}{6} = 23 \cdot 77$

### Доказательство c примером:

**Процесс кодирование сочитания:**

Давайте в качестве множества объектов исходного возьмем любимые буквы русского алфавита. Давайте подумаем, что такое сочетание с повторениями, для примера рассмотрим $4$-сочетания с повторениями. Вместо слова «лягушка», в которой все буквы были разные, теперь рассмотрим слово «жаба», в котором есть совпадающие буквы «а». {ж, а, б, а} — это вполне себе $4$-сочетания с повторениями. При этом «абаж», «бажа» — это все то же самое сочетание, что {а, б, а, ж}. Важен состав, что $2$ буквы «а», $1$ буква «б» и $1$ буква «ж», а в каком порядке они не имеет значение.

Идея такова, давайте закодируем наше сочетание, сопоставим ему некий номер. Действуем следующим образом: двигаемся по вот этому исходному множеству объектов от начала до конца, они упорядочены у нас, это буквы русского алфавита. Давайте возьмем первый объект, и в данном случае это, конечно, «а», первая буква русского алфавита — это «а». Давайте посчитаем, сколько раз эта буква встречается в нашем $4$-сочетании с повторениями. Сколько раз встречается буква «а»? Ведь нам же важен не порядок, нам важно количество встречаемости. Вот мы и посчитаем, сколько раз буква «а» встречается вот в этом множестве. Ну ясно, $2$ раза. Давайте нарисуем в честь этого две еденицы:

## $11$

Вот он начинается с двух единиц, потому что у этого конкретного $4$-сочетания с повторениями ровно $2$ раза встречается буква «а», и всякий раз, когда она встречается, мы как бы рисуем очередную единицу в номер будущего паспорта. Все, с буквой «а» разобрались. Двигаемся дальше. Внутри изначального множества берем букву «б» и смотрим, сколько раз она встречается в нашем $4$-сочетании с повторениями. Но прежде чем посмотреть это, ставим $0$. $0$ — это такая перегородка, которая отделяет количество встречаемости буквы «а» в нашем $4$-сочетании от количества встречаемости буквы «б». Ну «б» встречается $1$ раз, в честь этого рисуем только одну единицу, потому что она встречается, буква «б» встречается в баране только $1$ раз.

## $1101$

Так, снова ставим нолик как перегородку между вот этими вот кодируемыми количествами встречаемости, сколько раз встречается очередная буква. Смотрим сюда, на русский алфавит, буква «в». Но буква «в» вообще-то в слове «жаба» не встречается, поэтому я не могу нарисовать ни одной единицы, а вот перегородку я все равно поставлю. 

## $110100$

Ну, дальше будет занудство, потому что мы двигаемся дальше по алфавиту, смотрим букву «г», и она — опять не встречается. Ну значит опять единицу нарисовать не можем, а рисуем $0$: перегородку каждый раз надо рисовать. Буква «д» не встречается — единицу не рисуем, $0$ рисуем. Так, следующая у нас буква «е». Она не встречается, поэтому единицы нет, $0$ есть. Следующая буква «ё». То же самое: единицы нет, $0$ есть. Ой, какое счастье, появилась буква «ж», которая есть в нашем слове, поэтому мы рисуем одну единицу, указывающую на то, что «ж» встречается вот в этом наборе букв ровно $1$ раз.

## $11010000001$

Дальше снова рисуем одни сплошные перегородки. То есть формально мы обязаны проверить, встречается ли буква «з» здесь, встречается ли буква «и», «й» и так далее вплоть до «о», «п», «р», «с», «т» и чего хотите. Вот, но всякий раз придется рисовать нули. Вот смотрите: буква «ю», это предпоследняя буква русского алфавита. Естественно, в наборе букв вот этого слова «жаба» она не встречается, поэтому опять не рисуем ни одной единицы, а перегородку перед количеством встречаемости буквы «я» все-таки ставим, то есть вот этот $0$, он отгораживает несуществующее количество букв «ю» от несуществующего количества букв «я». Внимание, вот очень важный момент: здесь ничего не рисуем, потому что «я» не встречается в слове «жаба» никак, но больше, внимание, нулей не ставим, так как отгораживать «я» еще от чего-то не нужно.

## $1101000000100 \ldots 0$


Значит, давайте осознаем все-таки, сколько в этом номере паспорта единиц и сколько нулей, для того чтобы полостью понять конструкцию. 

**Сколько едениц у кода?**

Единиц, очевидно, столько, сколько есть элементов в этом сочетании, то есть $4$, просто по определению.

**Сколько нулей в этой последовательности?**

Мы каждые две соседние буквы разделяем ровно одним нулем. Вот здесь в конце никакого нуля нет. Сколько стало быть нулей? Если бы у нас было всего $2$ буквы в русском алфавите, то нулей был бы один, если бы у нас было $3$ буквы в русском алфавите, то нулей было бы $2$. Поскольку у нас $33$ буквы в русском алфавите, то нулей $32$, на единицу меньше, чем число букв.

А теперь видим: n у нас равняется $33$ в этом примере, — $33$ буквы русского алфавита и из них извлекаются $4$ сочетания, — $k$ равняется $4$. Суммарное количество единиц и нулей — это 36. В номере паспорта 36 всего элементов: $4$ единицы и $32$ нуля, всего $36$. Но это и есть $(33 + 4 – 1)$ — то самое, что было $(n + k – 1)$ в основании $C$. Ну это совершенно не удивительно, потому что нам будет нужно потом посчитать количество номеров паспорта, и вот это количество номеров паспорта — это и есть количество наших $4$-сочетаний.

### Доказательство в общем виде:

У нас есть просто какой-то набор объектов ${a1,..., an}$. 

И вот из этих объектов мы составляем всевозможные $k$-сочетания с повторениями. Каждому из этих $k$-сочетаний с повторениями мы сопоставляем паспорт с номером, который представляет собой последовательность из единиц и нулей. Мы просто рисуем в начале столько единиц, сколько в данном конкретном $k$-сочетании раз встречается $a1$. И не рисуем единицы вовсе, если $a1$ в данном конкретном $k$-сочетании, которое мы пытаемся закодировать, не встречается. Дальше рисуем перегородку в виде $0$, отделяющего встречаемость символа $a1$ в нашем сочетании от встречаемости символа $a2$, и рисуем столько единиц, сколько раз встречается $a2$ и так далее до элемента $a_n$.

Взяли $k$-сочетание с повторениями, посмотрели, сколько раз $a_1$, сколько раз $a_2$ и так далее, сколько раз $a_n$, нарисовали соответствующие числа единиц, количества единиц, а эти количества единиц разделили перегородками-нулями. Вот получился такой паспорт.

**Количество едениц:**

Каждому $k$-сочетанию с повторениями отвечает уникальный паспорт — такая вот последовательность из единиц и нулей, в которой, внимание, сколько единиц в общем случае? Ну, понятно сколько. Мы посмотрели, сколько раз встречается одна буква, посмотрели, сколько раз встречается другая, посмотрели, сколько раз встречается последняя, — в сумме, конечно, получится просто $k$. Сколько всего есть объектов внутри нашего $k$-сочетания? Ну их, по определению, $k$ штук. То есть, действительно, единиц и в общем случае тоже $k$ штук.

**Количество нулей:**

Перегородки возникают между соседними буквами. Если бы объектов было $2$, перегородок было бы $1$, если объектов было бы $3$, перегородок $2$. Но объектов $n$. Если этих объектов $n$ штук, то перегорок $n – 1$, естественно. Поэтому нулей в этой последовательности $n – 1$ штука.


Тогда, величина $\bar{C}_n^k$, количество $k$-сочетаний с повторениями из наших $n$ объектов, в точности равна количеству, количеству всех различных последовательностей, номеров паспорта из $0$ и $1$, в каждой из которых ровно $k$ единиц и ровно $n – 1$ нулей. Длина всей последовательности — это сколько? Ну, если в ней $k$ единиц и $n – 1$ нулей, значит, это $(n + k – 1)$.

И вот нас интересует, сколько всего бывает последовательностей из нулей и единиц, которые состоят всего из $(n + k – 1)$ этих самых нулей и единиц, причем мы в точности знаем, что единиц, например, $k$ штук среди вот этих $(n + k – 1)$. Ну давайте, стало быть, посчитаем количество этих последовательностей, тогда мы найдем и $\bar{C}_n^k$.

Давайте посчитаем количество последовательностей. Вот что значит нарисовать такую последовательность из нулей и единиц, длина которой вся вот целиком $(n + k – 1)$ и в которой ровно $k$ единиц, вот что это в точности означает?  

Вот что на самом деле характеризует эту последовательность? Внимание, эту последовательность характеризует множество тех позиций, на которых стоят единицы. Если мы выбрали $k$ позиций среди наших $(n + k – 1)$, на которые мы захотим потом поставить единицы, то все, последовательность задана однозначно. Нам лишь надо из $(n + k – 1)$ позиций выбрать некоторые $k$ позиции, на которые мы поставим единицы, тогда на все остальные $n – 1$ позиции мы, естественно, поставим нули. И, заметьте, порядок-то единиц неважен: мы выбрали $k$ позиций, — какая разница, какую единицу куда поставить.Важно лишь, какие $k$ позиции вот из этих $(n + k – 1)$ возможных вы займете единицами. Никакие порядки, никакие размещения здесь не важны.

Количество последовательностей, равно количеству способов выбрать $k$ позиций из $(n + k – 1)$ возможных позиций для простановки на эти $k$ позиции единиц. Позиция — это объект нашего изучения. Вот есть $(n + k – 1)$ объектов, каждый из которых называется красивым словом «позиция». Вот там есть позиция $b_1$, есть позиция $b_2$ и так далее, есть позиция $b$ с индексом $n + k – 1$. Вот всего столько есть позиций. И вот из этого множества объектов мы хотим выбрать $k$ штук, $k$ позиций. Но, внимание, как мы их выбираем? Понятно, что позиции разные, то есть повторений здесь быть не может. Те самые k позиций, которые мы собирается извлечь отсюда, они безусловно разные, потому что мы же не будем $2$ единицы ставить на одну позицию, то есть никаких повторений здесь нет.

У нас есть всего $(n + k – 1)$ объект, мы из этого множества, мощности, размера $(n + k – 1)$, выбираем $k$ объектов без учета порядка и при этом различные — конечно, это есть сочетания без повторений:

### $\bar{C}_n^k = C_{n + k - 1}^k$

Теорема полностью доказана.

<h3 style="color:#2E8B57" align="right">Задача на применение сочетания с повторениями</h3>

В продаже имеются розы трех цветов. При этом считается, что розы одного цвета между собой никак различаются. Cколькими способами можно составить букет из $7$ роз?

**Решение:**

У нас есть $3$ типа объектов ${a_1, a_2, a_3}$. $a_1$ — это объект, который называется «красная роза», $a_2$ — это объект, который называется «белая роза», $a_3$ — это роза третьего цвета какого-нибудь. И объектов первого типа у нас, в принципе, неограниченное количество, хоть $7$: мы можем все $7$ роз взять первого типа. Ну и то же самое касается объектов второго типа, то же самое касается объектов третьего типа.

Сколько же существует букетов из $7$ роз, которые можно составлять из таких объектов? 

$\bar{C}_3^7 = C_9^7$

<h3 style="color:#2E8B57">Задача № 1</h3>

В одном маленьком королевстве выбирают царя, его советника и пятерых стражников из $30$ кандидатов. Сколькими способами это можно сделать?

<h4 align='center'>Решение:</h4>

Должности царя, советника и стражника различны, поэтому важно кому из кандидатов какая должность достанется. Если царя можно выбрать $30$ способами, то стать его советником сможет уже один из $29$ претендентов. После чего останется выбрать $5$ стражников из оставшихся $28$ человек. Поскольку должности стражников между собой никак не отличаются, то речь идет о сочетании без повторений $C_{28}^5$.

В итоге получаем, что выбрать людей на искомые должности можно $30\cdot 29\cdot C_{28}^5$ способами.

<h3 style="color:#2E8B57">Задача № 2</h3>

У королевы есть 12 одинаковых зеркал. Сколькими способами их можно повесить в 8 разных залах замка так, чтобы в каждом зале было хотя бы одно зеркало?

<h4 align='center'>Решение:</h4>

Попробуем понять, что мы из чего выбираем. Так как в каждом зале висит хотя бы одно зеркало, то можно считать, что мы уже повесили в каждом зале по зеркалу, и остались $4$ зеркала, которые надо развесить произвольным образом по $8$ залам (тогда в зале может висеть от одного зеркала до пяти). Если занумеровать все залы, то получится, что нам нужно выбрать четыре числа (некоторые из них могут повторяться) из восьми возможных, причем порядок этих чисел не важен (сколько раз число встретилось в выбранном наборе, столько "дополнительных" зеркал в соответствующем зале). Это в точности количество сочетаний с повторениями $\overline{C}_{8}^4 = C_{11}^{4} = 330$.

<h3 style="color:#2E8B57">Задача № 3</h3>

В летнем лагере $10$ отрядов. Каждый день дежурят по $3$ отряда. Сколькими способами можно выбрать дежурные отряды? 

<h4 align='center'>Решение:</h4>

Для каждого дня нам надо выбрать $3$ отряда из $10$. Последовательность выбора не важна, следовательно, это сочетание. Отряды не повторяются, поэтому это сочетание без повторений. Количество способов выбрать $3$-сочетание без повторений из $10$ элементов равно $C_{10}^3 = \frac{10!}{3!\cdot7!}$.

<h3 style="color:#2E8B57">Задача № 4</h3>

В палитре художника имеется $10$ разных красок. Сколько разных оттенков он может получить, смешивая $3$ краски (краски могут быть одинаковыми, получаемый оттенок не зависит от порядка смешивания красок)?

<h4 align='center'>Решение:</h4>

Мы выбираем $3$ краски из $10$. Порядок роли не играет, следовательно, мы выбираем сочетание. Краски могут быть одинаковыми, следовательно, это сочетание с повторениями. Итоговый ответ:  $\overline{C}_{10}^3 = \frac{12!}{3! \cdot 9!}$.

<h3 style="color:#2E8B57">Задача № 5</h3>

Команда <<Турин>> состоит из $10$ баскетболистов, а команда <<Ювента>> -− из $12$. Сколькими способами команды могут сформировать стартовый состав (в баскетболе на поле играют $5$ человек)?

<h4 align='center'>Решение:</h4>

Для каждой команды мы выбираем $5$ баскетболистов из $10$ или $12$. Порядок выбора не важен, баскетболисты не повторяются, следовательно, это сочетание без повторений. Для <<Турина>> количество способов выбора равно $C_{10}^5$ для <<Ювенты>> − $C_{12}^5$. По правилу умножения получаем $C_{10}^5 \cdot C_{12}^5 = 199584$.

<h3 style="color:#2E8B57">Задача № 6</h3>

Сколькими способами в течение $5$ дней можно выбирать на дежурство по $4$ ученика из класса в $20$ человек так, чтобы каждый день состав дежурных был разным?

<h4 align='center'>Решение:</h4>

Каждый день мы выбираем $4$ учеников из $20$. Так как порядок выбора не важен, выбор без повторений, то это $4$-сочетание без повторений из $20$ элементов. Количество таких сочетаний равно $C_{20}^4$. Общее количество способов − это выбор дежурных на $5$ дней (порядок выбора важен, дежурные не повторяются) − $5$-размещение без повторений из $C_{20}^4$ элементов. Значит, искомое количество равно $A_{C_{20}^4}^5 = \frac{(C_{20}^4)!}{(C_{20}^4 - 5)!} = C_{20}^4\cdot (C_{20}^4 - 1)\cdot (C_{20}^4 - 2)\cdot (C_{20}^4 - 3)\cdot (C_{20}^4 - 4)$.

<h3 style="color:#2E8B57">Задача № 7</h3>

Имеется $10$ разных чашек и $5$ разных ложек. Количество способов разложить ложки по чашкам (в одну чашку влезает сколько угодно ложек) является по своей сути чем?

<h4 align='center'>Решение:</h4>

Для каждой ложки мы выбираем чашку. Ложки различные, значит порядок раскладывания важен. В одну чашку можно положить несколько ложек, следовательно, мы размещаем ложки по чашкам с повторением.

Для наглядности посчитаем количество этих способов. Для каждой ложки есть $10$ вариантов чашек, поэтому общее количество вариантов по правилу умножение равно $10^5$.

<h3 style="color:#2E8B57">Задача № 8</h3>

В первый класс набрали $75$ учеников. Сколькими способами можно распределить их по трем классам по $25$ человек в каждом? 

<h4 align='center'>Решение:</h4>

Можно организовать распределение следующим образом: сначала выбрать $25$ учеников из $75$ в первый класс, а потом из оставшихся $50$ выбрать $25$ во второй класс, тогда последние $25$ учеников автоматически попадут в третий класс. В каждом выборе порядок не важен и повторов нет, поэтому это сочетания без повторений: $C_{75}^{25}$, $C_{50}^{25}$. По правилу умножения получаем, что искомое количество способов равно $C_{75}^{25} \cdot C_{50}^{25}$.

Можно объяснить и по другому: занумеруем учеников от $1$ до $75$. Способов нумерации будет столько же, сколько перестановок из $75$ элементов, то есть $75!$. Если считать, что ученики с $1$ по $25$ образуют первый класс, с $26$ по $50$ − второй, с $51$ по $75$ − третий класс, то один и тот же способ распределения посчитан $(25!)^3$ раз. Поэтому итоговое количество способов равно $\frac{75!}{(25!)^3}$. Несложно проверить, что он совпадает с предыдущим ответом.
