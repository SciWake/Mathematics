<h2 style="color:black" align="center">Основные комбинаторные величины и их свойства</h2>

<h2 style="color:#008B8B">Неформальное обозначение</h2>


Ну эти комбинаторные величины, на самом деле, очень естественным образом связаны с организацией выбора объектов внутри какого-то множества. Нам дано какое-то множество объектов, естественно, конечное, которое, как обычно же, обозначается $A$ и состоит из элементов объектов **a1, ..., a** с индексом **n**.

Ну как можно, например, извлекать объекты из множества букв русского алфавита?

$A$ состоит из букв русского алфавита. Что можно с этими буквами делать? Ну можно представить себе, что каждая из этих букв нарисована на некоторой карточке. Эти все карточки там как-то перемешаны, заброшены в коробку. И вот мы запускаем пригоршней запускаем руки в эту коробку, в которой лежат наши карточки и извлекаем некоторое количество букв. 

**1. Взять и пригоршней зачерпнуть некоторое количество букв (k - число от 1 до 33).**

Представим себе, для примера, $k = 7$. Ну так вот захотелось нам вытащить $7$ букв. Вытащили пригоршней, и оказалось, что это буквы такие: л, я, г, у, ш, к, а. То есть вот этот вариант он не учитывает ни в коем случае порядка букв, не говорит ни о каких конкретных словах, а он говорит лишь о том, что мы зачерпнули пригоршней и получили кучку букв, из которых дальше еще можно что-то составлять. 

**2. Последовательно извлекаем буквы из коробки, буквы из коробки, покуда не извлечем их k штук. Получаем на выходе на выходе конкретное слово из k букв.**

Вот лежат те же самые карточки внутри той же самой коробки. Но мы теперь не пригоршней их зачерпываем, а начинаем их извлекать последовательно. Извлекли одну букву – положили перед собой. Извлекли вторую букву – положили вслед за первой. Извлекли третью букву – положили вслед за второй. И так далее. Так, покуда не извлечем $k$ букв, ну или скажем $7$ букв, если говорить о совершенно конкретном примере. 

Например, лягушка. Или, например, гуляшка. То есть, с точки зрения первого способа выбора, лягушка и гуляшка – суть одно и то же, это одна и та же кучка букв, но с точки зрения второго варианта выбора, лягушка и гуляшка – это разные вещи, это разные слова. 

**3. Зачерпнуть пригоршней k букв из божественной коробки.**

Лягушка, замечательным образом, состоит из разных букв русского алфавита, а вот любимый кролик – другой любимый зверь – он состоит из, вообще-то, иногда совпадающих букв. У него два раза повторяется, как минимум, буква «к». Но ни в рамках первого варианта, ни в рамках второго варианта мы слово «кролик» не получим.

А все дело в том, что в первом, и во втором подходе мы требовали действительно различия букв. То есть либо зачерпывали их пригоршней из коробки, в которой на карточках были написаны разные буквы, либо последовательно извлекали их из такой же коробки.

Ну так представьте себе теперь, что у вас в коробке лежит не 33 карточки, на каждой из которых написана своя отдельная буква русского алфавита. Представьте себе, что в вашей коробке лежит бесчисленное множество карточек. Такая божественная коробка, которая содержит бесчисленное множество карточек: 
* Есть бесчисленное множество карточек, на которых написана буква «а»; 
* бесчисленное множество карточек, на которых написана буква «б»;

И вот теперь из этой божественной коробки вы можете, опять же, двумя способами извлекать наборы букв. Поэтому возникает третий вариант.

**4. Последовательно друг за другом извлекать k букв из божественной коробки.**

<h2 style="color:#008B8B">Обозначение основных комбинаторных величин</h2>

Если у нас все-таки есть абстрактное множество объектов $A = \{a_1, \ldots , a_n\}$, то:

**1. k-сочетания без повторений** $C_n^k$

**2. k-размещения без повторений** $A_n^k$

**3. k-сочетания с повторениями** $\bar{C_n^k}$

**4. k-размещения с повторениями** $\bar{A_n^k}$

Но, интересно, а чему же они на самом деле равны, как их выразить? А для того чтобы из выразить, давайте введём такую функцию, которая в математике называется факториалом.

$0! = 1$, $1! = 1$ А в общем виде: $n! = (n - 1)! n$

То есть по большому счёту, если говорить совершенно простыми и понятными словами, n! — это просто произведение всех чисел от 1 до n. 


<h2 style="color:black" align="center">Размещения</h2>

<h2 style="color:#008B8B">Теорема 1. K-размещение с повторениями</h2>

### $\bar{A_n^k} = n^k$

**Доказательство:**

У нас есть некоторое множество объектов, которое состоит из n штук этих самых объектов $\{a_1, \ldots ,a_n\}$. И вот из этого множества мы извлекаем объекты с повторениями. То есть у нас имеется вот эта самая божественная коробка, в которой содержится бесконечно много объектов $a_1$, бесконечно много объектов $a_2$ и так далее, бесконечно много объектов $a_n$.

Получается, это задача на правила умножения. На первую позицию в будущем k-размещении с повторениями, мы можем выбрать любой из этих объектов. На вторую позицию в строящемся k-размещении мы тоже можем выбрать любой из наших $n$ объектов. И так далее. На k-тую позицию, мы тоже можем выбрать любой из имеющихся в нашем распоряжении объектов. 

Ну понятное дело, что мы количество таких вот номеров, количество таких размещений с повторениями и будем вычислять как $n \cdot n \cdot \ldots \cdot n$. И этих умножений будет ровно $k$ штук. То есть мы просто $n$ умножим на себя $k$ раз, потому что имеет место правило умножения. А это и есть $n$ в степени $k$.

<h2 style="color:#008B8B">Теорема 2. K-размещение без повторений</h2>

### $A_n^k = n(n-1) \cdot \ldots \cdot (n - k + 1)$

**Доказательство:**

1. `На первую позицию в k-размещении без повторений (видите, разница в том, что мы тепрерь работаем без повторов), значит, мы можем выбрать любой из объектов. Количество таких способов n;`

Мы на первой позиции уже какой-то объект зафиксировали, но размещение, которое мы хотим построить, оно ведь не допускает повторений. Поэтому когда мы будем ставить какой-то объект на вторую позицию, мы ни в коем случае не сможем использовать вот тот единственный объект, который уже стоит на первой позиции.

2. `На вторую позицию, на вторую позицию в строящемся k-размещении, размещении без повторений мы не можем взять тот объект, который уже стоит на первой позиции, и потому мы можем выбрать любой из оставшихся (n − 1) объектов.` 

И дальше, естественно, происходит абсолютно тоже самое: когда мы выбираем объект на третью позицию, мы уже помним, что мы зафиксировали некоторый объект на первой позиции, некоторый отличающийся от него объект – на второй позиции. Соответственно, на третью позицию у нас есть вариантов выбора сколько? Только, конечно же, $(n − 2)$. 

На первую позицию у нас есть $n$ вариантов выбора. На вторую позицию у нас уже есть $(n − 1)$. На третью позицию (три!) $(n − 2)$ варианта выбора; на четвертую (четыре!) – $(n − 3)$; на пятую – $(n − 4)$; на шестую – $(n − 5)$; на k-тую – разумеется, $(n − k + 1)$. 

**Правило умножения** дает нам, как следствие, ровно то, что было заявлено. Сначала у нас есть **n** вариантов выбора, вслед за ними $(n − 1)$, последовательность, по правилу умножения, можно составить $n \cdot (n − 1)$ способами, дальше $(n − 2)$ – это когда мы третий объект устанавливаем на свою позицию, и так вплоть до k-того объекта, который дает сомножитель $(n − k + 1)$.

### Дополнительно к теореме 2:

**Формат записи факториалом:**

Данное выражение можно записать следующим образом:

### $A_n^k = n(n-1) \cdot \ldots \cdot (n - k + 1) = \frac{n!}{(n - k)!}$

Ну, действительно, сверху стоит произведение всех чисел от $1$ до $n$, по определению факториала: $1 \cdot 2 \cdot , \ldots , \cdot n$, а снизу стоит произведение всех чисел от $1$ до $(n – k)$, то есть и здесь есть это произведение от $1$ до $(n – k)$, и здесь оно полностью содержится. Когда мы сократим, у нас ровно останутся вот эти вот сомножители, которые написаны слева. 

**Что такое A из n по 0?**

Что значит посчитать количество способов ничего не выбрать? Ведь $A_n^0$ — это 0-размещение. Что такое 0-размещение? Это мы ничего не размещаем, мы вообще не запускаем руку ни в какую коробку, мы просто ничего не делаем. Ну, сколькими способами можно ничего не сделать? Сколькими способами можно вытащить, так сказать, пустое множество объектов? Понятное дело, что есть только $1$ способ вытащить пустое множество объектов. 

### $A_n^0 = 1 = \frac{n!}{(n - 0)!}$

**Что такое A из n по n?**

Ну, что дает нам наша формула? Она дает нам, конечно, 

$A_n^n = n \cdot (n – 1) \cdot, \ldots, \cdot (n – n + 1)$

то есть здесь у нас стоит в точности $1$, то есть это получается произведение от $1$ до $n$, а это есть $n!$. И заметьте, если мы подставляем вот в эту формулу $k = n$, $(A_n^n = \frac{n!}{(n - n)!})$ то внизу оказывается $0!$. Но мы с вами договорились, $0!$ — это $1$. И тогда эта формула благополучно согласуется с тем, что у нас здесь написано. Именно ради этого, в особенности ради этого мы и договаривались о том, чтобы считать $0!$ равным $1$. 

$n!$, количество размещений, $n$ размещений без повторений в множестве из $n$ объектов — это на самом деле по существу что такое? Это число различных перестановок.
