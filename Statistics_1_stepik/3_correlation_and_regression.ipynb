В данном случае корреляция пложительная, так как с увеличением экспрессии одного гена, экспрессия гена № 2 тоже увеличивается.

Показатель, который отражает силу взаимосвзяи двух количественных переменных, называется **коэффициентом корреляции**.

<h1 style="color:#008B8B">Вывод формулы коэффициента корреляции</h1>

<img src="img/3_1_4.png" width="600" height="360">

Рассмотрим пример взаиммосвязи роста и веса, где

* Переменная вес - это x, 

* а переменная рос - это y.

Добавим на график две линии:

* Красная вертикальная линия соответствует среднему значению веса в выборке;

* Синяя горизонтальная линия соответсвтует среднему значению роста в выборке.

Из-за этого, вся диаграмма рассеивания будет разбита на 4 квадрата. Если большенство наблюдений лежит в верхнем правом и нижнем левом квадрате, то это говорит о положительной взаимосвязи. 

**Почему так?** Все наблюдения, которые лежат в верхем правом квадрате, от среднего значения роста и от среднего значения веса отклоняются в положительную сторону (Пример, точка # 1, которая хорошо отражает положительное отклонение от среднего).Но если человек оказывается ниже, чем в среднем по росту, то соответственно его весь становится меньше, чем в среднем по выборке.

Если большая часть наблюдений лежит в верхнем левом и нижнем правом квадрате, тогда взаимосвязь будет отрицательной.

**Рассчитаем показатели для каждого наблюдения:**

Начнём с испытуемого номер 1, рассчитаем его отклонение показателя веса от среднего значения по выборке умноженное на его отклонение показателя роста от среднего значения по выборке:

$$\Large (x_i - \bar{X}) \cdot (y_i - \bar{Y})$$

Так как отклонение от среднего в обоих случаях пложительно, то и произведение будет положительным.

Теперь возьмём испытуемого номер два на графике, и рассчитаем такой же показатель:

$$\Large (x_2 - \bar{X}) \cdot (y_2 - \bar{Y}) > 0$$

То и в этом случае, произведение будет положительным, так как отрицательное отклонение одного среднего умноженное на отрицательно отклонение другого среднего, даст положительный результат при произведении.

Следовательно, для всех точек внутри этих квадратов, произведение будет положительным.

А для значений которые находятся в других квадратах, произведение будет отрицательным. Если взять нижний правый квадрат, то отклонение от среднего значения веса будет положительным, но отклонение роста уже отрицательным, что при перемножении этих двух отклонений даст отрицательное значение.

**Вывод:** Если взять каждого из участников исследования и для всех наблюдений рассчитаем произведение отклонений по двум переменным, то в нашем случа, большенство таких произведений будут положительными.

**Нахождение ковариации:**

Следующим шагом необходимо найти среднее значение произведения отклонений от средних значений:

$$ \Large cov = \frac{1}{n -1 } \cdot \sum^{n}_{i=1} {(x_i - \bar{X}) \cdot (y_i - \bar{Y})}$$

Это и есть количественный показатель силы и направления взаимосвязи двух количественных переменных. Чем больше значение корреляции, тем сильнее взаимосвязь.

Добавляем минус 1, так как это связано с число степеней свободы, как и в случае дисперсии.

**Нахождение коэффициента корреляции:**

$$\Large r_{xy} = \frac{cov}{\sigma_x \cdot \sigma_y}$$

Нахождение коэффициента ковариации позваляет преобразовать все значение на промежутке от [-1, 1]. Так как значение cov напрямую будет зависить от данных, из-за этого, могут получать огромные значения ковариации и будет трудно оценить уровнень взаимосвязи.

<h1 style="color:#008B8B">Формула коэффициента корреляции</h1>

Давайте остановимся на формуле коэффициента корреляции, которую мы получили:

$\Large r_{xy} = \frac{cov_{xy}}{\sigma_{x}\sigma_{y}}$
 
Запишем формулу чуть подробнее и выполним возможные преобразования:

$\Large r_{xy} = \frac{\sum{(x_i - \bar{X})(y_i - \bar{Y})}}{(N-1)\sigma_{x}\sigma_{y}} = $ 

Распишем подробнее стандартные отклонения:

$\large \frac{\sum{(x_i - \bar{X})(y_i - \bar{Y})}}{(N-1)\sqrt{\frac{\sum(x_i - \bar{X})^2}{N-1}}\sqrt{\frac{\sum(y_i - \bar{Y})^2}{N-1}}} = $ 

Теперь вынесем $1/ (N - 1)$ из под корней:

$\large \frac{\sum{(x_i - \bar{X})(y_i - \bar{Y})}}{(N-1) \frac{1}{(N - 1)} \sqrt{\sum(x_i - \bar{X})^2}\sqrt{\sum(y_i - \bar{Y})^2}} =$

Сократим $(N - 1)$:

$\large \frac{\sum{(x_i - \bar{X})(y_i - \bar{Y})}}{ \sqrt{\sum(x_i - \bar{X})^2}\sqrt{\sum(y_i - \bar{Y})^2}}$ 

Таким образом, мы сократили $N - 1$ в знаменателе и получили финальную формулу для коэффициента корреляции, которую вы часто сможете встретить в учебниках:

$\large r_{xy} = \frac{\sum{(x_i - \bar{X})(y_i - \bar{Y})}}{ \sqrt{\sum(x_i - \bar{X})^2\sum(y_i - \bar{Y})^2}}$

<h1 style="color:#008B8B">Взаимосвязь двух переменных</h1>

Когда коэффициент корреляции равен 0, тогда в каждом квадрате находится равное количество кочек данных. С ростом коэффициента корреляции, все точки данных будут стремится к одной линии под углом 45 градусов.

В случае отрицательного коэффициента корреляции, точки данные будут распологаться вдоль прямой под 45 градусов, которая направлена в нижний левый угол.

Что из себя представляет показатеь общей дисперсии (Shared variance), который составляет 45%?

<img src="img/3_1_7.png" width="600" height="360">

Можно заметить, что это значение равняется квадрату коэффициента корреляции. Это называется коэффициентом детерминации. Елси коэффициент корреляции показывает силу и направление корреляции, то коэффициент детерминации показывает, какая часть наблюдений изменяется посредством другой переменной (как одна часть наблюдений зависит от значения другой переменнйо).

<h1 style="color:#008B8B">Вывод о пройденном материале</h1>

**Коэффициент корреляции:**

Чтобы исследовать взаимосвязь двух количественных переменных, мы можем рассчитать коэффициент корреляции Пирсона (Формула, которая была выведена).

$r_{xy}$ - показатель силы и направления взаимосвязи двух количественных переменных, который принимает значения от [-1, 1], где знак коэффициента корреляции показывает направление взаимосвязи.

**Коэффициент детерминации:**

$R^2$ - показывает, в какой степени дисперсия одной переменной обусловленна влиянием другой переменной. Коэффициент детерминации равен квадрату коэффициента корреляции, поэтому он принимает значение от [0, 1].

<h1 style="color:#008B8B">Проверка статистических гипотез</h1>

Как проводить проверку статистических гипотех используя коэффициент корреляции Пирсона?

Вернёмся к примеру с ростои и весом.

<img src="img/3_1_8.png" width="400" height="260">

**Формулирование гипотез:**

$H_0: r_{xy} = 0$ - в ГС коэффициент корреляции между ростом и весом равен нулю;

$H_1: r_{xy} \ne 0$ - Коэффициент корреляции между ростом и весом не равен нулю.

**Рассчёт p-value для проверки нулевой гипотезы:**

Необходимо рассчитать t-значение с соответствующим числом степеней свободы:

$ \large df = N -2$ - вычитаем два, так как мы работаем с двумя переменными $x$ и $y$.

t критерий в данном примере составляет 2.067, число степеней свобод df = 48, коэффициент корреляции 0.28589. P-value составил 0.04415, что позволяет отклонить нулевую гипотезу и принять альтернативную, что истинная корреляция между этими переменными не равна нулю.

### Задача 1

Верно ли это утверждение:  если коэффициент корреляции равен 0,7, то такая взаимосвязь всегда будет статистически достоверна (p - уровень значимости обязательно будет меньше 0,05)

**Ответ:**  Нет, высокая корреляция не обязательно означает статистически значимую взаимосвязь.

<h1 style="color:#008B8B">Влияние выбросов</h1>

Перед использование коэффициента корреляции Пирсона необходимо построить диаграмму рассеивания и посмотреть на характер взаимосвязи двух переменных. В случае, когда мы мысленно можем представить некую линию, которая описывает точки данных, тогда можно использовать корреляцию Пирсона. В случае, когда взаимосвязь не монотонная и не линейная, тогда этот коэффициент будет показывать ложные результаты.

Второе, необходимо чтоб характеристики переменных были распределны нормальным образом. Это связано с тем, что идея формулы корреляции Пирсона связано с понятием среднего значения как по переменной x так и по переменной y. Поэтому выбросы, ассиметрия... будет негативно сказываться на использовании среднего значения в качестве некоторого показателя (он перестаёт хорошо описывать распределение данных).

<h1 style="color:#008B8B">Коэффициенты корреляции</h1>

Если распределение не нормлаьное и взаимосвязь не линейна, тогда можно использовать непараметрические аналоги. Такие аналоги мы использовали в случае критерия t-Стьюдента мы использовали критерий Манна-Уитни, который переходит к ранжированным значениям. 

Есть несколько непараметрических аналогов коэффициента корреляции Пирсона:

* Коэффициент корреляции Спирмена;

* Коэффициент корреляции Тау Кендалла.

Рассмотрим коэффициент корреляции Спирмена для нашего примера. Если для нашего примера рассчитать коэффициент корреляции Спирмена, то он станет равен 0.23, что уже гворит о положитеьной взаимосвязи, нежел без его использования.

<img src="img/3_2_6.png" width="600" height="360">

Идея заключается в том, что мы ранжируем наши перменные (Таблица). Если у нас есть некоторая закономерность, первому рангу по X, соотносится первый ранг по Y, тогда это говорит о положительной взаимосвязи. А когда самый последний ранг по X, соотносится с третьим рангом по Y - говорит об отрицательной взаимосвязи.

$d^2$ представляет разность рангов, возведённая в квадрат. Основываясь на сумме всех расностей рангов возведённых в квадрат, рассчитывается коэффициент корреляции Спирмена, который имеет меньшую чувствительность к нарушению нормальности и к линейности взаимосвязи двух переменных.

<h1 style="color:#008B8B">Результаты корреляционных исследований</h1>

**Ошибка корреляции**

Положительная или отрицательная взаимосвязь между двумя переменными не говорит о причино-следственной зависимости. Такие примеры мы уже рассматривали. 

Иногда причино-следственная зависимость может быть обусловленна логическими соображениями. Например, возьмём взаимосвязь количества ядер процессора и его стоимостью в магазине. Если мы увидим положительную взаимосвязь, то это скорее всего означает, что технические характеристики компьютера определяют его стоимость, а не стоимость определяет технические характеристики.

**Сокрытая переменная**

Третья переменная является сокрытой переменной, которая и обуславливает корреляцию между двумя признаками. Например, размер ноги школьника будет положительно коррелировать с его знаниями в математике. Это связано с тем, чем школьник старше, тем больше у него размер ноги и знание математики будут  возрастать. Тем самым, размер ноги и знание математики обусловленно знанием третьей переменной. 

**Вывод:** Коэффициент корреляции показывает силу и направление взаимосвязи двух количественных переменных. Если взаимосвязь линейная, монотонная и соответствует нормальному распределению, тогда лучше использовать коэффициент корреляции Пирсона. При нарушении этих условий, необходимо применять непараметрические критерии корреляции Пирсона.

<h1 style="color:black" align="center">3.3 Регрессия с одной независимой переменной</h1>

В этой и в следующих главах мы научимся работать с одномерным регрессионным анализом, который позволяет проверять гипотезы о взаимосвязи одной  количественной зависимой переменной и нескольких независимых.

Сначала мы познакомимся с самым простым вариантом -  простой линейной регрессией, при помощи которой можно исследовать взаимосвязь двух переменных. Затем перейдем к множественной регрессии с несколькими независимыми переменными.

<h1 style="color:#008B8B">Что такое регрессионный анализ?</h1>

Регрессионный анализ представляет собой совокупность методов, которые позволяют исследовать взаимосвязь различных переменных между собой.

Начнём с одномерного регрессионного анализа, который как коэффициент корреляции позволяет нам исследовать взаимосвязь двух количественных переменных.

### Линия регрессии

Одномерный регрессионный анализ применяется для исследования взаимосвязи между двумя переменными. Та переменная, которую наносят на ось Y называетсся зависимой переменной. Переменная, которую наносят на ось X, называют независимой переменной (Предиктор).

Регрессионный анализ очень часто применяется для того, чтоб установить то, как одна переменная объясняет другую. Например, мы хотим посмотреть как на качество товара влияет его цена.

<img src="img/3_3_1.png" width="600" height="460">

Если две переменные каким-то образом взаимосвязанны между собой, удобно изобразить эту взаимосвязь проведя линию. Мы хотим, чтобы линия максимально описывала направление взаимосвязи и хорошо описывала распределение данных (каждая точка данных, должна находиться максимально близко к линии).

**Как это получить?**

Уравнение прямой имеет следующий вид: $\large b_0 + b_1\cdot x$, где 

$\large b_0$ - это свободный член (intercept), который отвечает за то, где линия пересекает ось Y;

$\large b_1$ (slope) - отвечает за направление и угол наклоной прямой с осью х.

Наша задача заключается в том, чтобы подобрать два параметра - $b_0$ и $b_1$. Для этого используется метод наименьших квадратов.

<h1 style="color:#008B8B">Метод наименьших квадратов</h1>

МНК - метод нахождения оптимальных параметров ($b_0, b_1$) линейной регрессии, таких, что сумма квадратов ошибок (остатков) была минимальна.

### Что такое остаток?

Посмотрим на пример, где необходимо сделать вывод, какя линия лучше всего описывает данные? Визуально понятно, что синяя линия лучше всего отражает набор данных.

<img src="img/3_3_2.png" width="500" height="260">

Но чем синяя линия лучше красной? Для каждой из точке рассчитаем расстояние от реального значения по оси y, до значения на регрессионной прямой, это расстояние и называется остатком. Если у нас есть наблюдение $y_1$, то остаток можно найти следующим образом:

$\large e_1 = y_1 - \hat{y_1}$, где

$\large y_1$ - Реально значение

$\large \hat{y_1}$ - Предсказанное значение на регрессионной прямой.

### Оценка регрессионной прямой

Если рассчитать сумму квадратов всех остатков. Квадрат необходим для того, чтобы положительные значения не компенсировали отрицательные. Следовательно, если найти такой показатель для красной линии, то он будет иметь большее значение.

Чтобы сумма квадратов ошибок была минимальная, коэффициенты рассчитывают следующим образом.

$\large b_1 = \frac{sd_y}{sd_x} \cdot r_{xy}$

Коэффициент корреляции опредлеяет положение нашей линии. На графике мы наблюдаем положительную взаимосвязь и линия имеет соотвествующее направление. В случае отрицательной взаимосвязи, точки данных будут расположены в других квадратов и направление линии будет другим.

$\large b_0 = \bar{Y} - b_1 \cdot \bar{x}$

Синяя линия построена с использованием данного метода. Точка пересечения с осью Y, это и есть значение коэффициента $b_0$, а коэффициент $b_1$ отвечает за угол наклона.

Тем самым, математическая модель, которая отражает взаимосвязь двух перменных будет: 

$\large \hat{y} = 5.4 + 0.83 \cdot x$

<h1 style="color:black" align="center">3.4 Гипотеза о значимости взаимосвязи и коэффициент детерминации</h1>

<h1 style="color:#008B8B">Коэффициенты линейной регрессии</h1>

Мы построили линю регрессии и теперь хотим ответить на вопрос, на сколько статистически значима взаимосвязь двух переменных?

Какой коэффициент уравнения отвечает за взаимосвязь? Этим коэффициентом являет тот, который стоит перед $b_1 \cdot x$. Так как именно этот коэффицент отвечает за угол наклона прямой.

**Что произойдёт если никакой взаимосвязи между данными нет?**

Тогда все точки, случайным образом будут распределены по диаграмме рассеивания и если коэффициент корреляции равен нулю $r_{xy}$, чему будет равен коэффициент $b_1$?

Вспомним формулу для коэффициент $b_1$

$\large b_1 = \frac{sd_y}{sd_x} \cdot r_{xy}$

Так как коэффициент корреляции равен нулую, значит и $b_1$, тоже будет равен нулю. А если $b_1$ равен нулю, тогда $b_0$:

$\large b_0 = \bar{Y} - b_1 \cdot \bar{x}$

Примет среднее значение по $\bar{Y}$ и тогда регрессионная линия проходила горизонально через среднее значение зависимой переменной $Y$.

**Формулирвоание гипотез**

Если мы предпологаем что в ГС две переменны не имеют взаимосвязи, тогда коэффициент $b_1$ в ГС будет равен нулю:

$\large H_0: b_1 = 0$

Альтернатианя гипотеза будет утверждать обратное:

$\large H_1: b_1 \ne 0$

**Использование t-критерия:**

Если бы многократно повторяли наш эксперемент в ситуации, когда никакой взаимосвязи между переменными нет, то выборочные значения коэффициентов $b_1$ распределились бы нормальным образом вокруг соответствующего значения ГС, то есть нуля. Поэтому, в такой ситуации лучше использовать t-распределение.

**Рассчёт t-критерия:**

Насколько выборочное значение коэффициента $b_1$ отклонилось от ожидаемого среднего в ГС (среднее значение ГС равно 0) и всё это делённое на стандартную ошибку распределения коэффициента $b_1$:

$\large t = \frac{b_1 - 0}{se} = \frac{b_1}{se}$

$t$-критерий проверяет гипотезу о том, что коэффициент $b_1$, который определяет угол наклона, отличный от нуля. Но формулу можно упростить, так как ноль не имеет смысла.

**Определение числа стпеней свобод:**

Число степеней свобод, определяется аналогично как при рассчёте коэффициента корреляции:

$\large df = N-2$

Отклонив нулевую гипотезу, о том что $\beta_{1}=0$, мы можем сделать следующий вывод: Обнаружена значимая взаимосвязь двух переменных/

<h1 style="color:#008B8B">Коэффициент детерминации</h1>

$R^2$ - доля димперсии зависимой переменной (Y), объясняемая регресионной моделью.

Предположим, мы решили выявить, как качество определяется ценой товара. Для этого мы построили регрессионную модель:

<img src="img/3_4_1.png" width="600" height="360">

На левом рисунке обозначена регрессионная прямая, которая имеет небольшой угол наклона. И мы хотими ответить на вопрсо, в какой мере изменчивость переменной качества объясняется взаимосвязью этой переменной с ценой (объясняется нашей моделью).

**Формула для коэффициента детерминации:**

$\large R^2 = 1 - \frac{SS_{res}}{SS_{total}}$, где

$\large SS_{res}$ - это сумма квадратов остатков. Она представляет растояние от нашего наблюдения, до регрессионной прямой возведённое в квадрат.

$\large SS_{total}$ - общая сумма квадратов (Общая изменчивость). Представляет собой расстояние от наблюдения до среднего значения. Среднее значение обозначено красной пунктирной линией.

**Пример слева:**

На левом рисунке расстояние до регрессионной прямой практически одинковое расстоянию до прямой проходящей через среднее значение. Это связано с тем, что не никакой взаимосвязи и наша регрессионная линия почти принимает вид линии среднего значения ($y = \bar{Y}$).

Что это означает с точки зрения квадрата детерминации? Если у нас сумма квадратов остатка практически равняется общей сумме квадратов, тогда их отношение будет стремится к еденице. А если это отношение стремится к еденице, тогда 1 минус 1, даст ноль. Это и будет говорить о том, что наша модель никак не объясняет изменчивость зависимой переменной.

**Пример справа:**

В обратной ситуации, когда выражена положительная корреляция $r=0.99$. Это означает, что наши точки очень близко расположены с регрессионной прямой и сумма квадратов остатков будет меньше общей суммы квадратов. Это приводит к тому, что отношение $SS_{res}$ и $SS_{total}$ будет становиться всё  меньше и меньше, чем лучше регрессионная модель. 

В нашем случае $R^2$ будет стремиться к 1, так как отношение стремится к 0. И такой $R^2 = 1$ говорит о том, что практически 100% изменчивости зависимой переменной обуславливается её взаимосвзью с независимой пременной, то есть регрессионная модель хорошо описывает поведение и изменчивость зависимой переменной.

Ну и в одномерном регрессионном анализе, сохраняется равенство о том, что коэффициент детерминации равен квадрату корреляции.

<h1 style="color:black" align="center">3.6 Применение регрессионного анализа и интерпретация результатов</h1>

<h1 style="color:#008B8B">Применение однофакторного регрессионного анализа</h1>

Загрузим набор данных о социально-экономических показателях для каждого штата США.

**Подключение библиотек:**

```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

### Загрузка и обзор данных

```python
data = pd.read_csv('input\states.csv')
data.head()
```

`state` - название шата;

`metro_res` - какой процент населения штата живёт в столичной области;

`white` - процент белокожего населения;

`hs_grad` - процент людей, которые имеют среднее образование;

`poverty` - процент людей проживающих в бедности;

`female_house` - процент семей, где женщина домохозяйка.

Применим линейную регрессию с одной независимой переменной и исследуем, как взаимосвязана бедность и уровень образования.

### Построим диаграмму рассеивания:

```python
data.plot(x='hs_grad', y='poverty', kind='scatter', figsize=(8,5))
plt.title('Корреляция', fontsize=14)
plt.xlabel("Уровень образования %")
plt.ylabel("Уровень бедности %")
plt.show()
```

Зависимая переменная - показатель бедности, в качестве независимой переменной выступает уровень образования.

**Построение регрессионной прямой:**

В нашем случае это будет уравнение регрессионной прямой: $\hat{y} = b_0 + b_1 \cdot x$. Для этого необходимо рассчитать коэффициент $b_0$ b $b_1$. 

```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats

b0 = scipy.stats.linregress(data['hs_grad'], data['poverty']).intercept
b1 = scipy.stats.linregress(data['hs_grad'], data['poverty']).slope
y = b0 + b1 * data['hs_grad']

plt.scatter(data['hs_grad'], data['poverty'])
plt.text(x=100, y=20, s='R = -0.75', fontsize=20)

plt.plot(data['hs_grad'], y, color='green')
plt.grid()
plt.xlabel('graduation (%)')
plt.ylabel('poverty (%)')
plt.title('poverty \ grad')
plt.show()
```

### Применим регрессионный анализ

```python
import statsmodels.api as sm # import statsmodels 

X = data["hs_grad"] ## X usually means our input variables (or independent variables)
y = data["poverty"] ## Y usually means our output/dependent variable
X = sm.add_constant(X) ## let's add an intercept (beta_0) to our model

# Note the difference in argument order
model = sm.OLS(y, X).fit() ## sm.OLS(output, input)
predictions = model.predict(X)

# Print out the statistics
model.summary()
```

**Результаты анализа для $b_0$**

По результатам анализа, $b_0$ (intercept) = 64.7810. $b_1$ = -0.6212.

Насколько значима эта взаимосвязь? Гипотезу мы формулирвоали следующим образом:

$\large H_0: b_1 = 0$

Для проверки этой гипотезы используется t-тест, результаты которого предствалены в таблице. Также есть стандартная ошибка и p-value, который сильно меньше чем 0.05, что и позволяет отклонить нулевую гипотезу, а значит найдена статистически значимая взаимосвязь между двумя переменными.

**Результаты анализа для $b_1$**

Аналогичные значения предствалены для интерсепта и тут тоже проверяется гипотеза о том что:

$\large H_0: b_0 = 0$

И тут тоже p-value меньше чем 0.05, что позволяет отклонить нулевую гипотезу. Если предположить о существовании штата, где процент людей с средним образованием равен 0, тогда мы можем предположить, что 64.7810 % проживают в бедности.

**Интерпритация коэффициента $b_1$**

Удостоверившись в том, что взаимосвязь между неременными статистически значимая. Теперь можно проинтерпретировать значение коэффициента $b_1$, который равен -0.6212 говорит о том, что взаимосвязь отрицательная. Это означает, что с увеличением каждого процента людей, которые имеют высшее образование, процент людей проживающих в бедности будет уменьшаться на 0.6212%. 

Поэтому коэффициенты перед независимыми переменными имеют определённый смысл, это то, как будет изменяться значение зависимой переменной с еденичным изменением независимой переменной.

**Значение коэффициента детерминации**

Коэффициент детерминации составил R-squared = 0.558. Это означает, что 0.558% изменчивость зависимой переменной объясняется моделью. Значение детерминации имеет небольшой показатель, так как половина изменчивости объясняется дгруими факторами, которые не включены в модель. 

Ещё можно заметить, что у коэффициента детерминации, рассчитано f-значение при помощи дисперсионного анализа, которое равно 61.81.

### Подведение итогов

Мы решили выявить, как взаимосвязаны две переменные между собой, а именно уровень бедности и образование. Для этого применили регрессионный анализ с одной независимой переменной.

Мы смогли посмтореть насколько значима данная взаимосвязь, она оказалась отрицательной. Была построена линейная модель, которая максимально удачно описывает взаимосвязь двух переменных. Выяснили какой процент дисперсии одной переменной, обуславливается взаимосвязью с другой переменной. И посмотрели, на сколько изменяется уровень бедности с еденичным изменением уровня образования.

Это предоставляет куда больше информации, чем значение корреляции Пирсона.

<h1 style="color:#008B8B">Применение однофакторного регрессионного анализа</h1>

```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
import numpy as np

data = pd.read_csv('input\states.csv')

# рассчет ковариационной матрицы двух векторов, на выходе ковариация и дисперсии каждого из массивов, 
# или можно рассчитать вручную по формулам из уроков cov - урок 3_1 третье видео; D - урок 1_5 третье видео
X = data['hs_grad']
Y = data['poverty']
ssxm, ssxym, ssyxm, ssym = np.cov(X,Y).flat

# рассчет корреляции по формуле из урока 3_1 третье видео: r = cov/stdx*stdy, 
# т. к. в ковариационной матрице у нас дисперсии, то в знаменателе нужно взять квадратный корень 
if ssxm == 0 or ssym == 0:
    r = 0
else:
    r = ssxym / (ssxm * ssym)**0.5
    
# расчет slope и intercept из урока 3_3 второе видео: slope = (stdy/stdx) * r
slope = r * (ssym/ssxm)**0.5
intercept = np.mean(Y) - slope * np.mean(X)

df = len(X) - 2
# рассчет t - значения, вероятности p и стандартной ошибки для X параметра подсмотренно тут: 
# https://github.com/scipy/scipy/blob/v1.4.1/scipy/stats/_stats_mstats_common.py#L15-L144
tx = r * np.sqrt(df / ((1.0 - r)*(1.0 + r)))
sterrestx = np.sqrt((1 - r**2) * ssym / ssxm / df)
px = 2 * stats.t.sf(np.abs(tx), df)

# Находим сумму квадратов по параметру X
s = [i**2 for i in X]
sterresty = (sterrestx**2/51*sum(s))**0.5

# рассчет t - значения, вероятности p и стандартной ошибки для Y параметра подсмотренно тут: 
# https://en.wikipedia.org/wiki/Simple_linear_regression
ty = intercept/sterresty
py = 2 * stats.t.sf(np.abs(ty), df)

# рассчет F-статистики подсмотренно тут:
# https://www.chem-astu.ru/science/reference/fischer.html
F = r**2/(1-r**2)*df
p_val = stats.f.sf(F, 1, df)

# создаем таблицу как в лекции
ttlinear = pd.DataFrame(data = {
    'Estimate':[round(num, 4) for num in [intercept,slope]], 'Std.Error':[round(num, 4) for num in [sterresty, sterrestx]], 
    't value':[round(num, 2) for num in [ty,tx]], 'Pr(>|t|)':[py,px]}, 
                        index = ['(Intercept)','hs_grand'])

# вычисляем остатки
#residuals = Y - intercept - slope*X

fig, ax1 = plt.subplots(nrows=1, ncols=1, figsize=(10, 7))

# точечный график
ax1.scatter(X,Y)

# регрессионная прямая
x1=np.linspace(min(X),max(X),51)
y1=intercept + slope*x1
plt.plot(x1,y1,'-r',label='b0 = {}, b1 = {}'.format(round(intercept,2), round(slope,2)))

plt.grid()
plt.title('Связь бедности и уровня образования')
plt.xlabel('Среднее образование (%)')
plt.ylabel('Бедность (%)')
plt.xticks(np.arange(76,93, step=4))
plt.yticks(np.arange(5,18, step=5))
plt.legend(loc='upper right')
plt.show()


print(ttlinear)
print('')
print('Multiple R-squared: {},'.format(round(r**2,4)))
print('F-statistic(1,{}) = {}, p-value = {}'.format(df, round(F,2), p_val))
```

<h1 style="color:black" align="center">3.8 Регрессионный анализ с несколькими независимыми переменными</h1>

<h1 style="color:#008B8B">Множественная регрессия</h1>

Множественная регрессия позволяет исследовать влияние сразу нескольких независимых переменных на одну зависимую переменную.

**Вспомним пример:**

Мы решили выяснить, как на показатель бедности (зависимая переменная) влияет фактор уровня образования населения. Если такая взаимосвязь является линейной, то можно построить линию регрессии, подобрав коэффициенты таким образом, чтобы сумма ошибок была минимальной:

$\large \hat{y} = b_0 + b_1 \cdot x$

Но в большенстве случаев много зависимых переменных, которые могут дать полезную информацию о зависимой переменной. Мы можем добавить независимые переменные следующим образом.

$\large \hat{y} = b_0 + b_1 \cdot x_1 + \ldots + b_n \cdot x_n$

### Две независимые переменные

Мы решили выяснить, как образование и процент белого населения влияет на показатель бедности.

**Визуализируем результат:**

<img src="img/3_8_1.png" width="auto" height="auto">

В этом случае, оптимально аппроксимировать данные будет уже не прямая линия, а плоскость, так как мы имеем дело с пространством. И в качестве остатков, выступает не расстояние до прямой линии, а расстояние до плоскости.

И коэффициенты можно подобрать также с использованием метода наименьших квадратов, чтобы сумма всех квадратов расстояния была минимальной.

<h1 style="color:#008B8B">Требования к данным</h1>

1) Линейная зависимость переменных;

2) Нормальное распределение остатков;

3) Гетероскедастичность;

4) Проверка на мультиколлинеарность;

5) Нормальное распределение переменных как зависимой переменной, так и независимых переменных.

Теперь построим модель, которая будет включать в себя все 4 независимые переменные.

### О интерпретации значения коэффициента регрессии

Когда мы интерпретитруем значение коэффициента, мы представляем, что все остальные переменные не изменяются. Таким образом, мы смотрим только на влияние одной переменной на одну зависимую переменную. Это важно, так как регрессионный анализ позволяет исключить факт того, что две переменные связаны между собой, засчё тольлко третьей переменной.

**Пример**

Разберёмся на примере: Мы решили выявить, как на интеллект человека, влияет размер его ноги. Проведя такой эксперемент, мы могли бы получить положительную корреляцию (чем больше размер ноги, тем больше интеллект). Это связано с тем, что в эксперементе заложена третья переменная, это возсраст человека, который обуславливает эти различия (чем больше возраст человека на определённом промежутке, тем выше интеллект, тем и больше размер ноги человека).

Поэтому, если бы мы включили третью переменную в регрессионную модель, то мы бы рассчитали на сколько размер ноги влияет на интеллект, при условии, что возраст не изменяется и тогда мы не получим значимую корреляцию. А если просто взять и посчитать между двумя переменными корреляцию, тогда она будет значимой, засчёт третьей переменной.

**Вывод:** Регрессия позволяет посмотреть отдельные взаимодействия каждой независмой переменной с зависимой, при условии того, что все остальные переменные фиксированные. Поэтому в регрессионном анализе, мы не поулчим ложную взаимосвязь, которая обусловлена третьим фактором.

**Исправленный R-квадрат** - Скорректированный коэффициент детерминации. Рассчитывается при включении в модель дополнительных независимых переменных.

Почему в случа множественной регресси рассчитывается исправленный R-квадрат? Как мы помним коэффициент детерминации позволяет нам ответить на вопрос: какая часть изменчивости зависимой переменной, объясняется нашей моделью. Другими словами, насколько хорошо реальные значения зависимой переменной согласуются с предсказанными.

Но тут мы включаем сразу несколько предикторов возникает эффект множественных сравнений. Если бы мы включили куда больше предикторов, тогда R-квадрат может стать ещё больше, это произошло не из-за того, что имеется реальная взаимосвязь предикторов с зависимой переменной, а из-за того, что мы будем чаще получать случайные взаимосвязи между предикторами и зависимой переменной.  Поэтому, когда в модели больше чем одна независимая переменная, то мы делаем поправку на это факт и рассчитываем исправленный R-квадрат, с учётом того, что предикторов больше, чем один.
