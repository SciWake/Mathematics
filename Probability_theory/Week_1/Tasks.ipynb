<h3 style="color:#2E8B57">Задача № 1</h3>

Книжная полка вмещает сорок книг. Среди имеющихся у нас сорока книг (все книги разные) есть три тома Пушкина. Ставим наши книги в случайном порядке на полку (все расстановки являются равновероятными). С какой вероятностью тома Пушкина окажутся стоящими в правильной последовательности (от меньшего к большему номеру тома, если считать книги, стоящие на полке, слева направо), но не обязательно рядом?

<h4 align='center'>Решение:</h4>

Определим пространство элементарных исходов $\Omega$ для нашей задачи. Так как случайным является порядок книг на полке, то элементарными исходами являются расстановки книг на полке (всего $40!$ элементарных исходов): $\Omega = \{\omega_1, \ldots, \omega_{40!}\}$. Таким образом, вероятность одного элементарного исхода равна $\frac{1}{40!}$. Посчитаем число расстановок, в которых тома Пушкина стоят в правильным порядке (соответствующее событие мы обозначим $A$). Выберем сначала три позиции, на которые можно поставить тома Пушкина. Таких способов $C_{40}^3$. Коль скоро мы зафиксировали эти позиции, мы уже обязаны поставить первый том Пушкина на первую позицию, второй - на среднюю и третий - на последнюю. Осталось поставить остальные $37$ книг. Это можно сделать $37!$ способами. Итак, $|A| = C_{40}^3 \cdot 37!$.

Тогда

### ${\sf P}(A)=\frac{|A|}{|\Omega|}=\frac{C_{40}^3 \cdot 37!}{40!}=\frac{40! \cdot 37!}{3! \cdot 37! \cdot 40!}=\frac{1}{3!}=\frac{1}{6}$

__Ответ:__ $\frac{1}{6}$

<h3 style="color:#2E8B57">Задача № 2</h3>

Из множества $N = \{a_1, \ldots a_N\}$ объектов выбирается случайное подмножество (выборы всех подмножеств равновероятны). Найдите вероятность того, что это случайное подмножество имеет четную мощность (мощностью множества называется число элементов в нем).

<h4 align='center'>Решение:</h4>

Обозначим данное множество $\{a_1,\ldots,a_N\}$. Определим пространство элементарных исходов $\Omega$ для нашей задачи: $\Omega=\{\omega_1,\ldots,\omega_{2^N}\}$ (так как во множестве $\{a_1,\ldots,a_N\}$ всего $2^N$ подмножеств).
