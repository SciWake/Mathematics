<h3 style="color:#2E8B57">Задача № 1</h3>

Книжная полка вмещает сорок книг. Среди имеющихся у нас сорока книг (все книги разные) есть три тома Пушкина. Ставим наши книги в случайном порядке на полку (все расстановки являются равновероятными). С какой вероятностью тома Пушкина окажутся стоящими в правильной последовательности (от меньшего к большему номеру тома, если считать книги, стоящие на полке, слева направо), но не обязательно рядом?

<h4 align='center'>Решение:</h4>

Определим пространство элементарных исходов $\Omega$ для нашей задачи. Так как случайным является порядок книг на полке, то элементарными исходами являются расстановки книг на полке (всего $40!$ элементарных исходов): $\Omega = \{\omega_1, \ldots, \omega_{40!}\}$. Таким образом, вероятность одного элементарного исхода равна $\frac{1}{40!}$. Посчитаем число расстановок, в которых тома Пушкина стоят в правильным порядке (соответствующее событие мы обозначим $A$). Выберем сначала три позиции, на которые можно поставить тома Пушкина. Таких способов $C_{40}^3$. Коль скоро мы зафиксировали эти позиции, мы уже обязаны поставить первый том Пушкина на первую позицию, второй - на среднюю и третий - на последнюю. Осталось поставить остальные $37$ книг. Это можно сделать $37!$ способами. Итак, $|A| = C_{40}^3 \cdot 37!$.

Тогда

### ${\sf P}(A)=\frac{|A|}{|\Omega|}=\frac{C_{40}^3 \cdot 37!}{40!}=\frac{40! \cdot 37!}{3! \cdot 37! \cdot 40!}=\frac{1}{3!}=\frac{1}{6}$

__Ответ:__ $\frac{1}{6}$

<h3 style="color:#2E8B57">Задача № 2</h3>

Из множества $N = \{a_1, \ldots a_N\}$ объектов выбирается случайное подмножество (выборы всех подмножеств равновероятны) Вернее, говорится так: давайте рассмотрим множество всех подмножеств этого множества, и вот из этого множества, множества всех подмножеств, возьмем один случайный элемент. Вот это будет случайное подмножество нашего множества из $N$ объектов. Найдите вероятность того, что это случайное подмножество имеет четную мощность (мощностью множества называется число элементов в нем).

<h4 align='center'>Решение:</h4>

Обозначим данное множество $\{a_1,\ldots,a_N\}$. Определим пространство элементарных исходов $\Omega$ для нашей задачи: $\Omega=\{\omega_1,\ldots,\omega_{2^N}\}$ (так как во множестве $\{a_1,\ldots,a_N\}$ всего $2^N$ подмножеств).Таким образом, вероятность одного элементарного исхода равна $\frac{1}{2^N}$. Обозначим $A$ событие, заключающееся в том, что случайное подмножество имеет четную мощность: $A=\{\omega:\,|\omega|\text{ - четное}\}$. Будем считать, что $0$ это четное число (тогда пустое множество имеет четную мощность). Очевидно, $|A|=C_N^0+C_N^2+\ldots$. Посчитаем это количество иным способом: каждый элемент (начиная с $a_1$) можно либо положить, либо не положить в строящееся случайное множество (две возможности для каждого элемента). Так как мы хотим построить множество четной мощности, то, возможно, последний элемент мы в наше множество положить не можем.

 Действительно, после того, как мы приняли решение относительно элемента $a_{N-1}$, мы либо обязаны положить элемент $a_N$, либо обязаны его не класть. Итак, по две возможности мы имеем ровно для $N-1$ элементов. Таким образом, $|A|=2^{N-1}$. (В частности, это означает, что $C_N^0+C_N^2+\ldots=2^{N-1}$).
 
__Ответ:__ ${\sf P}(A)=\frac{|A|}{|\Omega|}=\frac{2^{N-1}}{2^N}=\frac{1}{2}$.

<h3 style="color:#2E8B57">Задача № 3</h3>

Графом на $n$ вершинах называется пара $G=(V,E)$, где $V$ - множество из $n$ элементов, называемое множеством вершин графа, $E$ - множество пар вершин из $V$, называемое множеством ребер. Если пары в $E$ упорядочены, то граф называется ориентированным. Иначе - неориентированным. Петлей называется ребро графа, соединяющее вершину саму с собой. Граф на $n$ вершинах называется простым циклом, если его множество вершин можно занумеровать таким образом числами от $1$ до $n$, что каждая вершина $i$ соединена ребром с вершиной $i+1$, а вершина $n$ - с вершиной $1$ (и в графе нет других ребер).

<h4 align='center'>Решение:</h4>

У нас есть множество, которое состоит из $n$ элементов $\{1, \ldots, n\}$. Это множество мы считаем множеством вершин графа. мы можем любые две из этих точек между собою соединить отрезочком, который называется ребром. То есть вот каждая неупорядоченная пара элементов этого множества теоретически может быть ребром графа. А может и не быть таковым. Например, в качестве графа вполне можно рассмотреть вот такую вот конструкцию:

<img src="img/Seminar_3_1.svg" width="1200" height="1000">`

Пространство элементарных исходов $\Omega$ - это множество всех неориентированных графов без петель на множестве вершин $\{1,\ldots, n \}$. Так как каждая пара различных вершин либо образует, либо не образуют ребро (две возможности для $C_n^2$ пар вершин), то $|\Omega|=2^{C_n^2}$. Неупорядоченных пар вершин, которые можно взять из множества, имеющего мощность $n$, их, конечно, $C$ из n по $2$, и для любой такой пары либо она соединена ребром, либо не соединена, поэтому $2$ в степени $C$ из n по $2$ – это количество вариантов провести или не провести ребра. 

Вот такое вот пространство элементарных исходов, и каждый элементарный исход – то есть граф – имеет, согласно классическому определению, вероятность, равную $1$ поделить на $2$ в степени $C$ из $n$ по $2$. Вот, но нас в этой задаче интересует, с какой вероятностью такой вот случайный граф является циклом, простым циклом?

<img src="img/Seminar_3_2.svg" width="800" height="600">`

Простой цикл – это такая последовательность вершин и ребер, в которой они вот так чередуются, и ни ребра, ни вершины не повторяются, покуда мы не наткнемся на исходную.

С какой вероятностью случайный граф окажется простым циклом? $A$ – это вот то самое событие, которое нас интересует. И чтобы посчитать его вероятность, мы, как обычно, должны написать в знаменателе общее количество элементарных событий, то есть общее количество графов на $n$ вершинах, а в числитель мы должны записать столько графов, сколько благоприятствует этому событию.

Сколько вот есть циклов, простых циклов на $n$ вершинах? Ну вот, скажем, на трех вершинах $(1, 2, 3)$ сколько всего есть простых циклов? А очень просто – только один. Смотрите, как ни перетасовывай, как ни переставляй между собою единицу, двойку и тройку, у нас все тот же самый треугольник получается. 

<img src="img/Seminar_3_3.svg" width="1200" height="100">`

Как посчитать общее количество простых циклов, имеющих $n$ вершин?

Ну, конечно, если бы мы строили цепочку вот такую и как-то пытались расставить по этой цепочке числа, то таких цепочек ну столько, сколько есть всего перестановок на множестве из $n$ элементов, а перестановок их $n$ факториал. Вернуться к примеру с треугольником, то есть перестановка $(1, 2, 3)$, а есть перестановка $(1, 3, 2)$. Это разные перестановки. Но в тот момент, когда мы их зацикливаем получаем один цикл.

В общем случае мы посмотрим на какую-нибудь перестановку:

$$\begin{matrix}
a_1 & \ldots & a_n\\
a_2 & a_3 & \ldots & a_n & a_1 \\
a_3 & a_4 & \ldots & a_n & a_1 & a_2 \\
& \ldots & \ldots  & \ldots \\
a_n & a_1 & \ldots & a_{n-1}
\end{matrix}$$

Все эти различные цепочки зацикливаются в один и тот же цикл. Естественно, получается, что уже, как минимум, в $n$ раз циклов меньше, чем существует всех возможных вот таких вот цепочек. На самом деле, ведь мы же можем еще эту последовательность прочитать в обратном порядке. И уже вот эту вот в обратном порядке прочитанную последовательность снова всеми возможными $n$ способами по циклу попереставлять:

$$\begin{matrix}
a_n & a_{n-1} & \ldots & a_1\\
a_{n-1} & \ldots & a_1 & a_n\\
& \ldots & \ldots \\
a_1 & a_n & \ldots & a_2
\end{matrix}$$

Есть $n!$ цепочек и для каждой цепочки есть $2n$ перестановок, которые дают один и тот же самый цикл, поэтому надо поделить на $2n$, коль скоро мы считаем количество всех циклов, и это получится: 

### $\frac{n!}{2n} = \frac{(n – 1)!}{2}$

Пусть $A$ (интересующее нас событие) - множество простых циклов, входящих в $\Omega$. Тогда $|A|=\frac{n!}{2n}$, так как способов сделать цепь (расположить все вершины подряд) $n!$, но когда мы эту цепь "зациклим", то мы один и тот же цикл учтем $2n$ раз (nn цепей, в которых элементы переставляются "по кругу" дают один и тот же цикл, а каждую такую цепь можно "прочитать" в обратном порядке, что снова даст тот же самый цикл).

Возвращаемся к подсчету нашей вероятности.  

### ${\sf P}(A)=\frac{|A|}{|\Omega|}=\frac{(n-1)!/2}{2^{C_n^2}}$

Но в задаче еще предлагалось найти предел этой вероятности при $n$ маленькое, стремящимся к бесконечности.

Все-таки что вероятнее: вытащить из огромной кучи графов простой цикл или не простой цикл. Совершенно понятно, что вероятнее вытащить не цикл, потому что цикл – ну это что-то такое очень специфическое.

Заметим, наконец, что 

### $\frac{(n-1)!}{2}<n!\leq n^n=2^{n\log_2 n}$ 

Тогда

### ${\sf P}(A)<\frac{2^{n\log_2 n}}{2^{\frac{n^2}{2}-\frac{n}{2}}}\rightarrow 0,\quad n\rightarrow\infty$,

Так как порядок роста $\frac{n^2}{2}-\frac{n}{2}$ больше, чем порядок роста $n\log_2$.

<h3 style="color:#2E8B57">Задача № 4</h3>

Есть множество чисел $\{1, 2, \ldots, n\}$. Из этого множества чисел мы последовательно извлекаем 3 числа по схеме выбора без возвращения.

Без возвращения – это значит, что извлеченные числа не могут совпасть, то есть мы вытаскиваем какое-то число отсюда, называем его i, после чего в этом множестве уже нет элемента с номером i, и мы из оставшегося множества, имеющего мощность, понятное дело, n − 1, вытаскиваем следующий элемент j, независимо от того, как был вытащен элемент i, просто случайно вытаскиваем еще один элемент. Ну и наконец, остается у нас n − 2 элемента, после того, как вытащены i и j, и мы извлекаем третий элемент k. Вот это называется схема выбора без возвращения, и в рамках такой схемы мы извлекаем последовательно три элемента: i, j и k.

Найдем вероятность того, что k – третье вытащенное число – лежит между числами, по величине своей, лежит между i и j,  если известно, что первое число i меньше второго j.


<h4 align='center'>Решение:</h4>

Необходимо найти **P(K лежит между i и j | i < j)**

Ну давайте рассуждать, из чего изначально у нас состояло пространство элементарных исходов? 
Ну, $\omega$ состояло из всевозможных троек чисел i, j и k, причем упорядоченных троек различных чисел. Числа различные, потом что вытаскиваются без возвращений. $\Omega$ – это множество упорядоченных троек различных чисел i, j и k.

$\Omega = \{i, j, k\}$, где количество таких элементарных исходов $|\Omega| = A_n^3$

Никоим образом вот это знание нам не поможет осознать чему равняется искомая условная вероятность, потому что, внимание, мы ведь сейчас находимся уже не в $\Omega$, а в подмножестве $\Omega$. Мы находимся в той части $\omega$, в которой находятся тройки i, j, k, у которых i строго меньше чем j, потому что крупье нам сообщил вот это дополнительное знание. Мы точно знаем, что ни одна из тех троек i, j, k – упорядоченных троек, в которых i > j не реализовалась. Стало быть, мы находимся в той ситуации, когда i заведомо строго меньше, чем j. И это наше дополнительное знание и этим мы должны как-то пользоваться.

Как можно расставить номера у элементов (i, j, k), упорядочить элементы этой «кучки» так, чтобы i получилось меньше чем j?

Можно написать таким образом:

**i, j, что-то ещё**

**i, что-то ещё, j**

**что-то ещё, i, j**

Все остальные взаимные расположения чисел i и j не благоприятствуют вот этому условию. Но теперь, а если мы к этому условию дополнительно добавляем утверждение о том, что k лежит между i и j? Тогда сколько таких ситуаций среди заведомо имеющихся? Тем самым, нам подходит всего одна ситуация из всех (вторая).

Условию (i < j) – три ситуации, а двум ограничениям – только одна из этих трех. То есть понятно, что $\omega$ разбивается на кучки, в рамках каждой кучки только одна ситуация из трех нам благоприятствует, ну значит, и в целом только треть ситуаций благоприятствует тому, чего мы хотим. Ну и получается, что вероятность, которая нас интересует, в точности равняется $\frac{1}{3}$. 

__Ответ:__ $\frac{1}{3}$

<h3 style="color:#2E8B57">Задача № 5</h3>

Среди 25 экзаменационных билетов 5 "хороших". Два студента по очереди берут по одному билету (первый студент берет каждый из билетов с одинаковой вероятностью, второй - равновероятно любой из оставшихся). 

<h4 align='center'>Решение:</h4>

**Найдите вероятность того, что первый студент взял "хороший" билет?**

Всего 5 благоприятствующих исходов и общее количество исходов 25. Тем самым вероятность достать хороший билет для первого студента составит: 

$\frac{5}{25} = \frac{1}{5}$

**Найдите вероятность того, что второй студент взял "хороший" билет?**

Пусть A – второй студент вытянул хороший билет.

Для подсчета этой вероятности рассмотрим два случая: Первый случай – это что первый студент вытянул хороший билет, и вот это как раз плохо для второго. Тогда у него вероятность вытянуть хороший билет, действительно, понижается. Но второй случай состоит в том, что первый студент вытянул плохой билет, и тогда плохих билетов стало меньше.  

$B_1$ – это первый вытащил хороший и 

$B_2$ – это первый вытащил плохой. 

Поэтому желая посчитать вероятность события A, мы просто можем воспользоваться формулой полной вероятности.

$${\sf P}(A) = {\sf P}(A \mid B_1){\sf P}(B_1) + {\sf P}(A \mid B_2){\sf P}(B_2)$$

Найдём значения из формулы:

Вероятность того, что первый вытащил хороший билет: $P(B_1) = \frac{1}{5}$

Давайте посмотрим с какой вероятностью второй студент достанет хороший билет, если первый студент достал хороший билет. Если первый студент достал хороший билет, значит у нас осталось 4 хороших и еще 20 плохих.

$P(A \mid B_1) = \frac{4}{24} = \frac{1}{6}$

Вероятность того, что первый вытащил плохой билет: $P(B_2) = \frac{20}{25} = \frac{4}{5}$

Давайте посмотрим с какой вероятностью второй студент достанет хороший билет, если первый студент достал плохой билет. Если первый студент достал плохой билет, значит у нас осталось 5 хороших и еще 19 плохих.

$P(A \mid B_2) = \frac{5}{24}$


$${\sf P}(A) = {\sf P}(A \mid B_1){\sf P}(B_1) + {\sf P}(A \mid B_2){\sf P}(B_2) = \frac{1}{30} + \frac{1}{6} = \frac{1}{5}$$

__Ответ:__ Для всех студентов вероятность составит $\frac{1}{5}$

<h3 style="color:#2E8B57">Задача № 6</h3>

Брошено 3 игральных кости (кость - кубик с шестью гранями, на которых написаны числа от 1 до 6). Найдите вероятность того, что на всех костях выпала "шестерка" (выпадения всех граней равновероятны) при условии, что

(a) по крайней мере на одной кости выпала "шестерка";

(b) по крайней мере на двух костях выпало равное количество очков.

<h4 align='center'>Решение:</h4>

В нашем случае пространство элементарных исходов состоит из всех возможных реализаций броска $\Omega = \{\omega_{1,1,1}, \ldots, \omega_{6, 6, 6}\}$, а в промежутке записаны всевозможные способы выпадения 3 костей, где они могут как совпадать, так и не совпадать. $\mid \Omega \mid = 6^3 = 216$ комбинаций. Другими словами, событие $A = \{ \omega_{6, 6, 6} \}$ состоит в том, что реализовались 3 шестёрки.

Событие $A$ - все 3 кости выпали 6 к верху;

$${\sf P}(A \mid B) = \frac{{\sf P}(A \cap B)}{{\sf P}(B)},  {\sf P}(B) \ne 0 $$

**Событие $B_1$ - по крайней мере на одной кости выпала "шестерка";**

Необходимо найти условную вероятность $P(A \mid B_1)$. Для начало найдём вероятность условия $P(B_1)$

Давайте напишем так: в знаменателе, конечно, находится 216, потому что мы по-прежнему работаем с классическим определением вероятности, а в числителе мы напишем так: мы возьмем общее количество элементарных исходов и из них вычтем те элементарные исходы, которые, напротив, не благоприятствуют нашему событию $B_1$. Что это за элементарные исходы, которые ему не благоприятствуют? Это значит, что ни одна кость не выпала шестеркой кверху. Отрицание события $B_1$.

$P(B_1) = \frac{216 - 5^3}{216} = \frac{91}{216}$

А что такое вероятность A, пересеченного с B1? Вообще, что такое A, пересеченное с B1? Смотрите, B1 говорит, что хотя бы одна выпала шестеркой, а A говорит, что все три выпали шестеркой. Ну ясно, что в пересечении находится ровно вот это самое элементарное событие, которое благоприятствует A. 

Ну и вероятность выпадения 3 шестёрок составит $P(A \cap B_2)  = \frac{1}{216}$.

$P(A \mid B_1) = \frac{\frac{1}{216}}{\frac{91}{216}} = \frac{1}{91}$

**Событие $B_2$ - по крайней мере на двух костях выпало равное количество очков.**

Необходимо найти условную вероятность $P(A \mid B_2)$. Для начало найдём вероятность условия $P(B_2)$.

В знаменателе находится 216, а числитель посчитаем ровно по тому же принципу, то есть возьмем общее число элементарных исходов и вычтем из него те элементарные исходы, которые не благоприятствуют событию $B_2$. Что значит «не благоприятствуют»? Видите, событие B2 состоит в том, что хотя бы у двух значения совпадают. Значит, отрицание этого – никакие значения не совпадают вовсе, то есть все три числа различны. Ну понятно, первое число можно выбрать любым из шести способов, второе – из пяти (раз оно не должно совпадать с первым), и третье – из четырех. 

$P(B_2) = \frac{216 - 5 \cdot 5 \cdot 4}{216} = \frac{96}{216}$

Пересечение этих двух событий снова состоит ровно из того элементарного исхода, который единственный благоприятствует событию A, то есть из трех шестерок. 

Ну и вероятность выпадения 3 шестёрок составит $P(A \cap B_2)  = \frac{1}{216}$.

$P(A \mid B_1) = \frac{\frac{1}{216}}{\frac{96}{216}} = \frac{1}{96}$

<h3 style="color:#2E8B57">Задача № 7</h3>

Есть множество, которое состоит из $n$ элементов $\{1, 2, \ldots, n \}$. Из этого множества мы выбираем два случайных подмножества по схеме упорядоченного выбора с возвращением. 

Это значит, что мы из всех возможных подмножеств, которых, конечно, 2 в степени n штук. Сначала выбираем некоторое множество А, запоминаем как его выбрали. После чего, возвращаем его обратно в кучу, снова тщательно кучу перемешиваем и оттуда наугад достаем множество В. Которое, естественно, тем самым, с какой-то вероятностью может даже совпасть.

Найдите условную вероятность того, что множества $A$ имеет мощность $l_1$ и $B$ имеет мощность $l_2$ соответственно при условии, что эти множества не пересекаются $A \cap B = \varnothing$.  

<h4 align='center'>Решение:</h4>

Таким образом, пространство элементарных исходов — это есть множество всех упорядоченных пар подмножеств. Упорядоченных пар подмножеств А и В. Ну раз подмножество А можно выбрать 2 в степени n способами, и мы его потом возвращаем, то множество В тоже можно выбрать 2 в степени n способами, откуда получаем, что мощность омега — это есть 4 в степени n.

Нам необходимо найти вот такую условную вероятность:

$P(|A| = l_1, |B| = l_2 \mid A \cap B = \varnothing)$

В числителе мы пишем вероятность пересечения событий. То есть что и вот это выполнено, и еще, к тому же, выполнено вот это. Одновременно и мощности такие, и множества не пересекаются. Ну вот давайте так через запятую напишем. Запятая — это просто пересечение событий.

А в знаменателе, конечно, стоит вероятность условия. То есть вероятность того, что А, пересеченное с В, — пусто.

$P(|A| = l_1, |B| = l_2 \mid A \cap B = \varnothing) = \frac{P(|A| = l_1, |B| = l_2, A \cap B = \varnothing)}{P(A \cap B) = \varnothing}$

Поскольку каждая из этих вероятностей классическая, то есть мы всякий раз количество благоприятствующих исходов делим на общее количество исходов, на мощность омега большого, то и сверху стоит в знаменателе 4 в степени n, и снизу стоит в знаменателе 4 в степени n. 4 в степени n — это просто мощность омега большого, и эти четверки в n степени благополучно сократятся.

$\frac{|A| = l_1, |B| = l_2, A \cap B = \varnothing}{A \cap B = \varnothing}$

**Рассчёт числителя:**

Числитель считается совсем легко, потому что смотрите: давайте сначала будем выбирать множество А. Но как мы должны его выбрать? Мы его должны выбрать так, чтобы его мощность равнялась $l_1$, и выбирается оно, конечно, из n-элементов. То есть мы просто берем С из n по $l_1$, и это есть количество способов зафиксировать множество А. 

После того как множество А зафиксировано, мы обязаны выбрать множество В вот такой мощности, $l_2$, но при этом так, чтобы оно с А не пересекалось. У нас осталось n−l1 элементов, которые множеству А не принадлежат. И вот из этих n−l1 элементов мы вольны выбрать произвольный l2 так, чтобы у нас на выходе получилось как раз l2-элементное множество В, которое с А не пересекается. 

$C_n^{l_1} \cdot C_{n - l_1}^{l_2}$

**Рассчёт знаменателя по биному:**

То есть будет сумма, и мы будем смотреть на мощность множества А. Ну давайте ее как-нибудь, k обозначим. Она меняется от нуля до n теоретически. И вот при каждом таком k мы можем выбрать множество А С из n по k способами, а множество В, соответственно, мы можем выбрать вообще как угодно, ведь его мощность не регламентирована. Ну у нас остается n−k элементов после того, как мы зафиксировали k элементов для множества А, и вот только из этих n−k элементов мы можем извлекать В, коль скоро считаем, что они не пересекаются. Поэтому здесь возникает 2 в степени n−k. 

$\sum^{n}_{k=0} {C_n^k \cdot 2^{n-k}} = 3^n$

То что здесь написано, это просто бином Ньютона, а именно, вот такое вот выражение: это $(1 + 2)^n$. Ну действительно, если раскрыть с помощью стандартного бинома Ньютона.

**Рассчёт знаменателя более простым способом:**

У нас есть 2 множества, А и В, которые вовсе не пересекаются. Давайте каждой паре таких множеств однозначно сопоставим некую последовательность из нулей, единиц и двоек, которая бы имела длину n. 

$\{1, 2, \ldots, n\}$

Смотрим на элементы множества 1, 2 и так далее n, из которого выбирались наши подмножества А и В. Вот у нас 1, 2 и так далее n. Где-то здесь живет подмножество А где-то. Где-то живет подмножество В, и они не пересекаются.

Давайте каждому элементу множества А сопоставим еденицу. Каждому элементу множества B сопоставим двойку. А если, в свою очередь, элемент не попадает ни в множество А, ни в множество В, тогда нарисуем какой-нибудь нолик. И вот у нас каждой паре множеств однозначно соответствует такая последовательность.

Ну, в итоге получаем, что количество способов выбрать вот такую пару — это в точности количество способов построить такую последовательность из нулей, единиц и двоек. На каждую позицию ставим любое из трех чисел в позиции n. Ответ — 3 в степени n.


**Ответ:** $\frac{C_n^{l_1} \cdot C_{n - l_1}^{l_2}}{3^n}$

<h3 style="color:#2E8B57">Задача № 1</h3>

Верно ли, что событие $\Omega$ (наступление произвольного элементарного исхода) является независимым с любым другим событием?

<h4 align='center'>Решение:</h4>

Пусть $A$ - произвольное событие. Тогда $A\cap\Omega=A$. Кроме того, по определению вероятности ${\sf P}(\Omega)=1$. Следовательно, ${\sf P}(A\cap\Omega)={\sf P}(A)={\sf P}(A){\sf P}(\Omega)$. Поэтому события $A$ и $\Omega$ являются независимыми.

**Ответ:** верно.

<h3 style="color:#2E8B57">Задача № 2</h3>

 Верно ли, что для независимых событий $A_1, A_2$ и произвольного события $B$ выполнено ${\sf P}(A_1\cap A_2|B)={\sf P}(A_1|B){\sf P}(A_2|B)$?

<h4 align='center'>Решение:</h4>

Если $B=A_1\cup A_2$, то

${\sf P}(A_1\cap A_2|B)=\frac{{\sf P}(A_1\cap A_2)}{{\sf P}(A_1\cup A_2)}=\frac{{\sf P}(A_1){\sf P}(A_2)}{{\sf P}(A_1\cup A_2)}\neq\frac{{\sf P}(A_1){\sf P}(A_2)}{({\sf P}(A_1\cup A_2))^2}={\sf P}(A_1|B){\sf P}(A_2|B)$, 
если ${\sf P}(A_1\cup A_2)\neq 1$. 

Если $B=\Omega$, то условная вероятность совпадает с безусловной. Поэтому равенство ${\sf P}(A_1\cap A_2|B)={\sf P}(A_1|B){\sf P}(A_2|B)$ выполнено. 

Выберем теперь в качестве $B$, скажем, событие $A_1$. Тогда ${\sf P}(A_1|B)=\frac{{\sf P}(A_1)}{{\sf P}(A_1)}=1$, ${\sf P}(A_2|B)=\frac{{\sf P}(A_1\cap A_2)}{{\sf P}(A_1)}=\frac{{\sf P}(A_1){\sf P}(A_2)}{{\sf P}(A_1)}={\sf P}(A_2)$. 

Наконец, ${\sf P}(A_1\cap A_2|B)=\frac{{\sf P}(A_1\cap A_2)}{{\sf P}(A_1)}={\sf P}(A_2)$. Поэтому равенство ${\sf P}(A_1\cap A_2|B)={\sf P}(A_1|B){\sf P}(A_2|B)$ также выполнено.

**Как была получена формула?**

По определению, $P(A|B) = \frac{P(A\cap B)}{P(B)}$. Если $A = A_1 \cap A_2$, $B=A_1\cup A_2$ то $A\cap B = (A_1 \cap A_2) \cap (A_1 \cup A_2) = A_1 \cap A_2$, откуда следует первая формула. Вторая формула следует из того, что $P(A_1 \cap A_2) = P(A_1) P(A_2)$ (следует из независимости этих событий).  

А справа расписано определение $P(A_1|B)$.


**Ответ:** неверно. Хотя это верно при $B=\Omega$ и в некоторых других ситуациях.

<h3 style="color:#2E8B57">Задача № 3</h3>

Пусть $\Omega$ - множество элементарных исходов с вероятностью ${\sf P}, C\subseteq\Omega$ - произвольное событие. Определите множество элементарных исходов $\tilde\Omega$ так, чтобы любое событие $A\subseteq\tilde\Omega$ являлось также и событием из $\Omega$, причем вероятность события $A$ (как события в $\tilde\Omega$) была равна ${\sf P}(A|C)$.

<h4 align='center'>Решение:</h4>
    
Если $\tilde\Omega=\Omega$, то вероятность события $C$ в $\tilde\Omega$ равна ${\sf P}(C)\neq 1= {\sf P}(C|C)$ (при условии что $C\neq\Omega$).

Положим теперь $\tilde\Omega=C$. Тогда если $A\subset\tilde\Omega$, то вероятность события $A$ в этом новом множестве элементарных исходов равна $\frac{|A|}{|C|}=\frac{|A\cap C|}{|C|}=\frac{|A\cap C|/|\Omega|}{|C|/|\Omega|}=\frac{{\sf P}(A\cap C)}{{\sf P}(C)}={\sf P}(A|C)$.

Если $\tilde\Omega=\overline{C}$, то вероятность события $\overline{C}$ в $\tilde\Omega$ равна $1\neq 0= {\sf P}(\overline{C}|C)$ (при условии, что событие C не является пустым множеством).  

**Ответ:** $\tilde\Omega=C$
   

<h3 style="color:#2E8B57">Задача № 4</h3>

<h3 style="color:#2E8B57">Задача № 5</h3>

Пусть $\Omega=\{\omega_1,\ldots,\omega_6\}$ - множество элементарных исходов. Сколько существует таких событий $A$, что из наступления $\omega_1$ следует наступление $A$?

<h4 align='center'>Решение:</h4>
    
Из наступления элементарного исхода $\omega_1$ следует наступление $A$ тогда и только тогда, когда $\omega_1\in A$. Поэтому любое такое событие $A$ является объединением элемента $\omega_1$ с произвольным подмножеством множества $\{\omega_2,\ldots,\omega_6\}$. Так как количество подмножеств в $\{\omega_2,\ldots,\omega_6\}$ (включая пустое подмножество и все множество $\{\omega_2,\ldots,\omega_6\}$ равно $2^5=32$, то и количество всевозможных событий $A$, удовлетворяющих условию задачи, равно $32$.
    
**Ответ:** 32.

<h3 style="color:#2E8B57">Задача № 6</h3>

На одной из книжных полок библиотеки стоит $20$ книг. У некоторых из них красная обложка, у других - синяя, у третьих на обложке присутствует как красный цвет, так и синий, а остальные книги имеют зеленые обложки (при этом на полке может не быть книг любого из цветов). Читатель случайным образом берет с этой полки одну из книг (каждую книгу с равной вероятностью). Известно, что он с одинаковой вероятностью $\frac{1}{2}$ выбирает книгу, на обложке которой присутствует красный цвет, и книгу, на обложке которой присутствует синий цвет. Каково количество значений, которое может принимать вероятность выбора книги с зеленой обложкой?

<h4 align='center'>Решение:</h4>
    
Пусть $\Omega=\{1,\ldots,20\}$, где число $i$, $i\in\Omega$, соответствует выбору $i$-ой книги. Пусть $A$ - событие, заключающееся в том, что читатель выбирает книгу, на обложке которой присутствует красный цвет, а $B$ - книгу, на обложке которой присутствует синий цвет. По условию задачи ${\sf P}(A)={\sf P}(B)=\frac{1}{2}$. Тогда $|A|=|B|=\frac{1}{2}|\Omega|=10$. Пусть $C$ - событие, заключающееся в том, что читатель выберет книгу с зеленой обложкой. Тогда $C=\Omega\setminus(A\cup B)$. Следовательно,

${\sf P}(C)=\frac{|C|}{|\Omega|}=\frac{|\Omega|-|A\cup B|}{|\Omega|}=\frac{20-|A\cup B|}{20}$.

Мощность множества $A\cup B$ может принимать все натуральные значения от $10$ до $20$ $(|A\cup B|=10$, если $A=B$, $|A\cup B|=11$, если в пересечении множеств $A$ и $B$ находится $9$ элементов, $\ldots, |A\cup B|=20$, если множества $A$ и $B$ не пересекаются). Следовательно, ${\sf P}(C)$ может принимать $11$ значений $\frac{0}{20},\frac{1}{20},\ldots,\frac{10}{20}$. 
    
**Ответ:** $11$.

<h3 style="color:#2E8B57">Задача № 7</h3>

Среди $25$ экзаменационных билетов $5$ "хороших". Три студента по очереди берут по одному билету (первый студент берет каждый из билетов с одинаковой вероятностью, второй - равновероятно любой из оставшихся $24$, третий - равновероятно любой из оставшихся $23$). Найдите вероятность того, что третий студент взял "хороший" билет.

<h4 align='center'>Решение:</h4>
    
Пусть $(1,\ldots,25)$ - номера экзаменационных билетов. Без ограничения общности будем считать, что $1,\ldots,5$ - номера "хороших" билетов. Тогда пространство элементарных исходов $\Omega=\{\omega_{1,2,3},\ldots,\omega_{25,24,23}\}$, где $\omega_{i,j,k}$ - исход, при котором первый студент вытянул билет с номером $i$, второй - билет с номером $j$, третий - с номером $k$. Обозначим $B_i$, $i\in\{1,2,3\}$, событие, заключающееся в том, что $i$-ый студент вытянул "хороший" билет. Тогда $B_i=\{\omega_{j_1,j_2,j_3}:\,j_i\in\{1,2,3,4,5\}\}$. Очевидно, между этими тремя множествами существует взаимно однозначное соответствие. Так, например, отображение $B_3\rightarrow B_1$ можно построить по следующему правилу: $\omega_{j_1,j_2,j_3}\rightarrow\omega_{j_3,j_2,j_1}$. Тогда $|B_1|=|B_2|=|B_3|$. Поэтому

${\sf P}(B_1)={\sf P}(B_2)={\sf P}(B_3)=\frac{1}{5}$.

Таким образом, все студенты вытягивают "хороший" билет с одной и той же вероятностью.
    
**Ответ:** $\frac{1}{5}$

<h3 style="color:#2E8B57">Задача № 8</h3>

На этаже школы есть кабинеты математики, физики, химии и биологии (всего $4$ кабинета). Шесть школьников случайным образом расходятся по кабинетам. Какова вероятность того, что ни один кабинет не останется пустым?

<h4 align='center'>Решение:</h4>
    
В данном случае множество элементарных исходов $\Omega$ представляет собой множество $6$-элементных последовательностей $(x_1,\ldots,x_{6})$, $i$-ые элементы которых соответствуют $i$-му школьнику и принимают четыре значения М, Ф, Х и Б в зависимости от того, какой кабинет выбрал школьник. Тогда $|\Omega|=4^{6}$. Пусть $A$ - событие заключающееся в том, что ни один кабинет не остался пустым, а $\overline{A}$ - множество всех остальных элементарных исходов, т.е. событие, заключающееся в том, что хотя бы один кабинет остался пустым. Тогда $|\overline{A}|=4\cdot 3^{6}-6\cdot 2^{6}+4∣$, так как имеется $4$ способа выбрать пустой кабинет (здесь мы воспользовались формулой включений-исключений). Окончательно получаем

${\sf P}(A)=1-{\sf P}(\overline{A})=1-\frac{|\overline{A}|}{|\Omega|}=1-\frac{4\cdot 3^{6}-6\cdot 2^{6}+4}{4^{6}}=1-\frac{3^6-6\cdot 16+1}{1024}=1-\frac{729-96+1}{1024}=\frac{390}{1024}=\frac{195}{512}$.
    
**Ответ:** $\frac{195}{512}$

<h3 style="color:#2E8B57">Задача № 9</h3>

Из множества всех графов на $5$ вершинах (без петель и ориентации) наугад берется один граф (выборы графов равновероятны). Найдите вероятность того, что этот граф является простым циклом с листом.

<h4 align='center'>Решение:</h4>
    
Положим $n=5$. Пространство элементарных исходов $\Omega$ - это множество всех неориентированных графов без петель на множестве вершин $\{1,\ldots,n\}$. Так как каждая пара различных вершин либо образует, либо не образуют ребро (две возможности для $C_n^2$ пар вершин), то $|\Omega|=2^{C_n^2}$. Пусть $A$ (интересующее нас событие) - множество простых циклов с листом, входящих в $\Omega$. Тогда $|A|=\frac{n!}{2}$, так как количество способов выбрать лист равно $n$, число способов выбрать соединенную с ним вершину равно $n-1$, а, наконец, число способов выбрать оставшиеся вершины простого цикла равно $\frac{(n-2)!}{2}$ (число способов расположить $n-2$ вершины в цепи равно $(n-2)!$, но так как наша цепь присоединяется ребрами ко второй выбранной вершине обоими концами, то мы каждый способ построить цикл посчитали дважды). Окончательно получаем ответ

${\sf P}(A)=\frac{|A|}{|\Omega|}=\frac{n!/2}{2^{C_n^2}}=\frac{5\cdot 4\cdot 3}{2^{10}}=\frac{15}{256}$.

**Ответ:** $\frac{15}{256}$

<h3 style="color:#2E8B57">Задача № 1</h3>

Имеется $10$ белых и $10$ черных шаров. Вам предлагают каким-то образом разложить эти шары по двум урнам. Далее случайно выбирается одна из урн (выборы урн равновероятны), а из нее вытаскивается шар (для всех шаров, лежащих в урне, вероятности их выбора одинаковы). Если он оказывается белым, то Вы получаете приз. Как нужно разложить шары по урнам, чтобы вероятность выиграть приз была наибольшей?

<h4 align='center'>Решение:</h4>

 Пусть в первой корзине $a_1$ белых и $b_1$ черных шаров, а во второй - $a_2$ и $b_2$ соответственно, и событие $A$ состоит в том, что мы выбрали белый шар. Тогда вероятность получить приз

${\sf P}(A) = \frac{1}{2} \left( \frac{a_1}{a_1+b_1} + \frac{a_2}{a_2+b_2} \right)$.

Требуется найти такие $a_1$ и $b_1$, при которых эта вероятность максимальна. Если $a_1 = b_1$, то ${\sf P}(A) = \frac{1}{2}$.

Пусть теперь без ограничения общности $a_1 > b_1$, и, значит, $a_2 < b_2$. Тогда легко убедиться, что при $b_1 \geq 1$ выполнены неравенства  $\frac{a_1}{a_1+b_1} < \frac{a_1-1}{(a_1-1)+(b_1-1)}$ и  $\frac{a_2}{a_2+b_2} < \frac{a_2+1}{(a_2+1)+(b_2+1)}$. Это означает, что, перекладывая один черный и один белый шар из первой корзины во вторую, мы увеличим вероятность выигрыша. Рассмотрим теперь оптимальные $a_1^*$ и $b_1^*$. В силу предыдущего рассуждения можно считать, что $a_1^* > b_1^* = 0$. Тогда

${\sf P}(A) = \frac{1}{2} \left( 1 + \frac{10-a_1}{(10-a_1)+10} \right) \leq \frac{1}{2} \left( 1 + \frac{9}{19}\right)$ и равенство достигается при $a_1 = 1$.

Итак, $\max_{a_1,a_2,b_1,b_2} {\sf P}(A) = \frac{14}{19}$.

**Ответ:** $\frac{14}{19}$

<h3 style="color:#2E8B57">Задача № 2</h3>

На шахматной доске размера $n\times n$  случайно размещают $n$ ладей (ладья может двигаться на любое число полей по горизонтали и по вертикали). Найдите вероятность того, что ладьи не бьют друг друга.

<h4 align='center'>Решение:</h4>

 Заметим, что количество способов расставить $n$ ладей на поле $n \times n$ равно $C_{n^2}^n$. Найдем количество способов расставить ладьи так, чтобы они не били друг друга. В такой расстановке в каждом столбце и каждой строке ровно по одной ладье. В первом столбце ладью можно поставить на любое место, во втором - на все кроме одного, в третьем - на все кроме двух и так далее, положение последней ладьи определяется однозначно. Соответственно, количество расстановок ладей подобным образом равно $n!$. Таким образом, если $A$ - событие, заключающееся в том, что никакие две ладьи не бьют друг друга, то, с учетом того, что все расстановки равновероятны, мы получаем, что интересующая нас вероятность равна ${\sf P}(A) = \frac{n!}{C_{n^2}^n}$.  


**Ответ:** $\frac{n!}{C_{n^2}^n}$
