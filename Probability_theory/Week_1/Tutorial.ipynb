<h1 style="color:#008B8B">Классическое определение вероятности</h1>

**Пример.** Игральная кость представляет собой кубик, грани которого пронумерованы от одного до шести, обычно точками. Кость бросают на стол и фиксируют число на той грани, которая оказывается сверху в результате броска. Можно предположить, что форма кубика идеальная и на ребро он упасть не может. В таком случае всего имеют место $6$ различных исходов такого броска, причем для них выполняются следующие свойства:

* Исходы образуют полную группу событий, то есть хотя бы один из исходов обязательно произойдёт (Никаких вариантов быть не может, кроме перечисленных);
* Исходы попарно несовместны, то есть никакие два исхода одновременно произойти не могут;
* Все исходы равновероятны.

В результате рассмотрения игральной кости появилось классическое определение вероятности.

Исходы, обладающие перечисленными свойствами "образуют полную группу событий", "попарно несовместны" и "равновероятны" называются элементарными исходами.

**Классическое определение вероятности.** Пусть различные исходы $\omega_1, \ldots, \omega_n$ 
некоторого эксперимента являются элементарными. Тогда по определению вероятностью каждого такого исхода называют величину:

$$\large P(\omega_i) = \frac{1}{n}$$

**Пример.** Игральные карты. Пусть дана колода из $36$ тщательно перетасованных карт. В этом случае элементарными исходами являются различные последовательности карт внутри колоды, то есть различные варианты их взаимного расположения. Эти исходы, которых всего $36!$, образуют полную группу событий и попарно несовместны. Вероятность каждого исхода:

$$\large P(\omega_i) = \frac{1}{36!}$$

Часто бывает необходимым рассматривать события несколько более сложные, чем элементарные исходы. 

**Например**, событие $A$ состоит в том, что "в результате броска игральной кости выпало четное число очков". Такому событию соответствуют три благоприятных элементарных исхода $\omega_2, \omega_4, \omega_6$. 

**Вероятность события.** Тогда вероятность некоторого события $A$ полагают равной отношению количества элементарных исходов $k$, которые благоприятствуют событию, к количеству всех возможных элементарных исходов $n$:

$$\large P(A) = \frac{k}{n} = \frac{3}{6} = \frac{1}{2}$$

<h1 style="color:#008B8B">Основные свойства вероятности</h1>

Если принять классическое определение как определение вероятности, то несложно получить следующие свойства.

Под событием удобно понимать множество таких элементарных исходов, которые ему благоприятствуют. Множество же всех элементарных исходов $\Omega=\{\omega_1,\ldots,\omega_n\}$ называется пространством элементарных исходов, где $n$ — количество всевозможных элементарных событий.

**Свойство 1.** Вероятность ${\sf P}(\Omega)$ события $\Omega$, которое состоит в том, что реализуется любой элементарный исход, равняется $1$. Такое событие называется достоверным, так как происходит с веротяностью равной $1$.

**Свойство 2.** В классическом определении вероятностей, нулевую вероятность имеет событие, которое соответствует пустому множеству элементарных исходов:  ${\sf P}(A)=0 \Leftrightarrow A=\varnothing$.

**Свойство 3.** Вероятность дизъюнктного объединения двух событий $A$ и $B$, то есть таких событий, что $A$ и $B$ имеют пустое пересечение, равна сумме их вероятностей ${\sf P}(A\sqcup B) = {\sf P}(A) + {\sf P}(B)$

**Свойство 4.** Вероятность объединения двух произвольных событий $A$ и $B$ равна ${\sf P}(A\cup B) = {\sf P}(A) + {\sf P}(B) -{\sf P}(A\cap B)$

**Свойство 5.** Более того, верна формула включений-исключений, которая была доказана в курсе комбинаторики:

$P(A_1 \cup \ldots \cup A_k) = P(A_1) + ... + P(A_k) - P(A_1 \cap A_2) - P(A_1 \cap A_3) - \ldots - P(A_{k-1} \cap A_k) + \ldots + (-1)^{k-1} P(A_1 \cap \ldots \cap A_k) $

**Свойство 6.** ${\sf P}(A_1 \cup \ldots \cup A_k) \leq {\sf P}(A_1)+ \ldots+ {\sf P}(A_k)$.

Вероятность $\cup$ не превосходит суммы вероятностей события $A_1, A_2, \ldots , A_k$

**Свойство 7.** Отрицанием события $A$ называется событие $\bar{A}$, которому благоприятствуют только такие исходы, которые не благоприятствуют событию $A$. Тогда верно следующее:

Вероятность события $\bar{A} = \Omega \backslash A$ равна ${\sf P}(\bar{A})=1-{\sf P}(A)$

### Дополнительный вопрос по свойствам:

В классическом определении вероятности все элементарные исходы равновероятны. Из какого (ровно одного) свойства следует определение вероятности произвольного события $A$, состоящего из $k$ элементов ${\sf P}(A)=\frac{k}{n}$

__Ответ:__ ${\sf P}(A\sqcup B) = {\sf P}(A) + {\sf P}(B)$

Заметим, что из данного свойства, очевидно, следует аналогичное свойство для k непересекающихся событий $A_1,\ldots,A_k$: ${\sf P}(A_1\sqcup\ldots\sqcup A_k)={\sf P}(A_1)+\ldots+{\sf P}(A_k)$

Пусть $A=\{\omega_{i_1},\ldots,\omega_{i_k}\}$ Тогда 

${\sf P}(A)={\sf P}(\omega_{i_1})+\ldots+{\sf P}(\omega_{i_k})=\frac{1}{n}+\ldots+\frac{1}{n}=\frac{k}{n}$

Легко заметить, что любого другого одного свойства недостаточно, чтобы доказать это равенство.

<h3 style="color:#2E8B57" align="right">Задача о раскраске пятнадцати пятиэлементных множеств</h3>

Рассмотрим $15$ совершенно произвольных подмножеств некоторого тридцати элементного множества $M_1, \ldots, M_{15} \subset \{1, 2, \ldots, 30\}$ (У нас есть множество состоящее из $30$ натуральных чисел и в этом множестве, выбраны и зафиксированы какие-то $15$ подмножеств). Мощность каждого подмножества, количество элементов в нём $|M_i| = 5$. В зависимости от того, как мы зафиксируем пятиэлементые подмножества, они могут иметь пересечения между собой, так как сумарное количество элементов $75$, что больше чем $30$ исходного множества.

Верно ли, что какова бы ни была зафиксированная нами совокупность пятерок $M_1, \ldots, M_{15}$, всегда можно найти расскраску элементов множества $\{1, 2, \ldots, 30\}$ в два цвета (Каждому элементу присвоить либо красный, либо синий цвет), при которой все множества $M_i$ (каждая пятёрка) не одноцветные, то есть каждое множество содержит как элемент красного, так и элементы синего цвета.

Зададим вопрос более простыми словами. Можно ли аудиторию из $30$ человек разделить на две непересекающиеся части, рассадить по двум аудиториям, так чтобы в первой аудитории присутствовало по представителю из каждого множества лучших $M_1, \ldots, M_{15}$ ($M_1$ - пятёрка лучших учеников, которые лучше всех знают комбинаторику и так далее), так и во второй аудитории присутствовало хотябы по одному представителю из каждого множества лучших. Существует ли всегда такое разбиение аудитории на две непересекающиеся части.

Как в первой аудитории присутсвуют $15$ лучших математиков, физиков..., так и во второй аудитории найдутся $15$ других студентов, которые лучше всех знаю математику, физику. 

<h4 align='center'>Решение:</h4>

Рассмотрим случайную расскраску $X$ множества $\{1, 2, \ldots, 30\}$. Пространство $\Omega$ - это и есть множество всех расскарсок. Расскрасок $2^{30}$, мы каждому элементу независимо от остальных можем присвоить либо красный, либо синий цвет. $\Omega = \{X_1, \ldots, X_{2^{30}}\}$ и вероятность возникновения любой расскраски можно считать равной $P(X_i) = \frac{1}{2^{30}}$. 

У нас зафиксирована совокупнось пятёрок $M_1, \ldots, M_{15}$, возьмём любое множество $M_i$ из этой совокупности и сопоставим этому множеству событие $A_i$ состоящие в том, что в случайной расскраске множество $M_i$ одноцветное. 

Чему тогда равна вероятность $P(A_i)$? Надо найти количество всех элементарных исходов, которые благоприятствуют нашему событию, то есть количество всех раскрасок, в которых данное множество $M_i$-тое одноцветно, и разделить на общее количество раскрасок, то на $2$ в $30$-той степени. В знаменателе стоит $2^{30}$. В числителе: есть всего два варианта, как сделать пятиэлементное множество одноцветным, покрасить $5$ элементов в карсный цвет или в синий и есть ещё определённое количество способов покрасить все остальные элементы, которых $25$. И количество способов расскрасить остальные элемент эт $2^{25}$.

$\large P(A_i) = \frac{2\cdot 2^{25}}{2^{30}} = \frac{1}{16}$

Вероятность объединения $A_i$ от $1$ до $15$ $P(\bigcup\limits_{i=1}^{15} {A_i})$ не превосходит суммы от $1$ до $15$ вероятностей $A_i$ $P(\sum\limits_{i=1}^{15} {A_i})$, где $P(A_i) = \frac{1}{16}$

$\large P(\bigcup\limits_{i=1}^{15} {A_i}) \le P(\sum\limits_{i=1}^{15} {A_i}) < 1$

Что обозначет событие объединения $A_i$? $A_i$ состоит из расскрасок, где $M_i$ одного цвета (например, $A_1$ - это расскраска в которых $M_1$ одного цвета, $A_2$ - это расскраска в которых $M_2$ целиком одного цвета). Следовательно это объедение состоит в том, что $M_1$ одноцветно, либо $M_2$ одного цвет и так далее до $M_15$,  значит, это событие состоящее в том, что в случайно расскраске найдётся одноцветное множество среди $15$ зафиксированных. 

Воспользуемся свойством $7$, которое мы определили:

Отрицание события $P(\bigcup\limits_{i=1}^{15} {A_i}) > 0$, так как само событи строго меньше $1$, а вероятность его отрицания строго больше 0 по $7$ свойствую. В чём состоит отрицание, само событие состоит в том, что существует одноцветное множество, а отрицание этого события состоит в том, что нет ни одного одноцветного множества.

Вспоминаем наше условие задачи, "всегда можно найти расскраску элементов множества $\{1, 2, \ldots, 30\}$ в два цвета, при которой все множества $M_i$ не одноцветные". Наше отрицательное событие говорит о том, что все множества не одноцветные и это происходит с положительной вероятностью. Следовательно, по свойству $2$, вероятность равняетя нулю, тогда и только тогда, когда $A$ это пустое множество. Значит, отрицание нашего события существует, так как вероятность его больше нуля и ему сопоставленно определённое множество где все множества $M_i$ не одноцветные.

<h1 style="color:black" align="center">Условная вероятность</h1>

<h1 style="color:#008B8B">Условная вероятность</h1>

**Пример.** Пусть игральную кость бросают один раз. Событие $A$ состоит в том, что выпало $1$ очко, а $B$ — в том, что выпало четное число очков. Требуется найти вероятность ${\sf P}(A \mid B)$ события $A$, если уже известно, что событие $B$ произошло. Как изменится ответ, если в качестве события $A$ взять событие, состоящее в том, что выпало $2$ очка?

Ответ на первый вопрос кажется естественным: ${\sf P}(A \mid B) = 0$, поскольку  мы знаем, что кость выпала чётной стороной, зачит выпасть стороной $1$ она не может.

Второй вопрос задачи. Если выпало четное число очков, то пространство элементарных исходов сузилось до множества $B$. Множество элементарных исходов ($B$) состоит из $3$-х элементов: выпало $2$ очка, выпало $4$ очка и выпало $6$ очков, причем, только один из исходов благоприятный и состоит в том, что кость выпала числом $2$ кверху. Таким образом, искомая вероятность равняется ${\sf P}(A \mid B)=\frac{1}{3}$.

### Вывод формулы условной вероятности:

В общем случае условная вероятность определяется следующим образом. Пусть в пространстве элементарных исходов $\Omega = \{\omega_1,\ldots,\omega_n\}$ задано некоторое непустое событие $B=\{\omega_{i_1} , \ldots, \omega_{i_k}\}$ $k \geqslant 1$. Пусть, также задано другое событие $A$. Тогда вероятность ${\sf P}(A|B)$ события $A$ при условии, что событие $B$ уже реализовалось, можно определить следующим образом:

Действительно, если событие $B$ уже реализовалось, то $B$ становится новым множеством элементарных исходов, а множеством благоприятных исходов станет множество $A\cap B$. 

То есть, в знаменатель поставим количество элементарных исходов, которые благоприятсвуют событию $B$, а это и есть число $k$ или же мощность $|B|$, а в числитель надо поставить те из вот этих благоприятствующих событию $B$ исходов, которые, в свою очередь, благоприятствуют событию $A$. Этот естественно обозначить как мощность пересечения $A$ и $B$, то есть нам нужно посмотреть, какие элементарные исходы, благоприятствующие событию $A$, есть среди элементарных исходов, успевших реализоваться, коль скоро мы знаем, что $B$ произошло:

$$\large {\sf P}(A \mid B) = \frac{|A \cap B|}{|B|}$$

Делим на мощность $|\Omega|$, чтобы получить вероятности:

$$\large {\sf P}(A \mid B) = \frac{|A \cap B|}{|B|} = \frac{|A \cap B| \frac{1}{|\Omega|}}{|B| \frac{1}{|\Omega|}} = \frac{{\sf P}(A \cap B)}{{\sf P}(B)},  {\sf P}(B) \ne 0 $$

### `Формула услованой вероятности:`

$$\large {\sf P}(A \mid B) = \frac{{\sf P}(A \cap B)}{{\sf P}(B)},  {\sf P}(B) \ne 0$$

**Замечание:** Если множество $B$ пустое, то по соглашению условную вероятность ${\sf P}(A|B)$ полагают равной нулю.

### `Теорема умножения:`

Если справа и слева формулу условной вероятности домножить на ${\sf P}(B)$, тогда мы получим:

$$\large {\sf P}(A \mid B){\sf P}(B)={\sf P}(A \cap B)$$

<h1 style="color:#008B8B">Независимость событий</h1>

### Понятие независимости событий

Пусть в некотором пространстве элементарных исходов $\Omega$ заданы два события $A$ и $B$, причем $B$ не является пустым. Условие того, что событие $A$ независит от события $B$, имеет следующий вид:

$$\large {\sf P}(A \mid B)= {\sf P}(A), P(B) \ne 0$$

Грубо говоря, знание о том, что произошло непустое событие $B$ не даёт никакой новой информации о том, произошло ли событие $A$.

**Замечание:** Зачем упомянать условие ${\sf P}(A)\neq 0$?

Это условие понадобится, если поменять события $A$ и $B$ местами: ${\sf P}(B \mid A)={\sf P}(B)$

Вспомним наше соглашение: Если множество $B$ пустое, то по соглашению условную вероятность ${\sf P}(A \mid B)$ полагают равной нулю. Это работает аналогично, если $A$ и $B$ поменять местами ${\sf P}(B \mid A)$, тогда необходимо, чтобы ${\sf P}(A)\neq 0$, вот почему:

Пусть ${\sf P}(A)=0$. Разумеется, если $P(B)\neq 0$, то по нашему соглашению ${\sf P}(B \mid A) = 0$, следовательно ${\sf P}(B \mid A)\neq{\sf P}(B)$. Но если событие $A$ пустое, то вполне естественно считать, что ничего от него не зависит и условие независимости должно соблюдаться, хотя это не так.

Для этого же мы упоминаем то, что ${\sf P}(B)\neq 0$

### Вывод симметричной формулы независимости двух событий:

Более удобную (которая не зависит от того, является ли $B$ пустым событием) формулировку условия независимости событий $A$ и $B$ можно получить с помощью теоремы умножения:

$$\large {\sf P}(A \mid B)= {\sf P}(A)$$

Домнажаем на ${\sf P}(B)$ и получаем:

$$\large {\sf P}(A \mid B){\sf P}(B)={\sf P}(A){\sf P}(B)$$

Используя теорему умножения и получаем формулу вида:

$$\large {\sf P}(A \cap B)= {\sf P}(A){\sf P}(B)$$

То есть события $A$ и $B$ независимы тогда и только тогда, когда ${\sf P}(A\cap B) = {\sf P}(A){\sf P}(B)$.

**Дополнительное пояснение к формуле:**

И вот эта формула, вытекающая из нашего интуитивно понятного определения независимости, она уже, во-первых, совершенно симметрична относительно событий $A$ и $B$, их не имеет никакого смысла переставлять местами. Какая разница в каком порядке пересекать или в каком порядке переумножать? Ясно, что от перестановки мест сомножителей произведение не меняется. И более того, она устойчива к тому, что вероятность $B$ может равняться нулю. Если подставить событие, имеющее вероятность ноль, то вы справа получите ноль. Если слева  подставить пустое множество, то и пересечение будет пустым множеством, значит вероятность будет ноль. То есть у нас опять никаких проблем. Поэтому говорят, что $A$ и $B$ независимы, если вероятность $A$ пересеченного с $B$ – это вероятность $A$ помножить на вероятность $B$.

### `Независимость в совокупности:`

Однако, существует более сильное свойство независимости событий — независимость в совокупности. События $A_1,\ldots, A_k$ независимы в совокупности (взаимно независимы), если:

$$\large \forall{l} \leqslant k ~~~~ \forall{i_1,\ldots,i_l} ~~~~ {\sf P}(A_{i_1} \cap \ldots \cap A_{i_l}) = {\sf P}(A_{i_1}) \cdot \ldots \cdot {\sf P}(A_{i_l})$$

При этом, из совокупной взаимной независимости моментально вытекает попарная. Мы просто берем в качестве $l$ величину $2$, если $l$ равняется $2$, то мы берем произвольные два различных индекса и проверяем, что вероятность пересечения соответствующих двух событий – это произведение их вероятностей. То есть при $l$ равном $2$ мы как следствие из взаимной независимости получаем и попарную независимость тоже. 

<h1 style="color:#008B8B">Формула полной вероятности</h1>
 
Представьте себе, что у нас есть какое-то сложноустроенное пространство $\Omega$ (Рис ниже), состоящее из конечного числа  элементарных исходов, равновероятных. И у нас есть какое-то хитрое событие $A$ подмножество этого множества $\Omega$, состоящее из тех элементарных исходов, которые ему благоприятствуют и вероятность которого мы хотим вычислить. Вот зачастую, удобно вычислять эту вероятность, разбив все наше множество $\Omega$ на так сказать, случаи, на какие-то непересекающиеся подмножества — $B_1$, скажем, $B_2$ и так далее $B$ с индексом $k$. Непересекающиеся подмножества считаются, конечно, не пустыми.

<img src="img/1.jpg" width="300" height="260">

Исходя из свойств вероятности, мы можем найти вероятность события $A$:

$$\large {\sf P}(A) = {\sf P}(A \cap B_1) + {\sf P}(A \cap B_2) + \ldots + {\sf P}(A \cap B_k)$$

А дальше мы можем воспользоваться теоремой умножения ${\sf P}(A∣B){\sf P}(B)={\sf P}(A \cap B)$. Мы можем сказать, что вероятность пересечения, которая здесь присутствует, это есть условная вероятность, помноженная на вероятность условия.

$$\large {\sf P}(A) = {\sf P}(A \cap B_1) + {\sf P}(A \cap B_2) + \ldots + {\sf P}(A \cap B_k) = {\sf P}(A \mid B_1){\sf P}(B_1) + {\sf P}(A \mid B_2){\sf P}(B_2) + \ldots + {\sf P}(A \mid B_k){\sf P}(B_k)$$

### `Формула полной вероятности:`

Иначе, данную формулу можно записать короче, и эта запись называется формулой полной вероятности:

$$\large {\sf P}(A) = \sum^{k}_{i=1} {{\sf P}(A \mid B_i){\sf P}(B_i)}$$

<h3 style="color:#2E8B57" align="right">Задача с урнами на применение формулы полной вероятности:</h3>

**Отступление.** Раньше при голосовании использовались шары: если в определенную урну положили черный шар, то это значит голос "ПРОТИВ", а белый — "ЗА" данного кандидата.

**Пример.** Пусть даны две урны: в первой из них находятся $3$ черных и $2$ белых шара, а во второй — $1$ черный и $3$ белых. Из первой урны наугад достают два шара и кладут их внутрь второй урны. После этого шары во второй урне тщательно перемешиваются и из нее извлекается $1$ шар. Найти вероятность того, что этот шар — белый.

<img src="img/2.png" width="800" height="660">

**Формулируем события:**

В результате первого вытягивания реализуется одно из трех событий $B_1, B_2, B_3$:
* Событие $B_1$ состоит в том, что выпали два белых шара;
* $B_2$ — состоит в том, что выпали два черных;
* $B_3$ — состоит в том, что выпал один белый и один черный шар.

Надо вспомнить классическое определение вероятности и попробовать понять, какие элементарные исходы у нас сейчас имеются.

**Расчёт вероятности событий:**

Сколько всего возможностей из $5$ шаров выбрать $2$? Ну, ясное дело $C_{2}^5$, это можно сделать $10$ способами. А в числитель мы должны поставить те события, которые благоприятствуют нашему $B_1$. Оба шара белые. Но это всего одно событие. Мы могли только одним способом из этой урны достать $2$ белых шара.

$\large {\sf P}(B_1) = \frac{1}{C_{5}^2} = \frac{1}{10}$

$2$ черных шара мы можем вытащить тремя способами. $C_{2}^5$ это будет $\frac{3}{10}$.  

$\large {\sf P}(B_2) = \frac{C_{3}^2}{C_{5}^2} = \frac{3}{10}$

$1$ черный и $1$ белый шар мы можем вытащить шестью способами. $2 \cdot 3$ это будет $6$.

$\large {\sf P}(B_3) = \frac{2 \cdot 3}{C_{5}^2} = \frac{6}{10}$

**Расчёт условной вероятности событий:**

Вот, и чтобы применить формулу полной вероятности, нам еще нужно знать условные вероятности $A$ при условии $B_1, B_2, B_3$.

Если из первой урны были выбраны 2 белых шара, то во второй урне после перемешивания окажутся 5 белых и 1 черный шар.В этом случае вероятность вытянуть белый шар при условии, что произошло событие $B_1$ равно:

$\large {\sf P}(A \mid B_1) = \frac{5}{6}$

Аналогично с другими условными вероятностями:

$\large {\sf P}(A \mid B_2) = \frac{1}{2}$

$\large {\sf P}(A \mid B_2) = \frac{4}{6}$

**Согласно формуле полной вероятности**, искомая вероятность равна:

$$\large {\sf P}(A) = \frac{1}{10} \cdot \frac{5}{6} + \frac{3}{10} \cdot \frac{3}{6} + \frac{6}{10} \cdot \frac{4}{6} = 0.63$$

<h1 style="color:#008B8B">Формула Байеса</h1>

Другая важная с практической точки зрения формула, связанная с формулой полной вероятности, — это формула Байеса. Пусть события $B_1, \ldots , B_k$ попарно не пересекаются и в объединении дают все множество элементарных исходов. Тогда вероятность события $A$ по формуле полной вероятности равна:

$$\large {\sf P}(A) = \sum^{k}_{i=1} {{\sf P}(A \mid B_i){\sf P}(B_i)}$$

**Для чего нужна формула Байеса?**

Часто бывает необходимо найти вероятность ${\sf P}(B_i \mid A)$ события $B_i$ при условии, что событие $A$ произошло.

**Как посчитать A пересечённую с B?**

Как посчитать такую условную вероятность? Она-то вроде с ходу очевидным образом не считается. А давайте сделаем так: напишем вероятность $A$, пересеченного с $B$ i-тым: 

$$\large {\sf P}(A \cap B_i)$$

Конечно, эту вероятность можно посчитать с помощью теоремы умножения, причем двумя различными способами. С одной стороны, теорема умножения говорит конечно, что это $A$ при условии $B_i$ на вероятность $B_i$: 

$$\large {\sf P}(A \cap B_i) = {\sf P}(A \mid B_i){\sf P}(B_i)$$

Вот этим мы пользовались, когда создавали формулу полной вероятности. 

С другой стороны, эта вероятность $B_i$, при условии $A$ умножить на вероятность $A$, так как нет разницы каким образом производить пересечение событий. Абсолютно по той же самой причине, теореме умножения можно вывести следующую формулу:

$$\large {\sf P}(A \cap B_i) = {\sf P}(B_i \mid A){\sf P}(A)$$


**Формула условной вероятности:**

$$\large {\sf P}(A \mid B) = \frac{{\sf P}(A \cap B)}{{\sf P}(B)},  {\sf P}(B) \ne 0 $$

**Вывод формулы:**

$$\large {\sf P}(B_i \mid A) = \frac{{\sf P}(B_i \cap A)}{{\sf P}(A)}$$

Подставля полученные значения выше, можем получить формулу Байеса. Вспомним формулу выше, а именно ${\sf P}(A \cap B_i) = {\sf P}(B_i \mid A){\sf P}(A)$ и подставим её в ${\sf P}(B_i \cap A)$. Если что, разницы как перескать нет, поэтому можно приравнять ${\sf P}(A \cap B_i) = {\sf P}(B_i \cap A):$

$$\large {\sf P}(B_i \mid A) = \frac{{\sf P}(B_i \cap A)}{{\sf P}(A)} = \frac{{\sf P}(A \mid B_i){\sf P}(B_i)}{{\sf P}(A)}$$

Но вероятность события $A$ в свою очередь легко считается с помощью формулы полной вероятности, подставим её в формулу:

### `Формула Байеса:`

$$\large {\sf P}(B_i \mid A) = \frac{{\sf P}(A \mid B_i){\sf P}(B_i)}{\sum^{k}_{i=1} {{\sf P}(A \mid B_i){\sf P}(B_i)}}$$



<h3 style="color:#2E8B57" align="right">Задача на применение формулы Байеса:</h3>

Студент решает задачу с $k$ вариантами ответа. Если студент знает, как решать задачу, он дает правильный ответ. Если нет — может угадать с вероятностью $\frac{1}{k}$.

Какая вероятность того, что студент знает решение задачи, при условии, что был дан правильный ответ.

**P(Студент знает решение $B_1$ | $A$ решение правильное)**

Нам нужно еще знать, с какой вероятностью студент, вообще, в принципе, умел решать задачу. Пусть $p\in [0,1]$ — вероятность, что студент знает решение задачи. 
Событие $B_1$ соответствует ситуации, в которой студент знает решение задачи, а $B_2$ — не знает. Тогда:

$\large {\sf P}(B_1)=p$, $\large \quad {\sf P}(B_2)=1-p$, $\large \qquad {\sf P}(A \mid B_1)=1$, $\large \quad \quad {\sf P}(A \mid B_2)=\frac{1}{k}$

Теперь можно применить формулу Байеса:

$$\large {\sf P}(B_1 \mid A)= \frac{p}{p+\frac{1}{k}(1-p)}$$
