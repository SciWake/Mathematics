<h2 style="color:black" align="center">1.1 Событие и вероятность</h2>

<h3 style="color:#008B8B">Классическое определение вероятности</h3>

**Пример.** Игральная кость представляет собой кубик, грани которого пронумерованы от одного до шести, обычно точками. Кость бросают на стол и фиксируют число на той грани, которая оказывается сверху в результате броска. Можно предположить, что форма кубика идеальная и на ребро он упасть не может. В таком случае всего имеют место $6$ различных исходов такого броска, причем для них выполняются следующие свойства:

* Исходы образуют полную группу событий, то есть хотя бы один из них имеет место в результате броска (Никаких вариантов быть не может, кроме перечисленных);
* Исходы попарно несовместны, то есть никакие два одновременно произойти не могут;
* Все исходы равновероятны.

В результате рассмотрения игральной кости появилось классическое определение вероятности.

Исходы, обладающие перечисленными свойствами "образовывать полную группу событий", "быть попарно несовместными и равновероятными", называются элементарными исходами.

**Классическое определение вероятности.** Пусть различные исходы $\omega_1, \ldots, \omega_n$ 
некоторого эксперимента являются элементарными. Тогда по определению вероятностью каждого такого исхода называют величину:

<h3> $$P(\omega_i) = \frac{1}{n}$$ </h3>

**Пример.** Игральные карты. Пусть дана колода из $36$ тщательно перетасованных карт. В этом случае элементарными исходами являются различные последовательности карт внутри колоды, то есть различные варианты их взаимного расположения. Эти исходы, которых всего $36!$, образуют полную группу событий и попарно несовместны. Вероятность каждого исхода:

<h3> $$P(\omega_i) = \frac{1}{36!}$$ </h3>

Часто бывает необходимым рассматривать события несколько более сложные, чем элементарные исходы. Например, событию "в результате броска игральной кости выпало четное число очков" соответствуют три благоприятных элементарных исхода $\omega_2, \omega_4, \omega_6$. Итак, событием является любое множество элементарных исходов.

**Вероятность события.** Вероятность некоторого события $A$ полагают равной отношению количества элементарных исходов $k$, которые благоприятствуют событию, к количеству всех возможных элементарных исходов $n$:

<h3> $$P(\omega_i) = \frac{k}{n}$$ </h3>

<h3 style="color:#008B8B">Основные свойства вероятности</h3>

Если принять классическое определение как определение вероятности, то несложно получить следующие свойства.

Под событием удобно понимать множество таких элементарных исходов, которые ему благоприятствуют. Множество же всех элементарных исходов $\Omega=\{\omega_1,\ldots,\omega_n\}$ называется пространством элементарных исходов, где $n$ — количество всевозможных элементарных событий.

**Свойство 1.** Вероятность ${\sf P}(\Omega)$ события $\Omega$, которое состоит в том, что реализуется любой элементарный исход, равняется $1$. Такое событие называется достоверным.

**Свойство 2.** В классическом определении вероятностей нулевую вероятность имеет событие, которое соответствует пустому множеству элементарных исходов, и только оно:  ${\sf P}(A)=0 \Leftrightarrow A=\varnothing$.

**Свойство 3.** Вероятность дизъюнктного объединения двух событий $A$ и $B$, то есть таких событий, что $A$ и $B$ имеют пустое пересечение, равна сумме их вероятностей ${\sf P}(A\sqcup B) = {\sf P}(A) + {\sf P}(B)$

**Свойство 4.** Вероятность объединения двух произвольных событий $A$ и $B$ равна ${\sf P}(A\cup B) = {\sf P}(A) + {\sf P}(B) -{\sf P}(A\cap B)$

**Свойство 5.** Более того, верна формула включений-исключений, которая была доказана в курсе комбинаторики:

$P(A_1 \cup ... \cup A_k) = P(A_1) + ... + P(A_k) - P(A_1 \cap A_2) - P(A_1 \cap A_3) - ... - P(A_{k-1} \cap A_k) + ... + (-1)^{k-1} P(A_1 \cap ... \cap A_k) $

Естественным следствием является выражение для элементарной оценки сверху вероятности объединения нескольких событий.

**Свойство 6.** ${\sf P}(A_1 \cup \ldots \cup A_k) \leq {\sf P}(A_1)+ \ldots+ {\sf P}(A_k)$.

Вероятность $\cup$ не превосходит суммы вероятностей события $A_1, A_2, ... , A_k$


**Свойство 7.** Отрицанием события $A$ называется событие $\bar{A}$, которому благоприятствуют только такие исходы, которые не благоприятствуют событию $A$. Тогда верно следующее:

Вероятность события $\bar{A} := \Omega \backslash A$ равна ${\sf P}(\bar{A})=1-{\sf P}(A)$

___

**Дополнительный вопрос по свойствам:**

В классическом определении вероятности все элементарные исходы равновероятны. Из какого (ровно одного) свойства следует определение вероятности произвольного события $A$, состоящего из $k$ элементов ${\sf P}(A)=\frac{k}{n}$

__Ответ:__ ${\sf P}(A\sqcup B) = {\sf P}(A) + {\sf P}(B)$

Заметим, что из данного свойства, очевидно, следует аналогичное свойство для k непересекающихся событий $A_1,\ldots,A_k$: ${\sf P}(A_1\sqcup\ldots\sqcup A_k)={\sf P}(A_1)+\ldots+{\sf P}(A_k)$

Пусть $A=\{\omega_{i_1},\ldots,\omega_{i_k}\}$ Тогда 

${\sf P}(A)={\sf P}(\omega_{i_1})+\ldots+{\sf P}(\omega_{i_k})=\frac{1}{n}+\ldots+\frac{1}{n}=\frac{k}{n}$

Легко заметить, что любого другого одного свойства недостаточно, чтобы доказать это равенство.

<h2 style="color:black" align="center">1.2 Условная вероятность</h2>

<h2 style="color:#008B8B">Условная вероятность</h2>

**Понятие условной вероятности**

**Пример.** Пусть игральную кость бросают один раз. Событие $A$ состоит в том, что выпало $1$ очко, а $B$ — в том, что выпало четное число очков. Требуется найти вероятность ${\sf P}(A|B)$ события $A$, если уже известно, что событие $B$ произошло. Как изменится ответ, если в качестве события $A$ взять событие, состоящее в том, что выпало $2$ очка?

Ответ на первый вопрос кажется естественным: ${\sf P}(A|B) = 0$, поскольку $1$ является нечетным числом.

Второй вопрос задачи уже не такой тривиальный. Если выпало четное число очков, то пространство элементарных исходов сузилось до множества $B$. Множество элементарных исходов состоит из 3-х элементов: выпало $2$ очка, выпало $4$ очка и выпало $6$ очков, причем только один из них благоприятный и состоит в том, что кость выпала числом $2$ кверху. Таким образом, искомая вероятность равняется ${\sf P}(A|B)=\frac{1}{3}$.

___

**Получаем формулу условной вероятности:**

В общем случае условная вероятность определяется следующим образом. Пусть в пространстве элементарных исходов $\Omega = \{\omega_1,\ldots,\omega_n\}$ задано некоторое непустое событие $B=\{\omega_{i_1} , \ldots, \omega_{i_k}\}$ $k \geqslant 1$. Пусть также задано другое событие $A$. Тогда вероятность ${\sf P}(A|B)$ события $A$ при условии, что событие $B$ уже реализовалось, можно определить следующим образом:

Действительно, если событие $B$ уже реализовалось, то $B$ становится новым множеством элементарных исходов, а множеством благоприятных исходов станет множество $A\cap B$. 

То есть, в знаменатель поставим количество элементарных исходов, которые благоприятсвуют событию $B$, а это и есть число $k$ или же мощность $|B|$, а в числитель надо поставить те из вот этих благоприятствующих событию B исходов, которые, в свою очередь, благоприятствуют событию $A$. Этот момент очень естественно обозначить как мощность пересечения $A$ и $B$, то есть нам нужно посмотреть, какие элементарные исходы, благоприятствующие событию $A$, есть среди элементарных исходов, успевших реализоваться, коль скоро мы знаем, что $B$ произошло:

<h3>$${\sf P}(A) = \frac{|A \cap B|}{|B|}$$</h3>

Домножаем на мощность $|\Omega|$, чтобы получить вероятности:

<h3>$${\sf P}(A) = \frac{|A \cap B|}{|B|} = \frac{|A \cap B| \frac{1}{|\Omega|}}{|B| \frac{1}{|\Omega|}} = \frac{{\sf P}(A \cap B)}{{\sf P}(B)},  {\sf P}(B) \ne 0 $$</h3>

**Замечание:** Если множество $B$ пустое, то по соглашению условную вероятность ${\sf P}(A|B)$ полагают равной нулю.

___
**Теорема умножения:**

Если справа и слева формулу условной вероятности домножить на ${\sf P}(B)$, тогда мы получим:

<h3>$${\sf P}(A∣B){\sf P}(B)={\sf P}(A \cap B)$$</h3>

<h3 style="color:#008B8B">Независимость событий</h3>

**Понятие независимости событий**

Пусть в некотором пространстве элементарных исходов $\Omega$ заданы два события $A$ и $B$, причем $B$ не является пустым. Условие того, что событие $A$ не зависит от события $B$, имеет следующий вид:

<h3>$${\sf P}(A|B)= {\sf P}(A)$$</h3>

Грубо говоря, знание о том, что произошло непустое событие $B$ не даёт никакой новой информации о том, произошло ли событие $A$.

___

**Замечание:** Зачем упомянать условие ${\sf P}(A)\neq 0$?

Это условие понадобится, если поменять события $A$ и $B$ местами: ${\sf P}(B|A)={\sf P}(B)$

Вспомним наше соглашение: Если множество $B$ пустое, то по соглашению условную вероятность ${\sf P}(A|B)$ полагают равной нулю.

Пусть ${\sf P}(A)=0$. Разумеется, если $P(B)\neq 0$, то ${\sf P}(B|A)\neq{\sf P}(B)$. Но если событие $A$ пустое, то вполне естественно считать, что ничего от него не зависит.
___

**Выведение симметричной формулы независимости двух событий:**

Более удобную (которая не зависит от того, является ли $B$ пустым событием) формулировку условия независимости событий $A$ и $B$ можно получить с помощью теоремы умножения:

<h3>$${\sf P}(A|B)= {\sf P}(A)$$</h3>

Домнажаем на ${\sf P}(B)$ и получаем:

<h3>$${\sf P}(A∣B){\sf P}(B)={\sf P}(A){\sf P}(B)$$</h3>

Используя теорему умножения формулу вида:

<h3>$${\sf P}(A \cap B)= {\sf P}(A){\sf P}(B)$$</h3>

То есть события $A$ и $B$ независимы тогда и только тогда, когда ${\sf P}(A\cap B) = {\sf P}(A){\sf P}(B)$.

**Дополнительное пояснение к формуле:**

И вот эта формула, вытекающая из нашего интуитивно понятного определения независимости, она уже, во-первых, совершенно симметрична относительно событий $A$ и $B$, их не имеет никакого смысла переставлять местами. Какая разница в каком порядке пересекать или в каком порядке переумножать? Ясно, что от перестановки мест сомножителей произведение не меняется. И более того, она устойчива к тому, что вероятность $B$ может равняться нулю. Если вы сюда подставите событие, имеющее вероятность ноль, то вы справа получите ноль. Если сюда подставите пустое множество, то и пересечение будет пустым множеством, значит вероятность будет ноль. То есть у нас опять никаких проблем. Поэтому говорят, что $A$ и $B$ независимы, если вероятность $A$ пересеченного с $B$ – это вероятность $A$ помножить на вероятность $B$.

___

**Независимость в совокупности:**

Однако, существует более сильное свойство независимости событий — независимость в совокупности. События $A_1,\ldots, A_k$ независимы в совокупности (взаимно независимы), если:

<h3>$$\forall{l} \leqslant k ~~~~ \forall{i_1,\ldots,i_l} ~~~~ {\sf P}(A_{i_1} \cap \ldots \cap A_{i_l}) = {\sf P}(A_{i_1}) \cdot \ldots \cdot {\sf P}(A_{i_l})$$</h3>

При этом, из совокупной взаимной независимости моментально вытекает попарная. Мы просто берем в качестве $l$ величину $2$, если $l$ равняется $2$, то мы берем произвольные два различных индекса и проверяем, что вероятность пересечения соответствующих двух событий – это произведение их вероятностей. То есть при $l$ равном $2$ мы как следствие из взаимной независимости получаем и попарную независимость тоже. 

<h3 style="color:#008B8B">Формула полной вероятности</h3>
 
Представьте себе, что у нас есть какое-то сложноустроенное пространство «Омега» (Рис ниже), состоящее из конечного числа  элементарных исходов, равновероятных. И у нас есть какое-то хитрое событие «А» подмножество этого множества «Омега», состоящее из тех элементарных исходов, которые ему благоприятствуют и вероятность которого мы хотим вычислить. Вот зачастую, удобно вычислять эту вероятность, разбив все наше множество «Омега» на так сказать, случаи, на какие-то непересекающиеся подмножества — $B_1$, скажем, $B_2$ и так далее $B$ с индексом $k$. Непересекающиеся подмножества считаются, конечно, не пустыми, но это очень не естественно. 

<img src="img/1.jpg"  width="300" height="260">

Исходя из свойств вероятности:

<h3> $${\sf P}(A) = {\sf P}(A \cap B_1) + {\sf P}(A \cap B_2) + \ldots + {\sf P}(A \cap B_k)$$ </h3>

А дальше мы можем воспользоваться теоремой умножения ${\sf P}(A∣B){\sf P}(B)={\sf P}(A \cap B)$. Мы можем сказать, что вероятность пересечения, которая здесь присутствует, это есть условная вероятность, помноженная на вероятность условия.

<h3> $${\sf P}(A) = {\sf P}(A \cap B_1) + {\sf P}(A \cap B_2) + \ldots + {\sf P}(A \cap B_k) = {\sf P}(A∣B_1){\sf P}(B_1) + {\sf P}(A∣B_2){\sf P}(B_2) + \ldots + {\sf P}(A∣B_k){\sf P}(B_k)$$ </h3>

Иначе, данную формулу можно записать короче, и эта запись называется формулой полной вероятности:

<h3> $${\sf P}(A) = \sum^{k}_{i=1} {{\sf P}(A∣B_i){\sf P}(B_i)}$$ </h3> 

<h3 align="right">Задача на формулу полной вероятности:</h3>
