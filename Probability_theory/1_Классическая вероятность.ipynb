<h1 style="color:black" align="center">1. Классическая вероятность</h1>
<h1 style="color:black" align="center">1.1 Событие и вероятность</h1>

<h1 style="color:#008B8B">Классическое определение вероятности</h1>

**Пример.** Игральная кость представляет собой кубик, грани которого пронумерованы от одного до шести, обычно точками. Кость бросают на стол и фиксируют число на той грани, которая оказывается сверху в результате броска. Можно предположить, что форма кубика идеальная и на ребро он упасть не может. В таком случае всего имеют место $6$ различных исходов такого броска, причем для них выполняются следующие свойства:

* Исходы образуют полную группу событий, то есть хотя бы один из исходов обязательно произойдёт (Никаких вариантов быть не может, кроме перечисленных);
* Исходы попарно несовместны, то есть никакие два исхода одновременно произойти не могут;
* Все исходы равновероятны.

В результате рассмотрения игральной кости появилось классическое определение вероятности.

Исходы, обладающие перечисленными свойствами "образуют полную группу событий", "попарно несовместны" и "равновероятны" называются элементарными исходами.

**Классическое определение вероятности.** Пусть различные исходы $\omega_1, \ldots, \omega_n$ 
некоторого эксперимента являются элементарными. Тогда по определению вероятностью каждого такого исхода называют величину:

$$\large P(\omega_i) = \frac{1}{n}$$

**Пример.** Игральные карты. Пусть дана колода из $36$ тщательно перетасованных карт. В этом случае элементарными исходами являются различные последовательности карт внутри колоды, то есть различные варианты их взаимного расположения. Эти исходы, которых всего $36!$, образуют полную группу событий и попарно несовместны. Вероятность каждого исхода:

$$\large P(\omega_i) = \frac{1}{36!}$$

Часто бывает необходимым рассматривать события несколько более сложные, чем элементарные исходы. 

**Например**, событие $A$ состоит в том, что "в результате броска игральной кости выпало четное число очков". Такому событию соответствуют три благоприятных элементарных исхода $\omega_2, \omega_4, \omega_6$. 

**Вероятность события.** Тогда вероятность некоторого события $A$ полагают равной отношению количества элементарных исходов $k$, которые благоприятствуют событию, к количеству всех возможных элементарных исходов $n$:

$$\large P(A) = \frac{k}{n} = \frac{3}{6} = \frac{1}{2}$$

<h1 style="color:#008B8B">Свойства вероятности</h1>

Если принять классическое определение как определение вероятности, то несложно получить следующие свойства.

Под событием удобно понимать множество таких элементарных исходов, которые ему благоприятствуют. Множество же всех элементарных исходов $\Omega=\{\omega_1,\ldots,\omega_n\}$ называется пространством элементарных исходов, где $n$ — количество всевозможных элементарных событий.

**Свойство 1.** Вероятность ${\sf P}(\Omega)$ события $\Omega$, которое состоит в том, что реализуется любой элементарный исход, равняется $1$. Такое событие называется достоверным, так как происходит с веротяностью равной $1$.

**Свойство 2.** В классическом определении вероятностей, нулевую вероятность имеет событие, которое соответствует пустому множеству элементарных исходов:  ${\sf P}(A)=0 \Leftrightarrow A=\varnothing$.

**Свойство 3.** Вероятность дизъюнктного объединения двух событий $A$ и $B$, то есть таких событий, что $A$ и $B$ имеют пустое пересечение, равна сумме их вероятностей ${\sf P}(A\sqcup B) = {\sf P}(A) + {\sf P}(B)$

**Свойство 4.** Вероятность объединения двух произвольных событий $A$ и $B$ равна ${\sf P}(A\cup B) = {\sf P}(A) + {\sf P}(B) -{\sf P}(A\cap B)$

**Свойство 5.** Более того, верна формула включений-исключений, которая была доказана в курсе комбинаторики:

$P(A_1 \cup \ldots \cup A_k) = P(A_1) + ... + P(A_k) - P(A_1 \cap A_2) - P(A_1 \cap A_3) - \ldots - P(A_{k-1} \cap A_k) + \ldots + (-1)^{k-1} P(A_1 \cap \ldots \cap A_k) $

**Свойство 6.** ${\sf P}(A_1 \cup \ldots \cup A_k) \leq {\sf P}(A_1)+ \ldots+ {\sf P}(A_k)$.

Вероятность $\cup$ не превосходит суммы вероятностей события $A_1, A_2, \ldots , A_k$

**Свойство 7.** Отрицанием события $A$ называется событие $\bar{A}$, которому благоприятствуют только такие исходы, которые не благоприятствуют событию $A$. Тогда верно следующее:

Вероятность события $\bar{A} = \Omega \backslash A$ равна ${\sf P}(\bar{A})=1-{\sf P}(A)$

### Дополнительный вопрос по свойствам:

В классическом определении вероятности все элементарные исходы равновероятны. Из какого (ровно одного) свойства следует определение вероятности произвольного события $A$, состоящего из $k$ элементов ${\sf P}(A)=\frac{k}{n}$

__Ответ:__ ${\sf P}(A\sqcup B) = {\sf P}(A) + {\sf P}(B)$

Заметим, что из данного свойства, очевидно, следует аналогичное свойство для k непересекающихся событий $A_1,\ldots,A_k$: ${\sf P}(A_1\sqcup\ldots\sqcup A_k)={\sf P}(A_1)+\ldots+{\sf P}(A_k)$

Пусть $A=\{\omega_{i_1},\ldots,\omega_{i_k}\}$ Тогда 

${\sf P}(A)={\sf P}(\omega_{i_1})+\ldots+{\sf P}(\omega_{i_k})=\frac{1}{n}+\ldots+\frac{1}{n}=\frac{k}{n}$

Легко заметить, что любого другого одного свойства недостаточно, чтобы доказать это равенство.

<h1 style="color:black" align="center">1.2 Задача о раскраске пятнадцати 5-элементных множеств</h1>

<h1 style="color:#008B8B">Формулировка задачи</h1>

Рассмотрим $15$ совершенно произвольных подмножеств некоторого тридцати элементного множества $M_1, \ldots, M_{15} \subset \{1, 2, \ldots, 30\}$ (У нас есть множество состоящее из $30$ натуральных чисел и в этом множестве, выбраны и зафиксированы какие-то $15$ подмножеств). Мощность каждого подмножества, количество элементов в нём $|M_i| = 5$. В зависимости от того, как мы зафиксируем пятиэлементые подмножества, они могут иметь пересечения между собой, так как сумарное количество элементов $75$, что больше чем $30$ исходного множества.

Верно ли, что какова бы ни была зафиксированная нами совокупность пятерок $M_1, \ldots, M_{15}$, всегда можно найти расскраску элементов множества $\{1, 2, \ldots, 30\}$ в два цвета (Каждому элементу присвоить либо красный, либо синий цвет), при которой все множества $M_i$ (каждая пятёрка) не одноцветные, то есть каждое множество содержит как элемент красного, так и элементы синего цвета.

Зададим вопрос более простыми словами. Можно ли аудиторию из $30$ человек разделить на две непересекающиеся части, рассадить по двум аудиториям, так чтобы в первой аудитории присутствовало по представителю из каждого множества лучших $M_1, \ldots, M_{15}$ ($M_1$ - пятёрка лучших учеников, которые лучше всех знают комбинаторику и так далее), так и во второй аудитории присутствовало хотябы по одному представителю из каждого множества лучших. Существует ли всегда такое разбиение аудитории на две непересекающиеся части.

Как в первой аудитории присутсвуют $15$ лучших математиков, физиков..., так и во второй аудитории найдутся $15$ других студентов, которые лучше всех знаю математику, физику. 

<h1 style="color:#008B8B">Решение задачи</h1>

Рассмотрим случайную расскраску $X$ множества $\{1, 2, \ldots, 30\}$. Пространство $\Omega$ - это и есть множество всех расскарсок. Расскрасок $2^{30}$, мы каждому элементу независимо от остальных можем присвоить либо красный, либо синий цвет. $\Omega = \{X_1, \ldots, X_{2^{30}}\}$ и вероятность возникновения любой расскраски можно считать равной $P(X_i) = \frac{1}{2^{30}}$. 

У нас зафиксирована совокупнось пятёрок $M_1, \ldots, M_{15}$, возьмём любое множество $M_i$ из этой совокупности и сопоставим этому множеству событие $A_i$ состоящие в том, что в случайной расскраске множество $M_i$ одноцветное. 

Чему тогда равна вероятность $P(A_i)$? Надо найти количество всех элементарных исходов, которые благоприятствуют нашему событию, то есть количество всех раскрасок, в которых данное множество $M_i$-тое одноцветно, и разделить на общее количество раскрасок, то на $2$ в $30$-той степени. В знаменателе стоит $2^{30}$. В числителе: есть всего два варианта, как сделать пятиэлементное множество одноцветным, покрасить $5$ элементов в карсный цвет или в синий и есть ещё определённое количество способов покрасить все остальные элементы, которых $25$. И количество способов расскрасить остальные элемент эт $2^{25}$.

$\large P(A_i) = \frac{2\cdot 2^{25}}{2^{30}} = \frac{1}{16}$

Вероятность объединения $A_i$ от $1$ до $15$ $P(\bigcup\limits_{i=1}^{15} {A_i})$ не превосходит суммы от $1$ до $15$ вероятностей $A_i$ $P(\sum\limits_{i=1}^{15} {A_i})$, где $P(A_i) = \frac{1}{16}$

$\large P(\bigcup\limits_{i=1}^{15} {A_i}) \le P(\sum\limits_{i=1}^{15} {A_i}) < 1$

Что обозначет событие объединения $A_i$? $A_i$ состоит из расскрасок, где $M_i$ одного цвета (например, $A_1$ - это расскраска в которых $M_1$ одного цвета, $A_2$ - это расскраска в которых $M_2$ целиком одного цвета). Следовательно это объедение состоит в том, что $M_1$ одноцветно, либо $M_2$ одного цвет и так далее до $M_15$,  значит, это событие состоящее в том, что в случайно расскраске найдётся одноцветное множество среди $15$ зафиксированных. 

Воспользуемся свойством $7$, которое мы определили:

Отрицание события $P(\bigcup\limits_{i=1}^{15} {A_i}) > 0$, так как само событи строго меньше $1$, а вероятность его отрицания строго больше 0 по $7$ свойствую. В чём состоит отрицание, само событие состоит в том, что существует одноцветное множество, а отрицание этого события состоит в том, что нет ни одного одноцветного множества.

Вспоминаем наше условие задачи, "всегда можно найти расскраску элементов множества $\{1, 2, \ldots, 30\}$ в два цвета, при которой все множества $M_i$ не одноцветные". Наше отрицательное событие говорит о том, что все множества не одноцветные и это происходит с положительной вероятностью. Следовательно, по свойству $2$, вероятность равняетя нулю, тогда и только тогда, когда $A$ это пустое множество. Значит, отрицание нашего события существует, так как вероятность его больше нуля и ему сопоставленно определённое множество где все множества $M_i$ не одноцветные.

<h1 style="color:black" align="center">Семинар 1</h1>

<h3 style="color:#2E8B57">Задача № 1</h3>

Книжная полка вмещает сорок книг. Среди имеющихся у нас сорока книг (все книги разные) есть три тома Пушкина. Ставим наши книги в случайном порядке на полку (все расстановки являются равновероятными). С какой вероятностью тома Пушкина окажутся стоящими в правильной последовательности (от меньшего к большему номеру тома, если считать книги, стоящие на полке, слева направо), но не обязательно рядом?

<h4 align='center'>Решение:</h4>

Определим пространство элементарных исходов $\Omega$ для нашей задачи. Так как случайным является порядок книг на полке, то элементарными исходами являются расстановки книг на полке (всего $40!$ элементарных исходов): $\Omega = \{\omega_1, \ldots, \omega_{40!}\}$. Таким образом, вероятность одного элементарного исхода равна $\frac{1}{40!}$. Посчитаем число расстановок, в которых тома Пушкина стоят в правильным порядке (соответствующее событие мы обозначим $A$). Выберем сначала три позиции, на которые можно поставить тома Пушкина. Таких способов $C_{40}^3$. Коль скоро мы зафиксировали эти позиции, мы уже обязаны поставить первый том Пушкина на первую позицию, второй - на среднюю и третий - на последнюю. Осталось поставить остальные $37$ книг. Это можно сделать $37!$ способами. Итак, $|A| = C_{40}^3 \cdot 37!$.

Тогда

$\large {\sf P}(A)=\frac{|A|}{|\Omega|}=\frac{C_{40}^3 \cdot 37!}{40!}=\frac{40! \cdot 37!}{3! \cdot 37! \cdot 40!}=\frac{1}{3!}=\frac{1}{6}$

__Ответ:__ $\frac{1}{6}$

<h3 style="color:#2E8B57">Задача № 2</h3>

Из множества $N = \{a_1, \ldots a_N\}$ объектов выбирается случайное подмножество (выборы всех подмножеств равновероятны) Вернее, говорится так: давайте рассмотрим множество всех подмножеств этого множества, и вот из этого множества, множества всех подмножеств, возьмем один случайный элемент. Вот это будет случайное подмножество нашего множества из $N$ объектов. Найдите вероятность того, что это случайное подмножество имеет четную мощность (мощностью множества называется число элементов в нем).

<h4 align='center'>Решение:</h4>

Обозначим данное множество $\{a_1,\ldots,a_N\}$. Определим пространство элементарных исходов $\Omega$ для нашей задачи: $\Omega=\{\omega_1,\ldots,\omega_{2^N}\}$ (так как во множестве $\{a_1,\ldots,a_N\}$ всего $2^N$ подмножеств).Таким образом, вероятность одного элементарного исхода равна $\frac{1}{2^N}$. 


Обозначим $A$ событие, заключающееся в том, что случайное подмножество имеет четную мощность: $A=\{\omega:\,|\omega|\text{ - четное}\}$. Для нас благоприятствуют все возможные подмножества $N$ элементного множества, которые имеют четную мощность. И для подсчета вероятности, необходимо разделить количество элементарных исходов, которые входят в множество $A$ (благоприятсвуют событию $A$) на общее количество элементарных исходов:

$\large {\sf P}(A)=\frac{|A|}{|\Omega|}$

**Способ 1.** Значит, нам необходимо посчитать сколько есть подмножеств, которые имеют четную мощность в множестве состоящем из $N$ элементов. Будем считать, что $0$ это четное число (тогда пустое множество имеет четную мощность). Очевидно, $|A|=C_N^0+C_N^2+\ldots$, так как четных множеств $C_N^0$, дальше смотрим на все двухэлементые подмножества $C_N^2$. А последнее слогаемо зависит от четности числа $N$, если $N$ четное, тогда последнее слогаемое будет $C_N^N$, а если $N$ не четное, тогда последнее слогаемое будет $C_N^{N-1}$. И данный способ не очень нравится, так как необходимо смотреть на четность исходного множества.

**Способ 2.** Посчитаем это количество иным способом: каждый элемент (начиная с $a_1$) можно либо положить, либо не положить в строящееся случайное множество (две возможности для каждого элемента), дальше можем аналогично поступить с элементом $a_2$, положить, либо не положить. Так как мы хотим построить множество четной мощности, то, возможно, последний элемент мы в наше множество положить не можем. Действительно, после того, как мы приняли решение относительно элемента $a_{N-1}$, мы либо обязаны положить элемент $a_N$, либо обязаны его не класть, чтобы получилось четное множество. Значит, есть всего $2^{N-1}$ способов выбрать первые $N-1$ элементов и для каждого из этих способов имеется однозначный путь добавления элемента $a_N$, так чтобы в итоге получилось множество четной мощности.


Итак, по две возможности мы имеем ровно для $N-1$ элементов. Таким образом, $|A|=2^{N-1}$. (В частности, это означает, что $C_N^0+C_N^2+\ldots=2^{N-1}$).
 
__Ответ:__ $\large {\sf P}(A)=\frac{|A|}{|\Omega|}=\frac{2^{N-1}}{2^N}=\frac{1}{2}$


**Дополнительное пояснение:** Задача о случайном подмножестве $N-1$ - почему?

Это из курса комбинаторики. Можно разными способами объяснять это, например так. Всего подмножеств $N$-элементного множества $2^N$ (каждый элемент можем выбрать или не выбрать). А между "чётными" и "нечётными" подмножествами есть взаимно-однозначное соответствие, получаемое добавлением или выкидыванием 1-го элемента. То есть если в $A$ нет элемента $1$, то мы его добавляем. Если есть - выкидываем. Тогда $A$ из чётного становится нечётным и наоборот. Поэтому "чётных" множеств ровно половина, то есть $2^{N-1}$. (это всё при $N\geq 1$).

<h3 style="color:#2E8B57">Задача № 3</h3>

Графом на $n$ вершинах называется пара $G=(V,E)$, где $V$ - множество из $n$ элементов, называемое множеством вершин графа, $E$ - множество пар вершин из $V$, называемое множеством ребер. Если пары в $E$ упорядочены, то граф называется ориентированным. Иначе - неориентированным. Петлей называется ребро графа, соединяющее вершину саму с собой. Граф на $n$ вершинах называется простым циклом, если его множество вершин можно занумеровать таким образом числами от $1$ до $n$, что каждая вершина $i$ соединена ребром с вершиной $i+1$, а вершина $n$ - с вершиной $1$ (и в графе нет других ребер).

<h4 align='center'>Решение:</h4>

У нас есть множество, которое состоит из $n$ элементов $\{1, \ldots, n\}$ - это множество мы считаем множеством вершин графа. Мы можем любые две из этих точек между собою соединить отрезочком, который называется ребром. То есть вот каждая неупорядоченная пара элементов этого множества теоретически может быть ребром графа, а может и не быть таковым. Например, в качестве графа вполне можно рассмотреть вот такую вот конструкцию:

Пространство элементарных исходов $\Omega$ - это множество всех неориентированных графов без петель на множестве вершин $\{1,\ldots, n \}$. Так как каждая пара различных вершин либо образует, либо не образуют ребро (две возможности для $C_n^2$ пар вершин), то $|\Omega|=2^{C_n^2}$. Неупорядоченных пар вершин, которые можно взять из множества, имеющего мощность $n$, их, конечно, $C$ из n по $2$, и для любой такой пары либо она соединена ребром, либо не соединена, поэтому $2$ в степени $C$ из n по $2$ – это количество вариантов провести или не провести ребра. 

Вот такое вот пространство элементарных исходов, и каждый элементарный исход – то есть граф – имеет, согласно классическому определению, вероятность, равную $1$ поделить на $2$ в степени $C$ из $n$ по $2$. Вот, но нас в этой задаче интересует, с какой вероятностью такой вот случайный граф является циклом, простым циклом?

<img src="img/Seminar_3_2.svg" width="800" height="600">`

Простой цикл – это такая последовательность вершин и ребер, в которой они вот так чередуются, и ни ребра, ни вершины не повторяются, покуда мы не наткнемся на исходную.

С какой вероятностью случайный граф окажется простым циклом? $A$ – это вот то самое событие, которое нас интересует. И чтобы посчитать его вероятность, мы, как обычно, должны написать в знаменателе общее количество элементарных событий, то есть общее количество графов на $n$ вершинах, а в числитель мы должны записать столько графов, сколько благоприятствует этому событию.

Сколько вот есть циклов, простых циклов на $n$ вершинах? Ну вот, скажем, на трех вершинах $(1, 2, 3)$ сколько всего есть простых циклов? А очень просто – только один. Смотрите, как ни перетасовывай, как ни переставляй между собою единицу, двойку и тройку, у нас все тот же самый треугольник получается. 

<img src="img/Seminar_3_3.svg" width="1200" height="100">`

Как посчитать общее количество простых циклов, имеющих $n$ вершин?

Ну, конечно, если бы мы строили цепочку вот такую и как-то пытались расставить по этой цепочке числа, то таких цепочек ну столько, сколько есть всего перестановок на множестве из $n$ элементов, а перестановок их $n$ факториал. Вернуться к примеру с треугольником, то есть перестановка $(1, 2, 3)$, а есть перестановка $(1, 3, 2)$. Это разные перестановки. Но в тот момент, когда мы их зацикливаем получаем один цикл.

В общем случае мы посмотрим на какую-нибудь перестановку:

$$\large \begin{matrix}
a_1 & a_2 & \ldots & a_n\\
a_2 & a_3 & \ldots & a_n & a_1 \\
a_3 & a_4 & \ldots & a_n & a_1 & a_2 \\
& \ldots & \ldots  & \ldots \\
a_n & a_1 & \ldots & a_{n-1}
\end{matrix}$$

Все эти различные цепочки зацикливаются в один и тот же цикл. Естественно, получается, что уже, как минимум, в $n$ раз циклов меньше, чем существует всех возможных вот таких вот цепочек. На самом деле, ведь мы же можем еще эту последовательность прочитать в обратном порядке. И уже вот эту вот в обратном порядке прочитанную последовательность снова всеми возможными $n$ способами по циклу попереставлять:

$$\large \begin{matrix}
a_n & a_{n-1} & \ldots & a_1\\
a_{n-1} & \ldots & a_1 & a_n\\
\ldots & \ldots & \ldots \\
a_1 & a_n & \ldots & a_2
\end{matrix}$$

Есть $n!$ цепочек и для каждой цепочки есть $2n$ перестановок, которые дают один и тот же самый цикл, поэтому надо поделить на $2n$, коль скоро мы считаем количество всех циклов, и это получится: 

$\large \frac{n!}{2n} = \frac{(n – 1)!}{2}$

**Обобщим что мы нашли:**

Пусть $A$ (интересующее нас событие) - множество простых циклов, входящих в $\Omega$. Тогда $|A|=\frac{n!}{2n}$, так как способов сделать цепь (расположить все вершины подряд) $n!$, но когда мы эту цепь "зациклим", то мы один и тот же цикл учтем $2n$ раз ($n$ цепей, в которых элементы переставляются "по кругу" дают один и тот же цикл, а каждую такую цепь можно "прочитать" в обратном порядке, что снова даст тот же самый цикл).

Возвращаемся к подсчету нашей вероятности.  

$\large {\sf P}(A)=\frac{|A|}{|\Omega|}=\frac{(n-1)!/2}{2^{C_n^2}}$

Но в задаче еще предлагалось найти предел этой вероятности при $n$ маленькое, стремящимся к бесконечности.

Все-таки что вероятнее: вытащить из огромной кучи графов простой цикл или не простой цикл. Совершенно понятно, что вероятнее вытащить не цикл, потому что цикл – ну это что-то такое очень специфическое.

Заметим, наконец, что 

$\large \frac{(n-1)!}{2}<n!\leq n^n=2^{n\log_2 n}$ 

Тогда

$\large {\sf P}(A)<\frac{2^{n\log_2 n}}{2^{\frac{n^2}{2}-\frac{n}{2}}}\rightarrow 0,\quad n\rightarrow\infty$,

Так как порядок роста $\frac{n^2}{2}-\frac{n}{2}$ больше, чем порядок роста $n\log_2 n$.

<h3 style="color:#2E8B57">Задача № 4</h3>

Случайно бросаются два $M$-гранных кубика, на гранях которых написаны числа от $1$ до $M$. Найдите вероятность того, что сумма чисел, выпавших на кубиках, равна $i$, где $i\in\{2,3,\ldots,M+1\}$.

<h4 align='center'>Решение:</h4>

Пусть $\Omega=\{(j_1,j_2),\quad j_1,j_2\in\{1,2,\ldots,M\}\}$. Тогда $|\Omega|=M^2$. Обозначим искомое событие $A_i$. Если $i\in\{2,3,\ldots,M+1\}$, то сумма чисел равна $i$ тогда и только тогда, когда на первом кубике выпало произвольное число $j\in\{1,2,\ldots,i-1\}$, а на втором - в точности $i-j$. Поэтому $A_i=\{(j,i-j),\quad j\in\{1,2,\ldots,i-1\}\}$, т.е. $|A_i|=i-1$. Таким образом, ${\sf P}(A_i)=\frac{i-1}{M^2}$.

<h3 style="color:#2E8B57">Задача № 5</h3>

По схеме случайного выбора с возвращением из множества натуральных чисел $1, \ldots, 2016$ выбираются числа $X$ и $Y$. Верно ли, что вероятность того, что разность квадратов этих чисел является четным числом, больше, чем вероятность того, что эта разность делится на $3$?

<h4 align='center'>Решение:</h4>

Пусть $\Omega=\{(i,j):\quad i,j\in\{1,\ldots,2016\}\}$, $|\Omega|=2016^2$. Положим

$A_2=\{(i,j):\quad i,j\in\{1,\ldots,2016\},\,i^2-j^2\vdots 2\}$

$A_3=\{(i,j):\quad i,j\in\{1,\ldots,2016\},\,i^2-j^2\vdots 3\}$

Интуитивно может показаться, что $|A_2|>|A_3|$. На самом деле это не так. Действительно, разность двух чисел является четной тогда и только тогда, когда эти два числа имеют одинаковую четность. Следовательно,

$A_2=\{(2i,2j),(2i-1,2j-1):\quad i,j\in\{1,\ldots,1008\}\}$. 

Поэтому $|A_2|=2\cdot 1008^2$. Но у квадрата натурального числа может быть только два остатка при делении на  $3$ - $0$ и $1$. Поэтому разность квадратов двух чисел делится на  тогда и только тогда, когда либо оба эти два числа делятся на $3$, либо оба не делятся. Поэтому 

$A_3=\{(3i,3j),(3i-1,3j-1),(3i-1,3j-2),(3i-2,3j-1),(3i-2,3j-2):\quad i,j\in\{1,\ldots,672\}\}$.

Таким образом,

${\sf P}(A_2)=\frac{2\cdot 1008^2}{2016^2}=\frac{1}{2}\lt\frac{5}{9}=\frac{5\cdot 672^2}{2016^2}={\sf P}(A_3).$

Ответ: неверно.

<h3 style="color:#2E8B57">Задача № 6</h3>

В карточной игре покер игрок получает 5 карт из колоды в 52 карты. Найдите вероятности следующих комбинаций.

пара - две карты одного номинала (и три оставшиеся карты имеют другие попарно различные номиналы);

тройка - три карты одного номинала (и две оставшиеся карты имеют другие и различные номиналы);

стрит - пять последовательных по номиналу карт (предполагается, что за тузом по номиналу следует двойка), среди которых есть хотя бы две различные масти;

три+два - три карты одного номинала, две карты - другого;

ройал-флэш - туз, король, дама, валет и десятка одной и той же масти.

В ответ запишите сумму полученных вероятностей.

<h4 align='center'>Решение:</h4>


Количество всевозможных наборов из $5$ карт $|\Omega|=C_{52}^5$. Комбинация "пара" выпадает, если выбраны две карты одного номинала (один номинал выбирается из тринадцати, две масти выбираются из четырех), а остальные карты имеют три различных номинала из $12$ (каждая масть выбирается из четырех), т.е. встречается в $|A_1|=13C_4^2C_{12}^3 4^3=13\cdot 12\cdot 11\cdot 10\cdot 64$ наборах из $5$ карт. Таким образом, ${\sf P}(A_1)=\frac{13\cdot 12\cdot 11\cdot 10\cdot 64\cdot 5!}{52\cdot 51\cdot 50\cdot 49\cdot 48}=\frac{352}{833}$.

Комбинация "тройка" выпадает, если выбраны три карты одного номинала (один номинал выбирается из тринадцати, три масти выбираются из четырех), а остальные карты имеют два различных номинала из оставшихся  (каждая масть снова выбирается из четырех), т.е. встречается в $|A_2|=13C_4^3 C_{12}^2 4^2=13\cdot 4\cdot 12\cdot 11\cdot 8$ наборах из $5$ карт. Таким образом, ${\sf P}(A_2)=\frac{13\cdot 4\cdot 12\cdot 11\cdot 8\cdot 5!}{52\cdot 51\cdot 50\cdot 49\cdot 48}=\frac{264}{12495}$.

Комбинация "стрит" выпадает, если выбрана последовательность из $5$ подряд идующих номиналов ($13$ способов) и пять мастей ($4^5-4$ способов), т.е. встречается в $|A_3|=13(4^5-4)=13\cdot 1020$ наборах из $5$ карт. Таким образом, ${\sf P}(A_3)=\frac{13\cdot 1020\cdot 5!}{52\cdot 51\cdot 50\cdot 49\cdot 48}=\frac{1}{196}$.

Комбинация "три+два" выпадает, если выбраны три карты одного номинала (один номинал выбирается из тринадцати, три масти выбираются из четырех) и две карты одного номинала (один номинал выбирается из двенадцати, две масти выбираются из четырех), т.е. встречается в $|A_4|=13 С_4^3 12 C_4^2=13\cdot 4\cdot 12\cdot 6$ наборах из $5$ карт. Таким образом, ${\sf P}(A_4)=\frac{13\cdot 4\cdot 12\cdot 6\cdot 5!}{52\cdot 51\cdot 50\cdot 49\cdot 48}=\frac{6}{4165}$.

Комбинация "ройал-флэш" выпадает, если выбраны туз, король, дама, валет и десятка одной и той же масти, т.е. встречается в $4$ наборах из $5$ карт. Таким образом, ${\sf P}(A_5)=\frac{4\cdot 5!}{52\cdot 51\cdot 50\cdot 49\cdot 48}=\frac{1}{649740}$.

Окончательно получаем ответ

$\frac{352}{833}+\frac{264}{12495}+\frac{1}{196}+\frac{6}{4165}+\frac{1}{649740}=\frac{780\cdot 352+264\cdot 52+3315+6\cdot 156+1}{649740}=\frac{292540}{649740}=\frac{14627}{32487}.$
