<h1 style="color:#008B8B">Схема испытаний Бернулли</h1>

**Определение вероятности орла и решки:**

Изучив независимость событий мы можем приступить к схеме испытаний Бернулли. У нас есть монета, со смещенным центром тяжести. Значит, давайте считать, что при случайном бросании монетки на стол: 

* $p \in [0, 1]$ – это вероятность решки, соответственно, 

* $q = 1 - p$, которую еще принято в этой науке для краткости обозначать $q$ – это вероятность орла.

**Формирование последовательности из 1 и 0**

И вот теперь будем проделывать, собственно, схему испытаний Бернулли. А именно зафиксируем в свою очередь некоторое натуральное число $n \in N$ и будем бросать нашу монетку на стол ровно $n$ раз, каждый раз фиксируя какой стороной кверху она упала. Успех в данном случае соответствует выпадению решки при случайном бросании монеты на стол, и этот успех кодируется значением $1$. Соответственно, если монетка упала кверху орлом, то мы фиксируем $0$ и говорим о том, что произошла неудача.

В итоге у нас возникает после всех $n$ бросаний монетки некоторая последовательность чисел, каждое из которых – это либо $1$, либо $0$. Давайте эту последовательность обозначим, как водится в теории вероятности, буквой $\omega$, ну а сама последовательность будет состоять из символов, скажем $x_1, \ldots, x_n$, где, $x_i$-тое при любом $i$ от $1$ до $n$ – это либо нолик, либо единичка:

$\large \omega = \{x_1, \ldots, x_n\}$, где $\large x_i \in \{0, 1\}$

**Определение пространства элементарных исходво:**

То есть всего различных таких $\omega$ маленьких последовательностей, которые реализуются в результате схемы испытаний Бернулли их, конечно, $2$ в степени $n$ штук. Вот давайте введем обозначение $\Omega$ большое – это будет множество всех таких $\omega$ маленьких, которые состоят из ноликов и единичек:

$\large \Omega:= \{\omega = \{x_1, \ldots, x_n\}, \large x_i \in \{0, 1\}\}$

Значи мощность $\Omega$ – количество элементов в нем, есть просто: $\large |\Omega| = 2^n$ 

То есть в рамках схемы испытаний Бернулли, пространство элементарных исходов- состоит из каких-то $\omega$ маленьких. На сей раз $\omega$ маленькие, в отличие от классической ситуации, – это не какие-то равновероятные события, а это события, которые построены по совершенно понятной схеме. Это элементарные события, представляющие собой последовательности или векторы из ноликов и единичек, которых всего $2$ в степени $n$ штук. 

**Вывод:** И вот принципиальное отличие схемы испытаний Бернулли от классической схемы, с которой мы до сих пор имели дело, оно состоит в том, что эти элементарные исходы уже действительно не равновероятны.

### Вероятность возникновения элементарного исхода

С какой вероятностью возникает конкретный элементарный исход в рамках этой схемы испытаний Бернулли.

$\large P(\omega = (x_1, \ldots, x_n))$

Вот с какой вероятностью возникает совершенно конкретная последовательность из ноликов и единиц (эта последовательно их нулей и едениц обозначена ($x_1, \ldots, x_n$))? 

**С какой вероятностью возникает посследовательность из 1?**

Смотрите, если $x_1$ в этой конкретной последовательности – это единичка, тогда мы знаем, что эта единичка возникает с вероятностью $p$. И если вслед за $x_1$ появилось $x_2$, которое, тоже равняется единичке, то и эта еденица возникает с вероятностью $p$. Но поскольку бросание монетки в двух испытаниях очевидно были независимыми, то вероятности этих двух единиц нам надо просто перемножить. То есть, вероятность того, что $x_1$ и $x_2$ единичка, это уже $p$ умножить на $p$, то есть $p^2$. Если $x_1$ у нас ноль, и вслед за ним стоит единица, то точно так же надо $q$ умножить на $p$. Ну и так далее...

Понятно, что в результате число $p$ мы умножим само на себя столько раз, сколько вот в этой конкретной фиксированной последовательности есть единиц. А это очень легко сосчитать. Надо просто просуммировать по $i$ от единицы до $n$, $x_i$-ое.

$\large P(\omega = (x_1, \ldots, x_n)) = p^{\sum_{i=1}^{n} x_i}$

**C какой вероятностью возникает посследовательность из 0?**

Вот столько раз у нас там встречается единица, и каждая из этих единиц независимо от остальных возникает с вероятностью $p$. Поэтому $p$ надо возвести вот в эту степень, и дальше надо умножить $q$, то есть число ($1 – p$). И возвести в степень, а какая будет степень? Ну а сколько у нас осталось нулей? Понятное дело, всего координатных позиций $n$, и из них мы должны выкинуть те, которые равны единице. То есть, вот здесь получается количество нулей в нашей последовательности. 

$\large P(\omega = (x_1, \ldots, x_n)) = p^{\sum_{i=1}^{n} x_i} \cdot q^{n - \sum_{i=1}^{n} x_i}$ - Это есть вероятность возникновения конкретного элементарного исхода.

Вот тут, собственно, кроется принципиальная разница между классической вероятностной схемой, когда все элементарные исходы были равновозможны, рановероятны, и схемой испытаний Бернулли.

### Частный случай:

Есть один очень специальный, частный случай, когда $p$ (маленькая), вероятность успеха в схеме испытаний Бернулли, равняется $\frac{1}{2}$. В этом специальной случае $q$ тоже равняется $\frac{1}{2}$. Поэтому суммы благополучно друг друга сокращают:

$\large P(\omega=(x_1, \ldots, x_n)) = \frac{1}{2}^{\sum_{i=1}^{n} x_i} \cdot \frac{1}{2}^{n - \sum_{i=1}^{n} x_i} = \frac{1}{2}^n$

и у нас остается просто вероятность $\frac{1}{2}^n$, которую можно записать следующим образом $\frac{1}{2^n}$

Но вот если вспомнить, что у нас все $\Omega$ состоит из $2^n$ элементов, то тогда мы понимаем, что при $p = \frac{1}{2}$, то есть, когда монетка все-таки с несмещенным центром тяжести, тогда схема испытаний Бернулли, с которой мы сейчас работаем, это в точности классическое вероятностное пространство. Мы здесь единичку делим на общее количество элементарных исходов и получаем обычную классическую схему. 

**Вывод:** То есть, классическое вероятностное пространство, в принципе, является частным случаем схемы испытаний Бернулли, но только в очень специальной ситуации, когда $p = q = \frac{1}{2}$. А в более общей ситуации вероятности элементарных исходов не такие, как в классической ситуации. И это существенно отличает схему испытаний Бернулли от классической схемы.

<h1 style="color:#008B8B">Подсчет вероятности события наступления фиксированного количества успехов</h1>

**Определение события A:**

Рассмотрим событие $A$, которое состоит в том, что в $n$ испытаниях Бернулли ровно $k$ успехов. 

Что значит, что событие состоит в чем-то? Как и в классическом случае это означает, что нужно описать множество тех элементарных исходов, которые благоприятствуют реализации события $A$. То есть $A$, как и в классической ситуации – это просто множество $\large A = \{\omega_{i_1}, \ldots, \omega_{i_m} \} \in \Omega$ тех элементарных исходов среди всех возможных, каждый из которых благоприятствует событию $A$.

**Определение благоприятсвующих исходов:**

Какие элементарные исходы благоприятствуют этому событию?
Значит, в $n$ испытаниях Бернулли $k$ успехов. Это значит, мы должны рассмотреть только те последовательности из ноликов и единиц, в которых ровно $k$ единиц. То есть, иными словами, $A$ – это множество таких $\omega$ с координатами $x_1, \ldots, x_n$, что $x_i$ по-прежнему, это $0, 1$ и $x_1 + \ldots + x_n$ в точности равняется $k$:

$\large A = \{ \omega = (x_1, \ldots, x_n): x_i \in \{0, 1\}, x_1 + \ldots + x_n = k \}$

**Что такое m?**

Вот, мы рассматриваем только такие элементарные исходы. Ровно они и только они благоприятствуют нашему событию. То есть $m$ – это, собственно, количество таких $0,1$ векторов, у которых ровно $k$ единиц. Где $\omega$ - это различные вариатны векторов, которые обозначены $\omega = (x_1, \ldots, x_n)$. Значим $m$ равняется:

$\large m = |A| = C_n^k$ - Количество всех способов выбрать из $n$ координатных позиций $k$ координатных позиций, на которых будут стоять единички, ну а на остальных будут стоять нули соответственно. 

**Дополнительный вопрсо про m:**

Чему равна мощность события, заключающегося в том, что произошло либо ровно $m_1$ успехов, либо ровно $m_2$ успехов?

$\large C_n^{m_1} + C_n^{m_2}$

В событие $A$ входят все такие $\omega=(x_1,\ldots,x_n)$, что среди чисел $x_1,\ldots,x_n$ либо $m_1$, либо $m_2$ единиц. Количество $\omega$ с $m_1$  единицами равно $C_n^{m_1}$, с $m_2$ единицами - $C_n^{m_2}$. Поэтому всего таких событий $C_n^{m_1}+C_n^{m_2}$.

**Определение вероятности A:**

Тогда вероятность события $A$ естественно полагать равной cумме по всем элементарным исходам, которые благоприятствуют событию $A$, то есть принадлежать этому множеству $A$, вероятностей этих элементарных исходов:

$\large \sum\limits_{\omega \in A} P(\omega)$

Если бы все вероятности были равны, тогда мы получили бы просто мощность события $A$, поделенную на мощность всего $\Omega$, но поскольку в общем случае эти вероятности не совпадают между собой, тогда мы на самом деле получаем вот что:

мы получаем, что вероятность конкретного $P(\omega)$ – это $p^k$ умножить на $q^{n-k}$. Так как сумма равняется $k$, а мы уже записывали $p$ в степенеи суммы, где сумма - это $k$. Аналогично доказывали с $q^{n-k}$

Видно, что эта величина ни коим образом не зависит от того $\omega \in A$, по которому производится суммирование, поэтому мы так и пишем количество всех элементарных исходов, которое равняется $C_n^k$, умножить на величину вероятности одного исхода $\omega$:

$\large \sum\limits_{\omega \in A} P(\omega) = C_n^k \cdot p^{k} \cdot q^{n-k}$

Вот такая вот классическая, как говорят, биномиальная вероятность. Об этом мы еще обязательно чуть позже поговорим. Биномиальная вероятность, потому что тут возник, очевидно, кусочек бинома Ньютона.

<h3 style="color:#2E8B57" align="right">Задача о вероятности пересечения двух случайных множеств</h3>

**Формирование множества A**

Есть множество из $n$ элементов $\{1, 2, \ldots, n\}$. Из этого множества будем извлекать случайное подмножество $A$ по схеме Бернули. Формируем множество так: бросаем нашу любимую монетку со смещенным центром тяжести и с вероятностью успеха $p$. И если в первом бросании монетка падает решкой кверху, тогда берем элемент $1$ и кладем его в строящееся множество $A$. А если происходит неудача, тогда элемент $1$ в множество $A$ не кладем и так действуем до элемента $n$. 

$\large P(i \in A) = p$ - Веорятность того, что элемент $i$ попадает в мнжество $A$ независимо от того, какое $i$ мы взяли в этом множестве.

**Формирование множества B**

Теперь снова берем исходное множество $\{1, 2, \ldots, n\}$, как будто из него ничего не извлекалось. Из него извлекаем новое случайное множество $B$ согласно той же самой схеме испытаний Бернулли. Выбор элементов в строящееся множество $B$ осуществляется независимо от того, как выбирались элементы при построении множества $A$. 

**Постановка вопроса**

Они могут даже совпасть теоретически. Ведь когда мы строили $B$, мы никак не смотрели на то, как строилось $A$. Вот, с другой стороны, они могут и вовсе не пересекаться, например. Давайте зададимся как раз вот этим самым вопросом: с какой вероятностью эти два независимо выбранных случайных множества вообще не пересекаются? 

$\large P(A \cap B = \emptyset ) = ?$


**Интерпретация множества A и B**

Проинтерпретируем множество $A$ ровно так, как схема испытаний Бернулли устроена, то есть проинтерпретируем множество $A$ в терминах последовательности из ноликов и единичек. Ставим $1$ на первой позиции, если элемент с номером $1$ попал в множество $A$. Ноль ставим в случае, если элемент с номером $2$  не попал в множество $A$ и так деалее... 

$A = 1010111$ - Множество, которое состоит из элементов $1, 3, 5, 6, 7$ в данной случайно реализации. (Исходя из элементов изначального множества, так как $1$ означет, что элемент присутсвует в множестве $A$);

$B = 0011010$ - Множество, которое состоит из элементов $3, 4, 6$.

Очевиндо что два этих множества пересекаются в данной ситуации, так как столбец отвечающий за элемент $3$ как в множестве $A$ так и в множестве $B$ содержит $1$. В нашем случае, еденицы присутвуют у элементов $3$ и $6$ исходного множества:

$$\large 10\textbf{1}01\textbf{1}1$$

$$\large 00\textbf{1}10\textbf{1}0$$


$\large P(A \cap B = \emptyset)$ тогда и только тогда, когда в каждом столбце данной таблицы из нулей и единиц, не присутсвует две еденицы.

**Нахождение интересующей веротности**

Столбцы появляются независимо друг от друга, потому что элементы внутри множества $A$ выбирались независимо друг от друга, и элементы внутри множества $B$ выбирались независимо друг от друга. 

Вероятность того, что в данном конкретном столбце нет двух единиц:

$\large 1 - p^2$, где 

$p^2$ - Это вероятность того, что в столбце стоит две еденицы. Почему $p$ в квадрате? вспомним с какой вероятностью один элемент попадает в множество $A$ - это $p$. А в множество $B$ элемент попадает тоже с вероятностью $p$. Соответсвенно, элемент попапдает одновременно в множество $A$ и в множество $B$ с вероятностью $p \cdot p = p^2$.

А $1-p^2$ это противоположная вероятность, верояность того, что две еденицы не стоят в одном столбце.

А вероятность того, что в каждом столбце нет двух еденицы - это надо за счет независимости перемножить, то есть, другими словами возвести в степень $n$.

$\large P(A \cap B = \emptyset) = (1 - p^2)^n$
