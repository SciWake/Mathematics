<h1 style="color:black" align="center">2. Элементарная теория вероятностей: случайные величины</h1>

<h2 style="color:black" align="center">2.1 Случайные величины</h2>

### Случайная величина

Во многих задачах не важна природа элементарных событий, а важные численные характеристики связанные с элементарными событиями. Например, количество успехов при многократном подбрасывании монеты. Связи с этим, в теории вероятностей изучаются случайные величины.

$\Omega = \{\omega_1, \ldots, \omega_N\}$ - Пространство элементарных событий.

Случайная величина - это произвольная фукнция, которая задана на пространсвет $\Omega$ и в качестве значений принимает вещественные числа:

$\xi : \Omega \to \mathbb{R}$ - Случайная величина. 

**Пример.** $\Omega = \{(i, j): i, j = 1, 2, \ldots, 6\}$ - имеется два кубика, мы подбрасываем два кубика и на них выпадает два числа $i, j$. Тогда случайно величиной будет сумма очков выпавщих на кубиках или произведение очков:

$\xi = i + j \quad \eta = i \cdot j$ - случайные величины

### Распределение случайной величины

**Распределение случайной величины $\xi$** - набор вероятностей, с которыми она принимает свои значения $P(\xi = a)$. То есть, мы рассматриваем всевозможные $a$, которые лежат в множестве значений этой случайной величины и смотрим на вероятности событий таких, что случайна величина $\xi$ принимает значение $a$. То есть нас интересует, вероятность того, что $\omega$ такая, что $\xi$ от $\omega$ есть $a$: $\{\omega \in \Omega: \xi(\omega) = a\}$ и набор таких вероятностей есть распределение случайной величины.

Так как мы рассматриваем случай, когда пространство элементарных исходов состоит из конечно числа элементов, то и множество значений случайной величины должно быть конечным, соответсвенно множество чисел, которое мы можем взять в качестве $a$ оно тоже является конечным.

**Распределение случайной величины $\xi, \eta$ для примера**

Всевозможные значения, которые могу принимать случайные величины $\xi, \eta$ = $1, 2, \ldots, 36$.

Распределение вероятностей для случайной величины $\xi$, то есть вероятности того, что случайная величина принимает соответствующее значение:

$\large \frac{1}{36} \frac{2}{36} \frac{3}{36} \frac{4}{36} \frac{5}{36} \frac{6}{36} \frac{5}{36} \frac{4}{36} \frac{3}{36} \frac{2}{36} \frac{1}{36}$ сколько способов получить число $4$? Оно может получиться как $1+3=2+2=3+1$, следовательно $3$ варината из $36$ нам подходят и получается $4$ возникает с вероятностью $\frac{3}{36}$. И таким образом получаем распределение вероятностей.

Распределение вероятностей для случайной величины $\eta$ более замысловатое, так как если мы берем конеретное число, то количество способов представить его в виде произведение двух чисел от $1$ до $6$ куда большим числом способов, чем сумму:

$\large \frac{1}{36} \frac{2}{36} \frac{2}{36} \frac{3}{36} \frac{2}{36} \frac{4}{36}
\quad
\frac{2}{36} \frac{1}{36} \frac{2}{36}
\quad
\large \frac{4}{36} \frac{2}{36} \frac{1}{36} \frac{2}{36} \frac{2}{36} \frac{2}{36} \frac{1}{36} \frac{2}{36} \frac{1}{36}$

### Совместное распределение случайных велечин

$\xi$ и $\eta$ - случайные величины.

Совместрым распределением велечин $\eta \text{ и } \xi$ называется набор вероятностей таких событий, что $P(\xi = a, \eta = b)$, где $a$ пробигает всевозможные значения случайной величины $\xi$, а $b$ пробигает всевозсожные значения величины $\eta$.

### Независимые случайные величины

Случайные величины мы назовем независимыми, если вероятность двух случайных величина равна произведению вероятностей, того случайная величина $\xi$ принимает значение $a$, а случайная величина $\eta$ принимает значение $b$: 

$P(\xi = a, \eta = b) = P(\xi = a) P(\eta = b)$, то есть, при всех $a$ и $b$ из множества значений $\xi, \eta$, события $\{\omega: \xi(\omega) = a\}$ и $\{\omega: \eta(\omega) = b\}$ случайных величин называются независимыми.

По аналогии можно определить независимость в совокупности большого количества величин, для этого необходимо завести совместное распределение большего количества случайных величин...

**Пример 1.** Очки на одном кубике и очки на другом кубике - независимы.

**Пример 2.** Сумма и произведение очков - зависимы:

$P(\xi =2, \eta = 1) = \frac{1}{36} \ne (\frac{1}{36})^2 = P(\xi = 2) P(\eta = 1)$

<h3 style="color:#2E8B57">Задача № 1</h3>

В урне $a$ черных шаров и $b$ белых. Из нее последовательно вынимаются два шара. Случайная величина $\xi$ равна количеству вынутых черных шаров. Найдите распределение этой случайной величины. Если в ответе значениям $0$, $1$ и $2$ соответствуют вероятности $p_0, p_1$ и $p_2$, то введите в систему выражение $p_0\cdot x+p_1\cdot y+p_2\cdot z$

<h4 align='center'>Решение:</h4>

$P(\xi=0)=\frac{b(b-1)}{(a+b)(a+b-1)}$, $P(\xi=1)=\frac{2ab}{(a+b)(a+b-1)}P$ и $P(\xi=2)=\frac{a(a-1)}{(a+b)(a+b-1)}$

Случайная величина $\xi$ может принимать лишь значения $0$, $1$ и $2$. 


Значение $0$ соответствует событию, когда оба вынутых шара — белые. Вероятность вынуть первым белый шар равна $\frac{b}{a+b}$, после этого остается $b-1$ белый шар из $a+b-1$ шара, поэтому вероятность вынуть вторым белый шар (при условии, что первый - белый) равна $\frac{b-1}{a+b-1}$. Таким образом, вероятность вынуть оба белых шара равна $\frac{b(b-1)}{(a+b)(a+b-1)}$ и, значит, $P(\xi=0)=\frac{b(b-1)}{(a+b)(a+b-1)}$. 

Аналогично находим, что вероятность вынуть оба черных шара равна $\frac{a(a-1)}{(a+b)(a+b-1)}$, поэтому $P(\xi=2)=\frac{a(a-1)}{(a+b)(a+b-1)}$. Следовательно,

$$\begin{aligned} P(\xi=1) &= 1-P(\xi=0)-P(\xi=2) = \\ &= 1-\frac{b(b-1)}{(a+b)(a+b-1)}-\frac{a(a-1)}{(a+b)(a+b-1)}=\frac{2ab}{(a+b)(a+b-1)} . \end{aligned}$$

$P(\xi=1)=?$, сначало можем выбрать один черный шар, вслед за ним выбрать белый и ещё есть вариант сначало выбрать белый, а потом черный: $P(\xi=1) = \frac{a}{a+b} \cdot \frac{b}{a+b-1} + \frac{b}{a+b} \cdot \frac{a}{a+b-1} = \frac{2ab}{(a+b)(a+b-1)}$

Либо, $(\xi=1) = \frac{a b}{C_{a+b}^2} = \frac{2ab}{(a+b)(a+b-1)}$

$C_{a+b}^2 = \frac{(a+b)!}{(a+b-2)!2!} = \frac{(a+b)!}{(a+b-2)!} \cdot \frac{1}{2!} = (a+b)(a+b-1) \frac{1}{2!}$

<h3 style="color:#2E8B57">Задача № 2</h3>

В урне $a$ черных шаров и $b$ белых. Из нее по схеме с возвращением вынимаются два шара. Случайная величина $\xi$ равна количеству вынутых черных шаров. Найдите распределение этой случайной величины. Если в ответе значениям $0$, $1$ и $2$ соответствуют вероятности $p_0, p_1$ и $p_2$, то введите в систему выражение $p_0 \cdot x+p_1 \cdot y+p_2 \cdot z$

<h4 align='center'>Решение:</h4>

$P(\xi=2)=\frac{b^2}{(a+b)^2}$, $P(\xi=1)=\frac{2ab}{(a+b)^2}$ и $P(\xi=2)=\frac{a^2}{(a+b)^2}$

Случайная величина $\xi$ может принимать лишь значения $0$, $1$ и $2$. Значение $0$ соответствует событию, когда оба вынутых шара - белые. Вероятность вынуть белый шар равна $\frac{b}{a+b}$, поэтому вероятность оба раза вынуть белый шар равна $\frac{b^2}{(a+b)^2}$. Таким образом, $P(\xi=0)=\frac{b^2}{(a+b)^2}$. Аналогично находим, что вероятность вынуть оба раза черный шара равна $\frac{a^2}{(a+b)^2}$, поэтому $P(\xi=2)=\frac{a^2}{(a+b)^2}$. Следовательно,

$$P(\xi=1) = 1-P(\xi=0)-P(\xi=2) = 1-\frac{b^2}{(a+b)^2}-\frac{a^2}{(a+b)^2}=\frac{2ab}{(a+b)^2}$$

<h3 style="color:#2E8B57">Задача № 3</h3>

Подбрасываются три правильных игральных кубика, $\xi$, $\eta$ и $\zeta$ — количество очков, выпавших на первом, втором и третьем кубике. Отметьте пары независимых случайных величин.

<h4 align='center'>Решение:</h4>

1. $\xi$ и $\eta$

Верно.

2. $\xi$ и $\xi+\eta$

Неверно, поскольку, например $P(\xi=1)=\frac{1}{6}$ и $P(\xi+\eta=12)=\frac{1}{36}$, но $P(\xi=1,\xi+\eta=12)=0.$

3. $\xi$ и $\eta+\zeta$

Верно.

4. $\xi+\eta$ и $\eta+\zeta$

Неверно, поскольку, например $P(\xi+\eta=2)=\frac{1}{36}$ и $P(\eta+\zeta=12)=\frac{1}{36}$, но $P(\xi+\eta=2,\eta+\zeta=12)=0$

5. $\xi^2$ и $\eta^3$
 
Верно, поскольку $P(\xi^2=a)=P(\xi=\sqrt{a})$ и $P(\eta^3=b)=P(\eta=\sqrt[3]{b})$, а случайные величины $\xi$ и $\eta$ независимы.

6. $\xi+\eta$ и $\zeta^2$

Верно.

7. $7-\xi$ и $\eta+\zeta$

Верно.

8. $\min\{\xi,\eta\}$ и $\max\{\xi,\zeta\}$

Неверно, поскольку, например $P(\min\{\xi,\eta\}=6)=\frac{1}{36}$ и $P(\max\{\xi,\zeta\}=1)=\frac{1}{36}$, но $P(\min\{\xi,\eta\}=6,\max\{\xi,\zeta\}=1)=0$.

9. $\min\{\xi,\eta\}$ и $\max\{\xi,\eta\}$

Неверно, поскольку, например $P(\min\{\xi,\eta\}=6)=\frac{1}{36}$ и $P(\max\{\xi,\eta\}=1)=\frac{1}{36}$, но $P(\min\{\xi,\eta\}=6,\max\{\xi,\eta\}=1)=0$
