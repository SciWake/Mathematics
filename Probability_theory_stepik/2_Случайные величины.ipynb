<h1 style="color:black" align="center">2. Элементарная теория вероятностей: случайные величины</h1>

<h2 style="color:black" align="center">2.1 Случайные величины</h2>

### Случайная величина

Во многих задачах не важна природа элементарных событий, а важные численные характеристики связанные с элементарными событиями. Например, количество успехов при многократном подбрасывании монеты. Связи с этим, в теории вероятностей изучаются случайные величины.

$\Omega = \{\omega_1, \ldots, \omega_N\}$ - Пространство элементарных событий.

Случайная величина - это произвольная фукнция, которая задана на пространсвет $\Omega$ и в качестве значений принимает вещественные числа:

$\xi : \Omega \to \mathbb{R}$ - Случайная величина. 

**Пример.** $\Omega = \{(i, j): i, j = 1, 2, \ldots, 6\}$ - имеется два кубика, мы подбрасываем два кубика и на них выпадает два числа $i, j$. Тогда случайной величиной будет сумма очков выпавщих на кубиках или произведение очков:

$\xi = i + j \quad \eta = i \cdot j$ - случайные величины

### Распределение случайной величины

**Распределение случайной величины $\xi$** - набор вероятностей, с которыми она принимает свои значения $P(\xi = a)$. То есть, мы рассматриваем всевозможные $a$, которые лежат в множестве значений этой случайной величины и смотрим на вероятности событий таких, что случайна величина $\xi$ принимает значение $a$. То есть нас интересует, вероятность того, что $\omega$ такая, что $\xi$ от $\omega$ есть $a$: $\{\omega \in \Omega: \xi(\omega) = a\}$ и набор таких вероятностей есть распределение случайной величины.

Так как мы рассматриваем случай, когда пространство элементарных исходов состоит из конечно числа элементов, то и множество значений случайной величины должно быть конечным, соответсвенно множество чисел, которое мы можем взять в качестве $a$ оно тоже является конечным.

**Распределение случайной величины $\xi, \eta$ для примера**

Всевозможные значения, которые могу принимать случайные величины $\xi, \eta$ = $1, 2, \ldots, 36$.

Распределение вероятностей для случайной величины $\xi$, то есть вероятности того, что случайная величина принимает соответствующее значение:

$\large \frac{1}{36} \frac{2}{36} \frac{3}{36} \frac{4}{36} \frac{5}{36} \frac{6}{36} \frac{5}{36} \frac{4}{36} \frac{3}{36} \frac{2}{36} \frac{1}{36}$ сколько способов получить число $4$? Оно может получиться как $1+3=2+2=3+1$, следовательно $3$ варината из $36$ нам подходят и получается $4$ возникает с вероятностью $\frac{3}{36}$. И таким образом получаем распределение вероятностей.

Распределение вероятностей для случайной величины $\eta$ более замысловатое, так как если мы берем конеретное число, то количество способов представить его в виде произведение двух чисел от $1$ до $6$ куда большим числом способов, чем сумму:

$\large \frac{1}{36} \frac{2}{36} \frac{2}{36} \frac{3}{36} \frac{2}{36} \frac{4}{36}
\quad
\frac{2}{36} \frac{1}{36} \frac{2}{36}
\quad
\large \frac{4}{36} \frac{2}{36} \frac{1}{36} \frac{2}{36} \frac{2}{36} \frac{2}{36} \frac{1}{36} \frac{2}{36} \frac{1}{36}$

### Совместное распределение случайных велечин

$\xi$ и $\eta$ - случайные величины.

Совместрым распределением велечин $\eta \text{ и } \xi$ называется набор вероятностей таких событий, что $P(\xi = a, \eta = b)$, где $a$ пробигает всевозможные значения случайной величины $\xi$, а $b$ пробигает всевозсожные значения величины $\eta$.

### Независимые случайные величины

Случайные величины мы назовем независимыми, если вероятность двух случайных величина равна произведению вероятностей, того случайная величина $\xi$ принимает значение $a$, а случайная величина $\eta$ принимает значение $b$: 

$P(\xi = a, \eta = b) = P(\xi = a) P(\eta = b)$, то есть, при всех $a$ и $b$ из множества значений $\xi, \eta$, события $\{\omega: \xi(\omega) = a\}$ и $\{\omega: \eta(\omega) = b\}$ случайных величин называются независимыми.

По аналогии можно определить независимость в совокупности большого количества величин, для этого необходимо завести совместное распределение большего количества случайных величин...

**Пример 1.** Очки на одном кубике и очки на другом кубике - независимы.

**Пример 2.** Сумма и произведение очков - зависимы:

$P(\xi =2, \eta = 1) = \frac{1}{36} \ne (\frac{1}{36})^2 = P(\xi = 2) P(\eta = 1)$

<h3 style="color:#2E8B57">Задача № 1</h3>

В урне $a$ черных шаров и $b$ белых. Из нее последовательно вынимаются два шара. Случайная величина $\xi$ равна количеству вынутых черных шаров. Найдите распределение этой случайной величины. Если в ответе значениям $0$, $1$ и $2$ соответствуют вероятности $p_0, p_1$ и $p_2$, то введите в систему выражение $p_0\cdot x+p_1\cdot y+p_2\cdot z$

<h4 align='center'>Решение:</h4>

$P(\xi=0)=\frac{b(b-1)}{(a+b)(a+b-1)}$, $P(\xi=1)=\frac{2ab}{(a+b)(a+b-1)}P$ и $P(\xi=2)=\frac{a(a-1)}{(a+b)(a+b-1)}$

Случайная величина $\xi$ может принимать лишь значения $0$, $1$ и $2$. 


Значение $0$ соответствует событию, когда оба вынутых шара — белые. Вероятность вынуть первым белый шар равна $\frac{b}{a+b}$, после этого остается $b-1$ белый шар из $a+b-1$ шара, поэтому вероятность вынуть вторым белый шар (при условии, что первый - белый) равна $\frac{b-1}{a+b-1}$. Таким образом, вероятность вынуть оба белых шара равна $\frac{b(b-1)}{(a+b)(a+b-1)}$ и, значит, $P(\xi=0)=\frac{b(b-1)}{(a+b)(a+b-1)}$. 

Аналогично находим, что вероятность вынуть оба черных шара равна $\frac{a(a-1)}{(a+b)(a+b-1)}$, поэтому $P(\xi=2)=\frac{a(a-1)}{(a+b)(a+b-1)}$. Следовательно,

$$\begin{aligned} P(\xi=1) &= 1-P(\xi=0)-P(\xi=2) = \\ &= 1-\frac{b(b-1)}{(a+b)(a+b-1)}-\frac{a(a-1)}{(a+b)(a+b-1)}=\frac{2ab}{(a+b)(a+b-1)} . \end{aligned}$$

$P(\xi=1)=?$, сначало можем выбрать один черный шар, вслед за ним выбрать белый и ещё есть вариант сначало выбрать белый, а потом черный: $P(\xi=1) = \frac{a}{a+b} \cdot \frac{b}{a+b-1} + \frac{b}{a+b} \cdot \frac{a}{a+b-1} = \frac{2ab}{(a+b)(a+b-1)}$

Либо, $(\xi=1) = \frac{a b}{C_{a+b}^2} = \frac{2ab}{(a+b)(a+b-1)}$

$C_{a+b}^2 = \frac{(a+b)!}{(a+b-2)!2!} = \frac{(a+b)!}{(a+b-2)!} \cdot \frac{1}{2!} = (a+b)(a+b-1) \frac{1}{2!}$

<h3 style="color:#2E8B57">Задача № 2</h3>

В урне $a$ черных шаров и $b$ белых. Из нее по схеме с возвращением вынимаются два шара. Случайная величина $\xi$ равна количеству вынутых черных шаров. Найдите распределение этой случайной величины. Если в ответе значениям $0$, $1$ и $2$ соответствуют вероятности $p_0, p_1$ и $p_2$, то введите в систему выражение $p_0 \cdot x+p_1 \cdot y+p_2 \cdot z$

<h4 align='center'>Решение:</h4>

$P(\xi=2)=\frac{b^2}{(a+b)^2}$, $P(\xi=1)=\frac{2ab}{(a+b)^2}$ и $P(\xi=2)=\frac{a^2}{(a+b)^2}$

Случайная величина $\xi$ может принимать лишь значения $0$, $1$ и $2$. Значение $0$ соответствует событию, когда оба вынутых шара - белые. Вероятность вынуть белый шар равна $\frac{b}{a+b}$, поэтому вероятность оба раза вынуть белый шар равна $\frac{b^2}{(a+b)^2}$. Таким образом, $P(\xi=0)=\frac{b^2}{(a+b)^2}$. Аналогично находим, что вероятность вынуть оба раза черный шара равна $\frac{a^2}{(a+b)^2}$, поэтому $P(\xi=2)=\frac{a^2}{(a+b)^2}$. Следовательно,

$$P(\xi=1) = 1-P(\xi=0)-P(\xi=2) = 1-\frac{b^2}{(a+b)^2}-\frac{a^2}{(a+b)^2}=\frac{2ab}{(a+b)^2}$$

<h3 style="color:#2E8B57">Задача № 3</h3>

Подбрасываются три правильных игральных кубика, $\xi$, $\eta$ и $\zeta$ — количество очков, выпавших на первом, втором и третьем кубике. Отметьте пары независимых случайных величин.

<h4 align='center'>Решение:</h4>

1. $\xi$ и $\eta$

Верно.

2. $\xi$ и $\xi+\eta$

Неверно, поскольку, например $P(\xi=1)=\frac{1}{6}$ и $P(\xi+\eta=12)=\frac{1}{36}$, но $P(\xi=1,\xi+\eta=12)=0.$

3. $\xi$ и $\eta+\zeta$

Верно.

4. $\xi+\eta$ и $\eta+\zeta$

Неверно, поскольку, например $P(\xi+\eta=2)=\frac{1}{36}$ и $P(\eta+\zeta=12)=\frac{1}{36}$, но $P(\xi+\eta=2,\eta+\zeta=12)=0$

5. $\xi^2$ и $\eta^3$
 
Верно, поскольку $P(\xi^2=a)=P(\xi=\sqrt{a})$ и $P(\eta^3=b)=P(\eta=\sqrt[3]{b})$, а случайные величины $\xi$ и $\eta$ независимы.

6. $\xi+\eta$ и $\zeta^2$

Верно.

7. $7-\xi$ и $\eta+\zeta$

Верно.

8. $\min\{\xi,\eta\}$ и $\max\{\xi,\zeta\}$

Неверно, поскольку, например $P(\min\{\xi,\eta\}=6)=\frac{1}{36}$ и $P(\max\{\xi,\zeta\}=1)=\frac{1}{36}$, но $P(\min\{\xi,\eta\}=6,\max\{\xi,\zeta\}=1)=0$.

9. $\min\{\xi,\eta\}$ и $\max\{\xi,\eta\}$

Неверно, поскольку, например $P(\min\{\xi,\eta\}=6)=\frac{1}{36}$ и $P(\max\{\xi,\eta\}=1)=\frac{1}{36}$, но $P(\min\{\xi,\eta\}=6,\max\{\xi,\eta\}=1)=0$

<h3 style="color:#2E8B57">Задача № 1</h3>

Из $50$ студентов, сдававших экзамен по теории вероятностей, $12$ получили оценку отлично, $20$ — оценку хорошо, $7$ — удовлетворительно, а остальные студенты экзамен не сдали. После заполнения ведомости экзаменатор наугад выбирает строчку и смотрит на выставленную в ней оценку. Найдите математическое ожидание и медиану данной случайной величины. В ответе приведите найденные значения, разделенные пробелом (для записи используйте десятичные дроби).

<h4 align='center'>Решение:</h4>

Вероятность выбрать каждую из $50$ строчек равна $\frac{1}{50}$. Поэтому вероятность выбора пятерки равна $\frac{12}{50}$, четверки — $\frac{20}{50}$, тройки — $\frac{7}{50}$ и двойки равна $1-\frac{12}{50}-\frac{20}{50}-\frac{7}{50}=\frac{11}{50}$. Поэтому математическое ожидание будет равно

$5\cdot\frac{12}{50}+4\cdot\frac{20}{50}+3\cdot\frac{7}{50}+2\cdot\frac{11}{50} = \frac{183}{50}$.

Для того, чтобы найти медиану, заметим, что вероятность получить хотя бы пятерку равна $\frac{12}{50}\lt \frac{1}{2}$, а вероятность получить хотя бы четверку равна $\frac{12}{50}+\frac{20}{50}=\frac{32}{50}\geqslant\frac{1}{2}$. Следовательно, медиана не больше четырех. С другой стороны вероятность получить не больше тройки равна $\frac{7}{50}+\frac{11}{50}=\frac{18}{50}\lt \frac{1}{2}$, а вероятность получить не больше четверки равна $1-\frac{12}{50}\geqslant\frac{1}{2}$. Поэтому медиана и не меньше четверки.

<h3 style="color:#2E8B57">Задача № 2</h3>

Случайная величина $\xi$ принимает все натуральные значения от $1$ до $1000$. Вероятность значения $k$ пропорциональна $k$. Найдите медиану и математическое ожидание $\xi$. В ответе укажите найденные числа, разделенные пробелами.

<h4 align='center'>Решение:</h4>

Пусть $n=1000$. Сначала найдем вероятность события $P(\xi=k)$ при $k$ от 
$1$ до $n$. По условию вероятность значения $k$ пропорциональна $k$, поэтому $P(\xi=k)=ck$. С другой стороны

$1=\sum\limits_{k=1}^n P(\xi=k)=\sum\limits_{k=1}^n ck=c\frac{n(n+1)}{2}$,
откуда $c=\frac{2}{n(n+1)}$. Найдем математическое ожидание:
$\mathbb{E}\xi = \sum\limits_{k=1}^n kP(\xi=k) = \sum\limits_{k=1}^n ck^2 = \frac{2}{n(n+1)}\cdot\frac{n(n+1)(2n+1)}{6} = \frac{2n+1}{3}$. Осталось заметить, что $\frac{2\cdot1000+1}{3}=667$.

Для того, чтобы найти медиану, вычислим вероятность события $P(\xi\leqslant m)$ для натурального $m$:
$P(\xi\leqslant m) = \sum\limits_{k=1}^m P(\xi=k) = \sum\limits_{k=1}^m ck = \frac{2}{n(n+1)}\cdot\frac{m(m+1)}{2} = \frac{m(m+1)}{n(n+1)}$.Следовательно, вероятность события

$P(\xi\geqslant m)=1-P(\xi\leqslant m-1)=1-\frac{(m-1)m}{n(n+1)}$.

Число $m$ будет медианой, если $\frac{m(m+1)}{n(n+1)}\geqslant\frac{1}{2}$ и $\frac{(m-1)m}{n(n+1)}\leqslant\frac{1}{2}$. Таким образом, нужно найти такое $m$, что $m(m-1)\leqslant\frac{n(n+1)}{2}=500\cdot1001\leqslant m(m+1)$. Отсюда видно, что $m$ должно быть близко к $\sqrt{500\cdot1001}\approx 707{,}46$. Непосредственным вычислением убеждаемся, что число $707$ подходит.

### Свойства математических ожиданий

**Свойство 1.** $\large \xi \ge 0 \Rightarrow \mathbb{E} \ge 0$ это следует из формулы по определению:

$$\large \mathbb{E} \xi = \sum\limits_{x \in \xi(\Omega)} x \cdot P(\xi = x)$$

Если случайная величина не отрицательная, тогда $x \ge 0, P(\xi = x) \ge 0$, поэтому и математическое ожидание больше или равно нулю.

**Свойство 2.** $\large \mathbb{E}(a\xi + b\eta) = a\mathbb{E}\xi + b\mathbb{E}\eta$

**Свойство 3.** $\large \xi \ge \eta \Rightarrow \mathbb{E}\xi \ge \mathbb{E}\eta$ 

Это свойство выводится из 1 и 2. Если, случайная величина $\xi \ge \eta$, это означает для каждого $\omega$ значение случайной величи $\xi(\omega)$ больше значения $\eta(\omega)$. Посмотрим на разность математических ожиданий, которые сварачиваем по свойству линейность:

$\large \mathbb{E}\xi - \mathbb{E}\eta = \mathbb{E}(\xi - \eta) \ge 0$

Но это не отрицательная случайная величина и по первому свойству, математическое ожидание у случайной величины больше либо равно нулю.

**Свойство 4.** $\large |\mathbb{E}\xi| \le \mathbb{E}|\xi|$ выведем из 3-го свойства. Запишем следующее:

$\large -|\xi| \le \xi \le |\xi| \Rightarrow -\mathbb{E}|\xi| \le \mathbb{E}\xi \le \mathbb{E}|\xi|$


**Свойство 5.** $\large (\mathbb{E}|\xi\eta|)^2 \le \mathbb{E}\xi^2 \cdot \mathbb{E}\eta^2$


**Свойство 6.** Если $\large \xi$ и $\eta$ независимы, то $\large \mathbb{E}(\xi\eta) = \mathbb{E}\xi \mathbb{E}\eta$

**Доказательство  2. $\large \mathbb{E}(a\xi + b\eta) = a\mathbb{E}\xi + b\mathbb{E}\eta$**

Распишем математическое ожидание по определению:

$\large \mathbb{E}(a\xi + b\eta) 
= 
\sum\limits_{\omega \in \Omega} (a\xi(\omega) + b\eta(\omega)) P(\omega) 
= 
a \underbrace{
    \sum\limits_{\omega \in \Omega} \xi(\omega)P(\omega)
    }_{\mathbb{E}\xi}
+ 
b \underbrace{
    \sum\limits_{\omega \in \Omega} \eta(\omega)P(\omega)
    }_{\mathbb{E}\eta}$

**Доказательство  5. $\large (\mathbb{E}|\xi\eta|)^2 \le \mathbb{E}\xi^2 \cdot \mathbb{E}\eta^2$** - Неравенство Коши-Буняковского

Рассмотрим величину $f(t)$, где $t$ - это произвольное вещественное число. Раскроем рассмотренное математическое ожидание по линейности:

$\large f(t) = \mathbb{E}((\xi + t\eta)^2) = \mathbb{E}(\xi^2 + 2t\xi\eta + t^2\eta^2) = \mathbb{E}\xi^2 + 2t\mathbb{E}(\xi\eta) + t^2\mathbb{E}\eta^2$

Получаем, что $f(t)$ - квадратный трехчлен, мы знаем что это математическое ожидание некоторого неотрицательной случайной величины, следовательно этот квадратныхй трехчлен неотрицателен при любом $t$, иными словами: $f(t) \ge 0$. Так как квадратный трехчлен не отрицателен и его старший коэффициент тоже не отрциателен, тогда он принимает лишь не отрицательные значения и его дискриминант меньше либо равен нулю. Запишем его дискриминант и после сократим:

$\large \Rightarrow (2\mathbb{E}(\xi\eta))^2 - 4\mathbb{E}\xi^2\mathbb{E}\eta^2 \le 0 \Rightarrow (\mathbb{E}(\xi\eta))^2 \le \mathbb{E}\xi^2\mathbb{E}\eta^2$

В это неравенство вместо $\xi, \eta$ их модули:

$\large (\mathbb{E}|\xi\eta|)^2 \le \mathbb{E}|\xi|^2 \mathbb{E}|\eta|^2 = \mathbb{E}\xi^2 \mathbb{E}\eta^2$

Тогда, $\large (\mathbb{E}|\xi\eta|)^2 \ge |\mathbb{E}(\xi\eta)|^2$

Мы убедились, что $(\mathbb{E}|\xi\eta|)^2 \ge |\mathbb{E}(\xi\eta)|^2$, поэтому после возведения в квадрат, данное неравенство сохраниться и смотря по цепочке убеждаемся в том, что 5-е свойство доказано.

**Доказательство 6. Если $\xi$ и $\eta$ независимы, то $\large \mathbb{E}(\xi\eta) = \mathbb{E}\xi \mathbb{E}\eta$**

Посмотрим как устроены значения у случайной величины $\xi\cdot\eta$, для этого необходимо взять значения случайной величины $\xi$, например, $x$ и умножить его на $y$, который является значением случайной величины $\eta$. И всевозможные значения случайной величины $\xi\eta$ - это есть всевозможные $x, y$, где $x$ это значение $\xi$, а $y$ значение $\eta$. То есть, математическое ожидание может быть посчитано по следующей формуле:

$\large \mathbb{E}(\xi\eta) = \sum\limits_{x \in \xi({\Omega}) \\y \in \eta(\Omega)} xy P(\xi=x, \eta=y) =$

Так как случайная величина независима, тогда:

$\large = \sum\limits_{x \in \xi({\Omega}) \\y \in \eta(\Omega)} xy P(\xi=x)P(\eta = y)= \sum\limits_{x\in\xi(\Omega)} x P(\xi=x) \sum\limits_{y\in\eta(\Omega)} y P(\eta=y) = \mathbb{E}\xi\mathbb{E}\eta$

**Вернемся к примеру 1.** $\xi_1$ и $\xi_2$ очки, выпавшие на 1-м и 2-м кубиках.

$\large \mathbb{E}\xi_1 = \mathbb{E}\xi_2 =\frac{1+2+3+4+5+6}{6} = 3.5$

$\large \xi = \xi_1 + \xi_2$ - сумма очков на кубиках

$\large \mathbb{E}\xi = \mathbb{E}\xi_1 + \mathbb{E}\xi_2 = 7$

$\large \eta = \xi_1 \xi_2$ - произведения очков на кубиках

$\large \mathbb{E}\eta = \mathbb{E}\xi_1 \mathbb{E}\xi_2 = 3.5^2 = 12.25$

<h3 style="color:#2E8B57">Задача № 3</h3>

В урне $a$ белых шаров и $b$ черных. Из нее вынимаются два шара

* по схеме с возвращением
* по схеме без возвращения.

Случайные величины $\xi$ и $\eta$ равны количеству вынутых черных шаров (соответственно в первом и во втором случае). Найдите математическое ожидание случайных величин $\xi$ и $\eta$. Если в ответе получились значения $A$ и $B$, то введите в систему выражение $A\cdot x+B \cdot y$.

Постарайтесь понять, почему получился именно такой ответ.

<h4 align='center'>Решение:</h4>

$\mathbb{E}\xi = \mathbb{E}\eta = \frac{2b}{a+b}$

По второй задаче предыдущего модуля (там переставлены местами буквы $a$ и $b$)

$P(\xi=0)=\frac{a^2}{(a+b)^2}, \qquad P(\xi=1)=\frac{2ab}{(a+b)^2} \text{ и } P(\xi=2)=\frac{b^2}{(a+b)^2}$

поэтому

$\mathbb{E}\xi=0\cdot P(\xi=0)+1\cdot P(\xi=1)+2\cdot P(\xi=2)=\frac{2ab}{(a+b)^2}+2\cdot\frac{b^2}{(a+b)^2} = \dfrac{2b}{a+b}$

По первой задаче предыдущего модуля (там также переставлены местами буквы $a$ и $b$)

$P(\xi=0)=\frac{a(a-1)}{(a+b)(a+b-1)}, \ P(\xi=1)=\frac{2ab}{(a+b)(a+b-1)} \text{ и } P(\xi=2)=\frac{b(b-1)}{(a+b)(a+b-1)}$

поэтому

$\mathbb{E}\eta=0\cdot P(\xi=0)+1\cdot P(\xi=1)+2\cdot P(\xi=2)=\frac{2ab}{(a+b)(a+b-1)}+2\cdot\frac{b(b-1)}{(a+b)(a+b-1)} = \dfrac{2b}{a+b}$

Замечание. То, что математические ожидания оказались равны, объясняется тем, что случайные величины можно представить в виде суммы двух случайных величин $\xi_1$ и $\xi_2$ ($\eta_1$ и $\eta_2$), которые равны $1$, если при соответствующем вытаскивании шара он оказался черным. Очевидно, что $\mathbb{E}\xi_i=P(\xi_i=1)$ и $\mathbb{E}\eta_i=P(\eta_i=1)$. Но вероятность вытащить черный шар равна $\frac{b}{a+b}$ и не зависит от того с возвращением вынимаются шары или нет, поскольку в случае выемки шаров без возвращения никакой информации про другие вынутые шары случайная величина $\eta_i$ в себе не несет. Стало быть,

$\mathbb{E}\xi_1=\mathbb{E}\xi_2=\mathbb{E}\eta_1=\mathbb{E}\eta_2=\frac{b}{a+b}$

и

$\mathbb{E}\xi = \mathbb{E}\xi_1+\mathbb{E}\xi_2=\mathbb{E}\eta_1+\mathbb{E}\eta_2=\mathbb{E}\eta$

<h3 style="color:#2E8B57">Задача № 4</h3>

Водопроводчик Вася поздно вечером возвращается домой. У него в руках связка из $n$ ключей, один из которых подходит к дверям квартиры, а остальные гаечные. По причинам, о которых можно только догадываться, Вася пробует ключи наугад так, что при каждой попытке любой ключ, включая нужный, выбирается с одинаковой вероятностью. Найдите математическое ожидание числа попыток, которые потребуются Васе для того, чтобы попасть домой. По средам Вася в состоянии исключить неподошедшие ключи из дальнейших проверок, а по пятницам не в состоянии. Если $A$ математическое ожидание Васиных попыток в среду, а $B$ математическое ожидание васиных попыток в пятницу, то в ответе введите выражение $A\cdot x+B\cdot y$.

<h4 align='center'>Решение:</h4>

Сначала найдем математическое ожидание для среды. Вероятность того, что квартира будет открыта первым ключом равна $\frac{1}{n}$. Вероятность того, что квартира будет открыта вторым ключом равна $\big(1-\frac{1}{n}\big)\frac{1}{n-1}=\frac{1}{n}$ (для этого нужно чтобы сначала не подошел первый ключ, вероятность этого $1-\frac{1}{n}$, а затем подошел второй ключ, вероятность этого $\frac{1}{n-1}$). Вероятность того, что квартира будет открыта третьим ключом равна $\big(1-\frac{2}{n}\big)\frac{1}{n-2}=\frac{1}{n}$ (нужно чтобы не подошли ни первый, ни второй ключи, но подошел третий). Аналогично устанавливается, что вероятность того, что квартира будет открыта $k$-м ключом равна $\frac{1}{n}$. Тогда математическое ожидание количества проверок равно

$\large 1\cdot\frac{1}{n}+2\cdot\frac{1}{n}+3\cdot\frac{1}{n}+4\cdot\frac{1}{n}+\ldots+n\cdot\frac{1}{n} =\frac{n(n+1)}{2}\cdot\frac{1}{n} = \frac{n+1}{2}$

Перейдем теперь к нахождению математического ожидания для пятницы. Вероятность того, что подошел $k$-й ключ равна $\frac{1}{n}$. Рассмотрим вспомогательную случайную величину $\xi_k$, которая обращается в единицу, если при $k$-й проверке ключ не подошел, и обращается в ноль, если подошел. Тогда количество попыток до успешного попадания в квартиру равно 

$\large \xi=1+\xi_1+\xi_1\xi_2+\xi_1\xi_2\xi_3+\xi_1\xi_2\xi_3\xi_4+\dots$

Поскольку все случайные величины $\xi_k$ независимы и имеют одинаковое математическое ожидание $\mathbb{E}\xi_k=0\cdot\frac{1}{n}+1\cdot\big(1-\frac{1}{n}\big)=1-\frac{1}{n}$,

$\large \mathbb{E}\xi= 1+\mathbb{E}\xi_1+\mathbb{E}\xi_1\mathbb{E}\xi_2+\mathbb{E}\xi_1\mathbb{E}\xi_2\mathbb{E}\xi_3+\dots= 1+\Big(1-\frac{1}{n}\Big)+\Big(1-\frac{1}{n}\Big)^2+\Big(1-\frac{1}{n}\Big)^3+\dots = \frac{1}{1-\big(1-\frac{1}{n}\big)} = n.$


Альтернативный способ вычисления математического ожидания такой. Можно заметить, что вероятность того, что квартира будет открыта на $k$-й проверке равна $\frac{1}{n}\big(1-\frac{1}{n}\big)^{k-1}$ (нужно чтобы первые $k-1$ раз успех не был достигнут, а на $k$-м шаге был успех). Положим для краткости $p=1-\frac{1}{n}$. Тогда математическое ожидание равно

$\large \mathbb{E} = \sum\limits_{k=1}^\infty k(1-p)p^{k-1} = (1-p)\sum\limits_{k=1}^\infty kp^{k-1}$

Основная сложность на этом пути — вычисление этой суммы. Один из способов такой: заметить, что (пусть $0\lt x<1$)

$\large \sum\limits_{k=1}^m kx^{k-1} = \sum\limits_{k=1}^m \big(x^k\big)' = \bigg(\sum\limits_{k=1}^m x^k\bigg)' = \bigg(\frac{1-x^{m+1}}{1-x}\bigg)' = \frac{1-x^{m+1}}{(1-x)^2}-\frac{(m+1)x^m}{1-x}$

а при переходе к пределу при mm стремящемся к бесконечности получим
$\large \sum\limits_{k=1}^\infty kx^{k-1} = \frac{1}{(1-x)^2}$. 

Следовательно, 

$\large \mathbb{E} = (1-p)\frac{1}{(1-p)^2}=\frac{1}{1-p}=n$

<h2 style="color:black" align="center">2.3 Дисперсия</h2>

Другой важной характеристикой случайной величины является дисперсия. Дисперсия характеризует то, насколько отклоняется случайная величина от математического ожидания. И определяется она следующим образом:

$\large \mathbb{D} \xi = \mathbb{E}(\xi - \mathbb{E}\xi)^2$

Иногда удобнее рассматривать не саму диспресию, а корнь из дисперсии:

$\large \sigma = \sqrt{\mathbb{D}\xi}$ - стандартное отклонение

### Свойства Дисперсии

**Свойство 1.** $\large \mathbb{D}\xi \ge 0$ по определению видно, что дисперсия не отрицательна, так как это математическое ожидание от неотрицательной величины.


**Свойство 2.** $\large \mathbb{D}\xi = \mathbb{E}(\xi^2) - (\mathbb{E}\xi)^2$ более удобно считать дисперсию по данной формуле, а не по определению.

Обозначим математическое ожидание следующим образом $a = \mathbb{E}\xi$, тогда по определению:

$\large \mathbb{D}\xi = \mathbb{E}(\xi - a)^2 = $

Расскроем скобки:

$\large = \mathbb{E}(\xi^2 - 2\xi a + a^2) =$

Воспользуемся линейностью математического ожидания, расскрыв все в три слогаемых и вынсети константы за знак математического ожидания:

$\large = \mathbb{E}(\xi)^2 - 2a \mathbb{E}\xi + \mathbb{E}(a^2) =$

Математическое ожидание от $\mathbb{E}(a^2) = a^2$, так как среднее значение для постоянной величины, равно самой величине. Величина $\mathbb{E}\xi = a$, тогда:

$\large = \mathbb{E}(\xi^2) - 2a^2 + a^2 = \mathbb{E}\xi^2 - (\mathbb{E}\xi)^2$


**Свойство 3.** $\large \mathbb{D}(\xi + a) = \mathbb{D}\xi$ дисперсия не изменится, если сдвинуть случайную величину на константу.

Воспользуемся определением, посмотрим чему равно математическое ожидание для $\xi + a$. По линейности получаем:

$\large \mathbb{E}(\xi + a) = \mathbb{E}\xi + a \Rightarrow$

Тогда, выржаение $\xi - \mathbb{E}\xi$, которое используется в определении дисперсии для $\xi + a$ принимает вид:

$\large \Rightarrow (\xi + a) - \mathbb{E}(\xi + a) = \xi - \mathbb{E}\xi$ - поскольку $a$ выносится из под математического ожидания, получаем $\xi - \mathbb{E}\xi$.

Следовательно, выражение $\xi - \mathbb{E}\xi$ в определении при сдвиге $\xi$ на константу не изменяется, тем самым и дисперсия тоже не изменяется.


**Свойство 4.** $\large \mathbb{D}(c\xi) = c^2 \mathbb{D}\xi$ поймем как выносятся множетели из-под дисперсии. Воспользуемся формулой из 2-го свойсва:

$\large \mathbb{D}(c\xi) = \mathbb{E}(c\xi)^2 - (\mathbb{E}(c\xi))^2 =$

В выражении $\mathbb{E}(c\xi)^2$ можно вынести константку возведенную в квадрат из под знака математического ожидания. А во втором выражении тоже вынесем константу из под знака математического ожидания:

$\large = с^2\mathbb{E}\xi^2 - (c\mathbb{E}\xi)^2 = $

Остается вынести константу за оба выражения:

$\large = c^2(\mathbb{E}\xi^2 - (\mathbb{E}\xi)^2)$ - а под скобками и есть дисперсия случайно величины.

### Дисперсия суммы случайных величин

Поймем как устроена дисперсия для суммы случайных величин, если мы не предпологаем никаких дополнительных свойств у этих случайных величин. Распишем дисперсию по второму свойству:

$\large \mathbb{D}(\xi + \eta) = \mathbb{E}(\xi + \eta)^2 - (\mathbb{E}(\xi + \eta))^2 =$

В первом выражении раскроем квадрат, как квадрат суммы. А во втором выражении сначало по линейности распишем математическое ожидание:

$\large = \mathbb{E}(\xi^2 +2 \xi\eta + \eta^2) - (\mathbb{E}\xi + \mathbb{E}\eta)^2 =$

После, раскроем второе выражение как квадрат суммы. В первом выражении воспользуемся линейностью математического ожидания:

$\large = \mathbb{E}(\xi^2) + 2\mathbb{E}(\xi\eta) + \mathbb{E}(\eta)^2 - (\mathbb{E}\xi)^2 - 2\mathbb{E}\xi\mathbb{E}\eta - (\mathbb{E}\eta)^2$

Сгруппируем некоторые слогаемые $\mathbb{E}(\xi^2) - (\mathbb{E}\xi)^2 = \mathbb{D}\xi$ и $\mathbb{E}(\eta)^2 - (\mathbb{E}\eta)^2 = \mathbb{D} \eta$, тогда, получаем:

$\large = \mathbb{D} \xi + \mathbb{D}\eta 
+ 2\underbrace{
        (\mathbb{E}(\xi\eta) - \mathbb{E}\xi\mathbb{E}\eta)
        }_{
            \text{cov}(\xi, \eta) 
            = 
            \mathbb{E}((\xi - \mathbb{E}\xi)(\eta - \mathbb{E}\eta))
           }$

Выражение $(\mathbb{E}(\xi\eta) - \mathbb{E}\xi\mathbb{E}\eta)$ в общей ситуации, если мы ничего не предполагаем про $\xi$ и $\eta$, тогда это выражение называется **ковариацией случайных величин $\eta$ и $\xi$**. Не сложно понять аналогично тому, как доказывалось второе свойство, что ковариация может быть посчитана и по данной формуле тоже:

$\large \text{cov}(\xi, \eta) = \mathbb{E}((\xi - \mathbb{E}\xi)(\eta - \mathbb{E}\eta))$

Ковариация равна дисперсии, если $\eta = \xi$, тогда мы просто получим определение дисперсии случайной величины.

Иногда, вместо ковариации используют коэффициент корреляции - это следующая величина: 

Если $\mathbb{D}\xi > 0$ и $\mathbb{D}\eta > 0$, то 

$\large \varrho(\xi, \eta) = \frac{\text{cov}(\xi,\eta)}{\sqrt{\mathbb{D}\xi} \sqrt{\mathbb{D}\eta}}$ - **коэффициент корреляции.**

Можно првоерить, что данное выражение по модулю не превосходит $1$ и засчет этого, оно является более удобным.

### Запишем полученные формулы

$\large \mathbb{D}(\xi + \eta) = \mathbb{D}\xi + \mathbb{D}\eta + 2 \text{cov}(\xi, \eta)$, где ковариация определяется как:

$\large \text{cov}(\xi, \eta) = \mathbb{E}(\xi\eta) - \mathbb{E}\xi \mathbb{E}\eta$

Заметим, если случайные величины $\xi$ и $\eta$ независимы, то мы знаем, что математическое ожидание от произведения, это есть произведение математических ожиднаий:

$\large \mathbb{E}(\xi \eta) = \mathbb{E}\xi \mathbb{E}\eta \Rightarrow \text{cov}(\xi, \eta) = 0$

Тогда, разность в определении ковариации раввна нулю. Значит, в сумме дисперсии случайных величин мы получим нулевую добавку, так как $2 \text{cov}(\xi, \eta) = 0$:

$\large \Rightarrow \mathbb{D}(\xi + \eta) = \mathbb{D}\xi + \mathbb{D}\eta$

И мы получили формулу, в случае, если случайные величины независимы, тогда дисперсия суммы это есть сумма дисперсий.

### Свойства ковариации

**Свойство 1.** $\large \text{cov}(\xi, \eta) = \text{cov}(\eta, \xi)$ ковариация не зависит от порядка агрументов.

**Свойство 2.** $\large \text{cov}(\xi_1 + \xi_2, \eta) = \text{cov}(\xi_1, \eta) + \text{cov}(\xi_2, \eta)$ - ковариация линейна, ковариация суммы есть сумма ковариаций.

Раскроем по определению первую ковариацию:

$\large \mathbb{E}((\xi_1 + \xi_2)\eta) - \mathbb{E}(\xi_1 + \xi_2)\mathbb{E}\eta =$

по линейности распишем математическое ожидание:

$\large = \mathbb{E}(\xi_1 \eta) + \mathbb{E}(\xi_2\eta) - \mathbb{E}\xi_1 \mathbb{E}\eta - \mathbb{E}\xi_2 \mathbb{E}\eta$

Если сгруппировать слогаеммые, тогда данное выражение дает сумму ковариаций. Поскольку мы можем переставлять агрументы, тогда и по второму агрументу данное свойство справедливо.

**Свойство 3.** $\large \text{cov}(c\xi, \eta) = c \cdot \text{cov}(\xi, \eta)$ можно выностить константы из под ковариации. Таким образом получается, что ковариация линейна по каждому из агрументов.

**Свойство 4.** $\large \mathbb{D}(\xi_1 + \ldots + \xi_{n}) = \mathbb{D}\xi_1 + \ldots + \mathbb{D}\xi_n + 2 \sum\limits_{i < j} \text{cov}(\xi_i, \xi_j)$

Данное свойство проясняет зачем нужна ковариация. Ковариация возникает, если необходимо посчитать дисперсию для суммы большого количества случайных величин.

**Индукция:** Сделаем переход от $n \to n+1$. Сгруппируем $\xi_n$ и $\xi_{n+1}$, будем считать что это одно слогаемое, тогда по написанной выше формуле, все раскроется в следующую сумму дисперсий, плюс следующую сумму ковариаций:

$\large \mathbb{D}(\xi_1 + \ldots + \xi_{n+1}) 
= 
\mathbb{D}\xi_1 + \ldots + \mathbb{D}\xi_{n-1} 
+ 
\mathbb{D}(\xi_{n} + \xi_{n+1}) +$


$\large + 2 \sum\limits_{i < j < n} \text{cov}(\xi_i, \xi_j) 
+ 
2 \sum\limits_{i=1}^{n-1} \text{cov}(\xi_i, \xi_n + \xi_{n+1})$


Дисперсия для:

$\mathbb{D}(\xi_{n} + \xi_{n+1}) = \mathbb{D}\xi_n + \mathbb{D}\xi_{n+1} + 2 \text{cov}(\xi_n, \xi_{n+1})$.

Все ковариации раскрываются по линейности и получается:

$\sum\limits_{i=1}^{n-1} \text{cov}(\xi_i, \xi_n + \xi_{n+1}) = \sum\limits_{i=1}^{n-1} \text{cov}(\xi_i, \xi_n) + \text{cov}(\xi_i, \xi_{n+1})$ 

Если сгруппировать все, что было получено, тогда мы получим сумму всех дисперсий плюс все возможные ковариации между $\xi_i$ и $\xi_j$.

### Пример 1

Может показаться, что из обращения в ноль ковариации следует независимость случайных величин. Это кончено же не так. Из $\text{cov}(\xi, \eta) =0$ не следует независиомть $\xi$ и $\eta$.

Пусть вероятностное пространство состоит из 3-х точек $\omega = \{0, \frac{\pi}{2}, \pi\}$ и они равновероятны. Рассмотрим две случайные величины $\xi = \sin \omega$, $\eta = \cos \omega$. Посчитаем их математические ожидания:

Так как, все точки равновероятны, тогда математическое ожидание равно сренему арифметическому:

$\large \mathbb{E}\xi = \frac{\sin 0 + \sin \frac{\pi}{2} + \sin \pi}{3} = \frac{1}{3}$

$\large \mathbb{E}\eta = \frac{\cos 0 + \cos \frac{\pi}{2} + \cos \pi}{3} = 0$

Случайная величина $\xi \eta$, то она равна нулю, так как в каждой из точек либо косинус, либо синус равен нулю, следовательно и произведение равно нулю:

$\large \eta \xi = \sin \omega \cdot \cos \omega = 0$

Тогда:

$\large \text{cov}(\xi, \eta) = \mathbb{E}(\xi \eta) - \mathbb{E}\xi \mathbb{E}\eta = 0 - \frac{1}{3} \cdot 0 = 0$

Проверим формально, что нет незвисимости:

$\large P(\xi = 1, \eta = 1) = 0 \ne \frac{1}{9} = P(\xi=1)P(\eta=1)$

### Пример 2 - Испытания Бернулли

Рассмотрим $n$ испытаний Бернулли. Пусть "Орёл" выпадает с вероятностью $p$, а "Решка" с вероятностью $1 - p$.

Пусть $\xi$ - это случайная величина, которая равна количествую успехов в $n$ последовательных подбрасываниях монеты.

Обычно, для подсчета дисперсии, математического ожидания - удобно использовать более простые конструкции, расщепляя случайную величину в сумму более простых объектов. За случайную величину $\xi_k$ обозначим следующее:

* Она равна 1, если $k$-м подбрасывании выпал "Орёл"
* Она равна 0, если $k$-м подбрасывании выпал "Решка"

Тогда случайная величина может быть представлена как сумма случайных величин:

$\large \xi = \xi_1 + \ldots + \xi_n$

Поскольку математическое ожидание линейно, тогда это будет сумма математических ожиданий. При этом, все слогаемые одинаковы, поэтому:

$\large \mathbb{E}\xi = \mathbb{E}\xi_1 + \ldots + \mathbb{E}\xi_n = n \mathbb{E}\xi_1$

При этом, эти случайные величины независимы, тогда дисперсия для суммы случайных величин равна сумме дисперсий. Так как все дисперсии одинаковы с точки зрения распределения, тогда все дисперсии равны:

$\large \mathbb{D}\xi = \mathbb{D}\xi_1 + \ldots + \mathbb{D}\xi_n = n \mathbb{D}\xi_1$

**Посчитаем математическое ожидание:**
 
$\large \mathbb{E}\xi_1 = 1 \cdot p + 0 \cdot (1-p) = p$

**Посчитаем дисперсию**, но для этого необходимо посчитать математическое ожидание квадрата случайной величины. Поскольку это случайная величина принимает значение 0 и 1, тогда квадрат на неё не повлияет:

$\large \mathbb{E}\xi_1^2 = \mathbb{E}\xi_1 = p$, тогда

$\large \mathbb{D}\xi_1 = \mathbb{E}\xi_1^2 - (\mathbb{E}\xi_1)^2 = p - p^2 = p(1 - p)$

**Следовательно, в ответе получаем, что:**

$\large \mathbb{E}\xi = \mathbb{E}\xi_1 + \ldots + \mathbb{E}\xi_n = n \mathbb{E}\xi_1 = np$

$\large \mathbb{D}\xi = \mathbb{D}\xi_1 + \ldots + \mathbb{D}\xi_n = n \mathbb{D}\xi_1 = np(1-p)$

<h3 style="color:#2E8B57">Задача № 1</h3>

Отметьте верные утверждения.

<h4 align='center'>Решение:</h4>

1. Математическое ожидание линейно, т.е. для постоянных $a$ и $b$ верно соотношение $\mathbb{E}(a\xi+b\eta)=a\,\mathbb{E}\xi+b\,\mathbb{E}\eta$.

Верно.

2. Для независимых случайных величин $\xi$ и $\eta$ дисперсия линейна, т.е. для постоянных $a$ и $b$ верно соотношение $\mathbb{D}(a\xi+b\eta)=a\,\mathbb{D}\xi+b\,\mathbb{D}\eta$.

Неверно, $\mathbb{D}(a\xi+b\eta)=a^2\,\mathbb{D}\xi+b^2\,\mathbb{D}\eta$

3. Математическое ожидание всегда неотрицательно.

Неверно, например, $\mathbb{E}(-1)=-1$.

4. Дисперсия всегда неотрицательна.

Верно, обсуждалось в видео про дисперсию.

5. Если $\mathbb{D}\xi=0$, то $\xi$ с единичной вероятностью постоянная величина.

Верно. По определению дисперсии это означает, что $\mathbb{E}(\xi-\mathbb{E}\xi)^2=0$, т.е. математическое ожидание неотрицательной случайной величины равно нулю, а это значит, что сама случайна величина $\xi-\mathbb{E}\xi$ равна нулю с вероятностью $1$.

Неверно, $\mathbb{E}(\xi+c)=\mathbb{E}\xi+c$.

7. Если к случайной величине прибавить постоянную, то дисперсия не изменится.

Верно, обсуждалось в видео про дисперсию.

8. $\mathbb{E}\xi=\mathbb{E}(-\xi)$.

Неверно. Математическое ожидание линейно, поэтому $\mathbb{E}\xi=-\mathbb{E}(-\xi)$.

9. $\mathbb{D}\xi=\mathbb{D}(-\xi)$.

Верно, $\mathbb{D}(-\xi)=(-1)^2\,\mathbb{D}\xi=\mathbb{D}\xi$.

Неверно, для этого равенства достаточно чтобы $\mathop{\mathrm{cov}}(\xi,\eta)=0$, а это может выполняться и для зависимых случайных величин. Пример приводился в видео, предшествующем заданию.

11. Если случайные величины $\xi$ и $\eta$ независимы, то $\mathbb{E}(\xi\eta)=\mathbb{E}\xi\,\mathbb{E}\eta$.

Верно, обсуждалось в видео про свойства математических ожиданий.

Неверно, например, если $P(\xi=0)=P(\xi=1)=\frac{1}{2}$, то $\mathbb{E}\xi^2=\mathbb{E}\xi=\frac{1}{2}\neq\frac{1}{4}=(\mathbb{E}\xi)^2$.

13. Для любых случайных величин $\xi$ и $\eta$ имеет место равенство $\mathbb{E}(\xi-\eta)=\mathbb{E}\xi-\mathbb{E}\eta$.

Верно, математическое ожидание линейно.

14. Если случайные величины $\xi$ и $\eta$ независимы, то $\mathbb{D}(\xi-\eta)=\mathbb{D}\xi-\mathbb{D}\eta$.

Неверно, $\mathbb{D}(\xi-\eta) 
= \mathbb{D}(\xi) + \mathbb{D}(-\eta) 
= \mathbb{D}(\xi) + (-1)^2 \mathbb{D}(\eta)
= \mathbb{D}\xi+\mathbb{D}\eta$.

<h3 style="color:#2E8B57">Задача № 3</h3>

Случайные величины $\xi$ и $\eta$ независимы и имеют одинаковое распределение с математическим ожиданием $e$ и дисперсией $d$. Найдите коэффициент корреляции cлучайных величин $a\xi+b\eta$ и $a\xi−b\eta$ ($a$ и $b$ — вещественные числа).

<h4 align='center'>Решение:</h4>

По линейности математического ожидания имеем:

$\large \mathbb{E}(a\xi\pm b\eta) = a\,\mathbb{E}\xi\pm b\,\mathbb{E}\eta = ae\pm be = (a\pm b)e$.


Найдем теперь математическое ожидание произведения:

$\large \begin{aligned} \mathbb{E}\big((a\xi+b\eta)(a\xi-b\eta)\big) &= \mathbb{E}\big(a^2\xi^2-b^2\eta^2\big) = a^2\mathbb{E}\xi^2-b^2\mathbb{E}\eta^2 = a^2\big(\mathbb{D}\xi+(\mathbb{E}\xi)^2\big)-b^2\big(\mathbb{D}\eta+(\mathbb{E}\eta)^2\big) = \\ &= a^2(d+e^2)-b^2(d+e^2) = (a^2-b^2)(d+e^2) . \end{aligned}$
 
Следовательно,

$\large \begin{aligned} \mathop{\mathrm{cov}}(a\xi+b\eta,a\xi-b\eta) &= \mathbb{E}\big((a\xi+b\eta)(a\xi-b\eta)\big)-\mathbb{E}(a\xi+b\eta)\mathbb{E}(a\xi-b\eta) = \\ &= (a^2-b^2)(d+e^2)-(a+b)e\cdot(a-b)e = (a^2-b^2)d . \end{aligned}$

Осталось найти дисперсии

$\large \mathbb{D}(a\xi\pm b\eta) = a^2\,\mathbb{D}\xi+b^2\,\mathbb{D}\eta = (a^2+b^2)d$.

Таким образом,

$\large \rho(a\xi+b\eta,a\xi-b\eta) = \frac{\mathop{\mathrm{cov}}(a\xi+b\eta,a\xi-b\eta)}{\sqrt{\mathbb{D}(a\xi+b\eta)}\sqrt{\mathbb{D}(a\xi-b\eta)}} = \frac{(a^2-b^2)d}{(a^2+b^2)d} = \frac{a^2-b^2}{a^2+b^2}$.
