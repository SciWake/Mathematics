### Конечно число исходов

Пространство элементарных исходов/событий $\Omega = \{\omega_1, \ldots, \omega_N\}$, где $\omega$ - элементарное событие или элементарный исход.

**Пример 1.** Однократное подбрасывание монеты

$\Omega = \{\text{Орёл, Решка}\}$, $|\Omega| = 2$

**Пример 2.** Однократное подбрасывание игральной кости

$\Omega = \{1, 2, 3, 4, 5, 6\}$, $|\Omega| = 6$

**Пример 3.** $n$-кратное подбрасывание монеты

$\Omega = \{\omega: \omega = (a_1, \ldots, a_n), \text{где } a_i = \text{О или Р}\}$, $|\Omega| = 2^n$

**Пример 4.** Есть 4 красных шара и 6 зеленых. Знаумируем наши шары, где от 1 до 4 - это красные шары, а числа от 5 до 10 это зеленые шары. Выберем 3 шара

$\Omega = \{\omega: \omega = (a_1, a_2, a_3), \text{где } a_i  \text{- различные числа от 1 до 10}\}$, $|\Omega| = C_{10}^3$

**Пример 5** Выбор с возвращением. В урне $m$ шаров, выбирают $n$

Если важен порядок, тогда пространством элементарных исходов будет все возможные последовательности из $n$ чисел:

$\Omega = \{\omega: \omega = (a_1, \ldots, a_n), a_i \in \{1,\ldots, m \}\}$, $|\Omega| = \bar{A}_m^n$

Если не важен порядок, тогда пространством элементарных исходов будет все возможные неупорядоченные наборы из $n$ чисел:

$\Omega = \{\omega: \omega = [a_1, \ldots, a_n], a_i \in \{1,\ldots, m \}\}$, $|\Omega| = \bar{С}_m^n$

### Событие

Если имеется пространство элементарынх исходов $\Omega = \{\omega_1, \ldots, \omega_N\}$, тогда событием называется произвольное подмножество пространства элементарынх исходов $A \subset \Omega$.

**Пример 1.** 3-кратное подбрасывание монеты

$A = \{\text{ООР}, \text{ОРО}, \text{РОО} \}$ - выпало два орла 

$B = \{\text{РРР}\}$ - ни разу не выпал орел

$A \cup B$ - Выпало четное число орлов, значит добавится событие, когда орлы вообще не выподают.

**Множества** для событиямй справедливы операции множествами, так как событие является множеством.

$A \cup B = \{\omega \in \Omega: \omega \in A \text{ или } \omega \in B\}$

$A \cap B = \{\omega \in \Omega: \omega \in A \text{ и } \omega \in B\}$

$\bar{A} = \{\omega \in \Omega: \omega \notin A\}$


**Примеры событий**

Рассмотрим "Пример 5. Выбор с возвращением. В урне $m$ шаров, выбирают $n$":

Пример события для упорядоченного выбора: Достаем шары с номерами идущими в возрастающем порядке. Для неупорядоченного выбора, данное событие не будет иметь смысла.

Пример события для двух случаев, упорядоченного и неупорядоченного выбора шаров: Все время достаем разные шары, тогда если $n > m$, то такое событие будет невозможным. Изначально выбор с возращением, но так как мы достаем каждый раз разные шары, тогда $n > m$ не может быть.

<h3 style="color:#2E8B57">Задача № 1</h3>

Из какого количества элементарных событий состоит пространство элементарных событий для следующих испытаний:

* производится выстрел по мишени, представляющей собой $10$ концентрических кругов, занумерованных числами от $1$ до $10$;
* три раза подбрасывается игральная кость;
* наудачу извлекается одна кость из полной игры домино.

<h4 align='center'>Решение:</h4>

1) Можно попасть в любой из десяти занумерованных кругов или не попасть по мишени, то есть $11$ исходов.

2) При каждом из трех бросаний можно получить по шесть вариантов, то есть $6^3=216$ исходов.

3) Полная игра домино содержит $28$ различных костей, то есть 28 исходов.

<h3 style="color:#2E8B57">Задача № 2</h3>

Из какого количества элементарных событий состоит пространство элементарных событий для следующего испытания: производится выстрел по мишени, представляющей собой $10$ концентрических кругов, занумерованных числами от $1$ до $10$, а затем столько раз кидается игральная кость, сколько очков выбито на мишени.

<h4 align='center'>Решение:</h4>

Если при выстреле промахнулись, то кость не бросается. В таком случае исход единственный. Если при выстреле попали в $1$, то кость бросается один раз, то есть исходов $6$. Если при выстреле попали в $k$, то кость бросается $k$ раз, то есть $6^k$ исходов. Вычислим общее количество исходов как сумму геометрической прогрессии:

$$\sum\limits_{k=0}^{10}6^k=\frac{6^{11}-1}{6-1}=72559411.$$

<h2 style="color:black" align="center">1.3 Вероятностные пространства</h2>

### Определение веротяностного пространтсва

Пусть имеется пространство элементарных событий $\Omega = \{\omega_1, \ldots, \omega_N\}$, состоящее из $N$ исходов и каждому из возможных исходов припищем число от $0$ до $1$, обозначив через:

$$p_1, \ldots, p_N \ge 0$$ 

Добавим условие нормировки, указава, что сумма этих чисел должна равняться $1$:

$$p_1 + \ldots + p_N = 1$$

Тогда эти числа мы будем называть вероятностями элементарных событий, следовательно $p_i$ это вероятность события $\omega_i$.


**Вероятность события $A$**

Рассмотрим все возможные элементы $\omega_k$, которые содержатся в множестве $A$ и просуммируем соответствующие им $p_k$:

$\large P(A) = \sum\limits_{k: \omega_k \in A} p_k$ - вероятность события $A$

### Свойства вероятности

1. $P(\emptyset) = 0, \quad P(\Omega) = 1$ - в первом случае в сумме нет ни одного слогаемого, поэтому вероятность равна $0$. Во втором случае суммируются все $p$ и получем $1$.

2. $0 \le P(A) \le 1$ - Аналогично получаем из суммы.

3. Если $A \cap B = \emptyset$, то $P(A \cup B) = P(A) + P(B)$. Этот пунк следует из 4-го свойства, так как если $A \cap B = \emptyset$, тогда мы вычитаем ноль и получаем 3-е свойство

4. $P(A \cup B) = P(A) + P(B) - P(A \cap B)$

**Проверим 4 свойство**, записав все вероятности с помощью определения:

$\large \sum\limits_{k: \omega_k \in A \cup B} p_k = \sum\limits_{k: \omega_k \in A} p_k + \sum\limits_{k: \omega_k \in B} p_k - \sum\limits_{k: \omega_k \in A \cap B} p_k$

Если $\omega_k$ не лежит ни в множестве $A$ ни $B$, тогда соответсвующий ему $p_k$ не лежит ни в одной из сумм, следовательно для этих $p_k$ равенство выполняется, так как слева и справа ничего нет.

Если $\omega_k$ лежит только в множестве $A$, тогда $p_k$ есть в сумме слева и в первой сумме справа, но нет во всех остальных суммах, поэтму для таких $p_k$ выполняется равенство между суммами.

Аналогично, если $\omega_k$ лежит только в множестве $B$.

Если $\omega_k$ лежит в множестве $A$ и $B$, тогда соответсвующий ему $p_k$ учтено в сумме слева, в сумме для $\omega_k \in A$, $\omega_k \in B$ и учтено в сумме $\omega_k \in A \cap B$ с отрицательным знаком, следовательно, $p_k$ в правой часте учтено один раз как и в левой части, значит равенство выполняется.

И теперь, формула доказана.

5. $P(A \cup B) \le P(A) + P(B)$ Это свойство следует из 4-го, так как мы вчитаем не отрицательное слогаемое $P(A \cap B)$

6. $P(\bar{A}) = 1 - P(A)$ Это свойство является следствием из 3-го свойства, так как если в 3-ем свойстве вместо $B$ подставить событие $\bar{A}$, тогда мы получим пустое пересечение $A \cap \bar{A} = \emptyset$, тогда $P(A \cup \bar{A}) = P(A) + P(\bar{A})$




7. $P\big(\bigcup\limits_{k=1}^n A_k\big) \le \sum\limits_{k=1}^n P(A_k)$

**Доказательство по индукции** База $n = 2$, тогда $P(A_1 \cup A_2) \le P(A_1) + P(A_2)$ - это неравенство было 5-свойством, поэтому оно уже проверено.

Остается доказать переход $n \to n + 1$. Заведем вспомогательное множество $B := \bigcup\limits_{k=1}^n A_k$ и тогда:

$P\big(\bigcup\limits_{k=1}^{n+1} A_k \big) = P(B \cup A_{n + 1}) \le$

По 5 свойству или по базе, получаем, что:

$\le P(B) + P(A_{n + 1}) \le$

Но $B$ является объединением множеств, тогда по индукционному предположению вероятность события $P\big(\bigcup\limits_{k=1}^n A_k\big)$ не превосходит суммы вероятностей:

$P\big(\bigcup\limits_{k=1}^n A_k\big) \le \sum\limits_{k=1}^n P(A_k)$, значит, вместо вероятности $B$ можем подставить следуюую сумму и от этого выражение станет только больше:

$\le \sum\limits_{k=1}^n P(A_k) + P(A_{n+1})$

Итого, мы получаем, что $P\big(\bigcup\limits_{k=1}^{n+1} A_k \big) \le \sum\limits_{k=1}^n P(A_k) + P(A_{n+1})$, что вместе дает сумму веротяностей событий $a_1, \ldots, a_{n+1}$

8. $P\big(\bigcup\limits_{k=1}^n A_k\big) = \sum\limits_{k=1}^n P(A_k) - \sum\limits_{j < k} P(A_j \cap A_k) + \sum\limits_{i < j < k} P(A_i \cap A_j \cap A_k) - \ldots + (-1)^{n-1} P(A_1 \cap \ldots \cap A_n)$

**Доказательство по индукции** База $n = 2$, тогда $P(A_1 \cup A_2) = P(A_1) + P(A_2) - P(A_1 \cap A_2)$, а это соответствует свойствую $4$, которое доказано.

Остается доказать переход $n \to n + 1$. Заведем вспомогательное множество $B := \bigcup\limits_{k=1}^n A_k$ и тогда:

$P\big(\bigcup\limits_{k=1}^{n+1} A_k \big) = P(B \cup A_{n + 1}) =$

По 4 свойству или по базе, получаем, что:

$= P(B) + P(A_{n + 1}) - P(B \cap A_{n + 1}) =$

---
А что значит событие $P(B \cap A_{n + 1})$? Подставим $B$ и получим объединение всевозможных пересечений множества $A_k$ с $A_{n+1}$

$P(B \cap A_{n + 1}) = A_{n+1} \bigcup\limits_{k=1}^n A_k = \bigcup\limits_{k=1}^n (A_k \cap A_{n + 1})$

Следовательно $P(B \cap A_{n + 1})$ - вероятность объеденения некоторых $n$ множеств, а для такой вероятности мы уже доказали формулу включений-исключений.

---

Расписываем $P(B)$ по формуле включений-исключений и для $P(B \cap A_{n + 1})$ тоже, только заместо $P(B \cap A_{n + 1})$ возьмем $\bigcup\limits_{k=1}^n (A_k \cap A_{n + 1})$:


$= P(A_{n + 1}) 
+ \big(
    \sum\limits_{k=1}^n P(A_k) 
        - \sum\limits_{j < k \le n} P(A_j \cap A_k) + \ldots 
  \big) 
- \big(
    \sum\limits_{k=1}^n P(A_k \cap A_{n+1}) 
        - \sum\limits_{j < k \le n} P(A_j \cap A_k \cap A_{n + 1}) + \ldots 
  \big)$
  
Имеется вероятность $P(A_{n + 1})$ множества $A_{n + 1}$. После идет сумма вероятностей остальных $\sum\limits_{k=1}^n P(A_k)$ множеств $a$. Именно то, что и нужно в формуле включений-исключений - это сумма вероятностей всех одиночных событий. Что вычитаем? Вычитаем вероятность всех попарных пересечений где $i, j$ меньше либо равны $n$.

Что представляет последнее выражение? Вычитается вероятность попарных пересечений $\sum\limits_{k=1}^n P(A_k \cap A_{n+1})$ и это выражение совместно с выражением $\sum\limits_{j < k \le n} P(A_j \cap A_k)$ является суммой всевозможных вероятностей для попарных пересечений множеств с индексами от $1$ до $n+1$. 

После выражения $\sum\limits_{j < k \le n} P(A_j \cap A_k)$ в первой скобе должны прибавить вероятность всех возможных пересечений по трое, в случае если индекс от $1$ до $n$. Так как перед второй скобкой стоит минус и перед выражением $\sum\limits_{j < k \le n} P(A_j \cap A_k \cap A_{n + 1})$ стоит минус, тогда мы должны прибавить пересечения по три, множеств. Следовательно, мы рассматриваем все возможные пересечений из первой скобки, когда ни один из индексов не равен $n + 1$, а во второй скобке когда один из индексов равняется $n + 1$. 

Значит мы рассматриваем все возможные пересечения по 3 и считаем их вероятности и складываем и так далее разбираемся с этими суммами и убеждаемся в том, что получается искомая нами формула.

### Классическое определение вероятности

Пусть имеется пространство элементарных событий $\Omega = \{\omega_1, \ldots, \omega_N\}$, состоящее из $N$ исходов и все элементарные исходы являются равновероятными, тогда:

$p_1 = \ldots = p_N = \frac{1}{N}$, тогда вероятность произвольного события $A$:

$\large P(A) = \frac{\#A}{\#\Omega}$

Верно, так как благоприятных исходов для события $B$ не меньше, чем для события $A$.


**2.** $P(A\cap B)$ всегда меньше, чем $P(A)$

Неверно, так как событие $A\cap B$ может совпадать с событием $A$, тогда их вероятности будут равны.


**3.** $P(A\cap B)$ может быть больше, чем $P(A)$

Неверно, так как благоприятных исходов для события $A\cap B$ не больше, чем для события $A$.


**4.** $P(A\cup B)=P(A)+P(B)-P(A\cap B)$
    
Верно. Это 4-е свойство в видео.


**5.** Если $P(A)\gt1/2$ и $P(B)\gt1/2$, то $P(A\cup B)=1$
    
Неверно. Если события $A$ и $B$ совпадают, то $P(A\cup B)=P(A)$, что может быть меньше единицы.


**6.** $P(A\cap B)=P(A)P(B)$

Неверно для событий, которые не являются независимыми. Например, если $B=\bar{A}$ и $0\lt P(A) \lt 1$.


**7.** $P(A\cap B)\geqslant P(A)+P(B)-1$
    
Верно. Так как $P(A\cap B)\leqslant1$, то по 4-му свойству $P(A\cap B)=P(A)+P(B)-P(A\cup B)\geqslant P(A)+P(B)-1$


**8.** $P(A\cap \bar{B})=P(A\cup B)-P(B)$
    
Верно, поскольку $(A\cap \bar{B})\cup B = A\cup B$ и события $A\cap \bar{B}$ и $B$ несовместны.


**9.** Событие $A\cup B$ означает, что произошло ровно одно из событий $A$ и $B$
    
Неверно. Событие $A\cup B$ означает, что произошло хотя бы одно из событий $A$ и $B$, то есть содержит и случай, когда произошли оба события $A$ и $B$.


**10.** Событие $A\cap B$ означает, что произошли оба события $A$ и $B$
    
Верно по определению пересечения множеств.


**11.** $P(A)\lt 1$
    
Неверно, так как бывает, что $P(A)=1$.


**12.** Если $P(\bar{A}\cup\bar{B})=1$, то $P(A\cup B)=P(A)+P(B)$
    
Верно. Так как $\bar{A}\cup\bar{B}=\Omega\setminus(A\cap B)$, то $P(A\cap B)=1-P(\bar{A}\cup\bar{B})=0$, тогда событие $A\cup B$ означает, что произошло ровно одно событие $A$ или $B$.

<h2 style="color:black" align="center">1.5 Условная вероятность</h2>

### Определение условной вероятности

$\large P(B | A) = \frac{\# A \cap B}{\# A} = \frac{\# A \cap B / \# \Omega}{\# A / \# \Omega} = \frac{P(A \cap B)}{P(A)}$

**Вытекающие свойства**

1) $P(A|A) = 1$

2) Если $B \supset A$, то $P(B|A) = 1$, так как $A \cap B = A$

3) $P(\emptyset | A) = 0$

4) Если $B_1 \cap B_2 = \emptyset$, то $P(B_1 \cup B_2|A) = P(B_1|A) + P(B_2|A)$

5) $P(B|A) + P(\bar{B}|A) = 1$ 

при этом $P(B|A) + P(B|\bar{A}) \ne 1$

<h3 style="color:#2E8B57">Задача № 1</h3>

В урне $11$ красных, $10$ синих и $9$ зеленых шаров. Из нее последовательно вынимают три шара. Найдите вероятность того, что первый шар окажется красным, второй — синим, а третий — зеленым. В ответе приведите обыкновенную дробь.

<h4 align='center'>Решение:</h4>

Пусть событие $A$ означает, что первым был вытащен красный шар, событие $B$ означает, что вторым был вытащен синий шар, а событие $C$ означает, что третий шар — зеленый. Нам нужно найти вероятность $P(A\cap B\cap C)$. По определению условной вероятности:

$P(A|B) = \frac{P(A \cap B)}{P(B)} \Rightarrow P(A \cap B) = P(A|B)P(B)$

$P(A\cap B\cap C) = P(C|A\cap B)P(A\cap B) = P(C|A\cap B)P(B|A)P(A)$

Найдем указанные вероятности. Поскольку всего шаров $30$, а красных среди них — $11$, $P(A)=\frac{11}{30}$

Вероятность $P(B|A)$ означает, что уже наступило событие $A$ и нужно найти вероятность наступления события $B$ в этом случае. Т.е. первым был вытащен красный шар, после этого осталось $29$ шаров, из которых $10$ синих, стало быть, $P(B|A)=\frac{10}{29}$

Наконец, вероятность $P(C|A\cap B)$ означает, что уже наступили события $A$ и $B$ и нужно найти вероятность наступления события $C$ в этом случае. Т.е. первым был вытащен красный шар, а вторым — синий, после этого осталось $28$ шаров, из которых $9$ зеленых, следовательно, $P(C|A\cap B)=\frac{9}{28}$

Таким образом,

$P(A\cap B\cap C) = \frac{9}{28}\cdot\frac{10}{29}\cdot\frac{11}{30} = \frac{33}{812}$

<h2 style="color:black" align="center">1.6 Теорема Байеса</h2>

### Формула полной вероятности

Пусть пространство элементарных событий 

$\Omega = A_1 \cup A_2 \cup \ldots \cup A_n$, мы предпологаем, что события не совместны $A_j \cap A_k = \emptyset$ при $j \ne k$. И вероятность любого события не нулевая $P(A_k) > 0$.

$\Omega = A_1 \cup A_2 \cup \ldots \cup A_n, \quad A_j \cap A_k = \emptyset \text{ и } P(A_k) > 0$

Рассмотрим произвольное событие $B$, тогда $B$ можно представить как пересечение с объединением:

$B = B \cap \bigcup\limits_{k=1}^n A_k = \bigcup\limits_{k=1}^n B \cap A_k$ и множества $B \cap A_j$ и $B \cap A_k$ не пересекаются

Тогда вероятность $P(B)$ это сумма вероятностей $P(B \cap A_k)$:

$P(B) = \sum\limits_{k=1}^n P(B \cap A_k) = \sum\limits_{k=1}^n P(B | A_k) P(A_k)$ - **Формула полной вероятности**

**В частности**, если $0 < P(A) < 1$. Так как при $P(A)=0$ не определена условная вероятность $P(B|A)$, а при $P(\bar{A})=0$ (т.е. при $P(A)=1$) не определена условная вероятность $P(B|\bar{A})$. 

В качестве события $A_1$ возьмем $A$, а заместо $A_2$ возьмем событие $\bar{A}$, то есть $n=2$. Тогда вероятность $P(B)$ может быть посчитана по следующей формуле:

$P(B) = P(B|A) P(A) + P(B| \bar{A}) P(\bar{A})$

<h3 style="color:#2E8B57">Задача № 1</h3>

Из полного набора костей домино взята одна кость. Найдите вероятность того, что наудачу взятую вторую кость можно приставить к первой по правилам домино. В качестве ответа приведите обыкновенную дробь.

<h4 align='center'>Решение:</h4>

Пусть событие $A$ означает, что первая кость является дублем, а событие $B$ означает, что вторую кость можно приставить к первой. По формуле полной вероятности $P(B)=P(B|A)P(A)+P(B|\bar{A})P(\bar{A})$. Поскольку в домино $7$ дублей из $28$ костей, $P(A)=\frac{7}{28}=\frac{1}{4}$ и $P(\bar{A})=1-P(A)=\frac{3}{4}$. 

Найдем теперь вероятность $P(B|A)$. К дублю можно приставить только такую кость, на которой есть столько же очков, сколько на дубле. Их ровно $6$ штук, поэтому$ P(B|A)=\frac{6}{27}=\frac{2}{9}$. 

Далее найдем вероятность $P(B|\bar{A})$. К кости $m:n$ (при $m\neq n$) можно приставить те кости, на которых есть $m$ или $n$ очков. Таких костей $12$ штук, поэтому $P(B|\bar{A})=\frac{12}{27}=\frac{4}{9}$. Следовательно,

$P(B)=P(B\mid A)P(A)+P(B|\bar{A})P(\bar{A}) = \frac{2}{9}\cdot\frac{1}{4}+\frac{4}{9}\cdot\frac{3}{4} = \frac{7}{18}$.

### Формула Байеса

Есть два события $A$ и $B$, где $P(A) > 0$ и $P(B) > 0$, то

$\large P(A|B) = \frac{P(B|A) P(A)}{P(B)}$

**Доказательство формулы:**

$\large P(A|B) = \frac{A \cap B}{P(B)} = \frac{\frac{P(A \cap B)}{P(A)} P(A)}{P(B)} = \frac{P(B|A) P(A)}{P(B)}$

### Теорема Байеса

Пусть вероятностное пространство разбивается в объединение событий $\Omega = A_1 \cup A_2 \cup \ldots \cup A_k$, все эти события не совместы $A_j \cap A_k = 0$ и $P(A_k) > 0$, и некоторое событие $B$ тоже $P(B) > 0$. Тогда

$\large P(A_k | B) = \frac{P(B|A_k)P(A_k)}{P(B|A_1) P(A_1) + \ldots + P(B|A_n)P(A_n)}$

**Доказательство формулы:**

$\large P(A_k|B) = \frac{P(B|A_k)P(A_k)}{P(B)}$, остается только предстваить $P(B)$ с помощью формулы полной вероятности и подставить:

$\large P(A_k | B) = \frac{P(B|A_k)P(A_k)}{P(B|A_1) P(A_1) + \ldots + P(B|A_n)P(A_n)}$

В урне находится $n$ шаров, некоторые из них белые. Событие $A_k$ при $k=0, 1, \dots, n$ состоит в том, что в урне ровно $k$ белых шаров. $\begin{aligned} P(A_k|B) &= \frac{P(B|A_k)P(A_k)}{P(B|A_0)P(A_0)+P(B|A_1)P(A_1)+\ldots+P(B|A_n)P(A_n)} = \\ &= \frac{P(B|A_k)}{P(B|A_0)+P(B|A_1)+\ldots+P(B|A_n)} = \frac{\frac{k}{n}}{\frac{1}{n}+\frac{2}{n}+\ldots+\frac{n}{n}} = \frac{k}{1+2+\ldots+n} = \frac{2k}{n(n+1)} . \end{aligned}$

<h4 align='center'>Решение:</h4>

Пусть событие $B$ означает, что шуруп дефективный, а событие $A_k$, что шуруп выточен в $k$-й день недели. По условию
$$P(B|A_1)=\frac{1}{2}, \quad P(B|A_2)=\frac{1}{5}, \quad P(B|A_3)=P(B|A_4)=P(B|A_5)=\frac{1}{10}$$.

Кроме того $P(A_2)=P(A_3)=P(A_4)=P(A_5)=2P(A_1)$, поскольку в понедельник он вытачивает в два раза меньше шурупов, чем в каждый из остальных рабочих дней. Тогда $P(A_1)=\frac{1}{9}$ и $P(A_2)=P(A_3)=P(A_4)=P(A_5)=\frac{2}{9}$ По теореме Байеса имеем
  
$\large P(A_1|B) =\frac{P(B|A_1)P(A_1)}{P(B|A_1)P(A_1)\!+\!P(B|A_2)P(A_2)\!+\!\ldots\!+\!P(B|A_5)P(A_5)} =\frac{\frac{1}{2}\cdot\frac{1}{9}}{\frac{1}{2}\cdot\frac{1}{9}+\frac{1}{5}\cdot\frac{2}{9}+ 3\cdot\frac{1}{10}\cdot\frac{2}{9}} = \frac{1}{3}$.

**Note:**

Почему $P(A_1) = \frac{1}{9}$ ?

Представим $P(A_1) = x$. Тогда $P(A_2) = P(A_3) = P(A_4) = P(A_5) = 2x$. Гипотезы разбивают элементарное пространство, образуя полную группу, следовательно сумма вероятностей всех гипотез равна единице. Получаем 

$x + 4 \cdot 2x = 1$.

$9x = 1$

$x = \frac{1}{9}$

**Note:**

$P(B|A_{m+1})P(A_{m+1})$. Почему $m+1$? Если же $k\gt m$, то в качестве второго числа подойдут числа $1, 2, \ldots, k-m$...

<h3 style="color:#2E8B57">Задача № 6</h3>

Дано натуральное число $n\lt 52$. Из тщательно перемешанной колоды в $52$ карты одновременно были взяты $n$ карт. На одну из этих $n$ карт посмотрели, она оказалась тузом. После этого она возвращается в набор взятых карт и эти $n$ карт перемешиваются. После этого из них выбирается одна карта и открывается. Найдите вероятность того, что открытая карта является тузом.

<h4 align='center'>Решение:</h4>

Пусть событие $A$ означает, что во второй раз вытащили туза, а событие $B$ означает, что во второй раз вытащили ту же карту, что и в первый. Тогда по формуле для полной вероятности $P(A)=P(A|B)P(B)+P(A|\bar{B})P(\bar{B})$. 

Найдем нужные четыре вероятности: $P(B)=\frac{1}{n}$, поскольку это вероятность вытащить фиксированную карту из $n$ карт; тогда $P(\bar{B})=1-P(B)=\frac{n-1}{n}$. Кроме того очевидно, что $P(A|B)=1$. 

Найдем $P(A|\bar{B})$. Здесь нам нужно найти вероятность вытащить туза при условии, что мы вытащили не ту же карту, что и в первый раз. Это означает, что туз должен был оказаться среди оставшейся $n-1$ карты. Эти карты образовались так: взяли $n-1$ случайную карту из $51$ карты колоды (событие $\bar{B}$ запрещает брать вытащенного в первый раз туза), а затем из них выбрали наугад карту. Но это все равно, что взять наугад одну карту из $51$. Вероятность при этом вытащить туза равна $\frac{3}{51}=\frac{1}{17}$. Следовательно, $P(A|\bar{B})=\frac{1}{17}$. Таким образом, 

$P(A)=P(A|B)P(B)+P(A|\bar{B})P(\bar{B}) = 1\cdot\frac{1}{n}+\frac{1}{17}\cdot\frac{n-1}{n} = \frac{16+n}{17n}$.

<h2 style="color:black" align="center">1.7 Независимые события</h2>

### Понятие независимости событий

$A, B$ - некоторые события. Событие $B$ не зависит от $A$, если $P(B|A) = P(B)$.

$\large P(B|A) = \frac{P(A \cap B)}{P(A)} \Rightarrow P(B) = \frac{P(A \cap B)}{P(A)} \Rightarrow P(A \cap B) = P(A)P(B)$

**Тогда**

События $A, B$ независимы, если 

$P(A \cap B) = P(A)P(B)$


**Пример.** Бросаем кубик

* $A$ - Выпало четное число, тогда $P(A) = \frac{1}{2}$

* $B$ - Выпало число, кратное трем, тогда $P(B) = \frac{1}{3}$

* $C$ - Выпало число $\ge 4$, тогда $P(C) = \frac{1}{2}$

$P(A \cap B) = P(\{6\}) = \frac{1}{6} = P(A)P(B)$ - Независимы

$P(A \cap C) = P(\{4, 6\}) = \frac{1}{3} \ne \frac{1}{4} = P(A)P(C)$ - Зависимы

$P(B \cap C) = P(\{6\}) = \frac{1}{6} = P(B)P(C)$ - Независимы

<h3 style="color:#2E8B57">Задача № 1</h3>

В мешке лежит $12$ шаров для игры в петанк: $6$ белых и $6$ черных, причем $2$ белых и $4$ черных шара с насечкой. Из мешка вынимают наугад один шар. Пусть $A$ — событие, означающее, что вытащили белый шар, а $B$ — событие, означающее, что вынули шар с насечкой. Найдите вероятности $P(A)$, $P(B)$ и $P(A\cap B)$. Будут ли события $A$ и $B$ независимы? В ответе приведите разделенные пробелами найденные вероятности в виде обыкновенных дробей и одно из слов "зависимы" или "независимы" (без кавычек).

<h4 align='center'>Решение:</h4>

В мешке $12$ шаров, из которых $6$ белых, поэтому $P(A)=\frac{6}{12}=\frac{1}{2}$. Шаров с насечкой там тоже $6$, поэтому $P(B)=\frac{6}{12}=\frac{1}{2}$. Белых шаров с насечкой там $2$, поэтому $P(A\cap B)=\frac{2}{12}=\frac{1}{6}$. Поскольку $P(A\cap B)=\frac{1}{6}\neq\frac{1}{2}\cdot\frac{1}{2}=P(A)P(B)$, события зависимы.

<h3 style="color:#2E8B57">Задача № 2</h3>

Из колоды, содержащей $52$ карты, наугад извлекается одна карта. Событие $A$ означает, что извлеченная карта является дамой, событие $B$ — что извлечена карта пиковой масти. Найдите вероятности событий $A$, $B$ и $A\cap B$. Независимы ли события $A$ и $B$? В ответе для каждой из вариантов задачи приведите разделенные пробелами найденные вероятности в виде обыкновенных дробей и одно из слов "зависимы" или "независимы" (без кавычек).

<h4 align='center'>Решение:</h4>

Вероятность вытащить пиковую даму равна $\frac{1}{52}$, поэтому $P(A\cap B) = \frac{1}{52} = \frac{1}{13}\cdot\frac{1}{4} = P(A)P(B)$ и события $A$ и $B$ независимы.

### Независимые в совокупности события

События $A_1, A_2, \ldots, A_n$ независимы в совокупности если для любых $1 \le i_1 < i_2 < \ldots < i_k \le n$ - (для любого набора различных индексов $i_1, \ldots, i_k$ от $1$ до $n$) выполняется:

$P(A_{i_1} \cap A_{i_2} \cap \ldots \cap A_{i_k}) = P(A_{i_1})\cdot \ldots \cdot P(A_{i_k})$

**Пример.** На гранях тетраэдра $1, 2, 3, 4$

Рассмотрим события, что выпало $A = \{1, 2\}$, $B = \{1, 3\}$, $C = \{1, 4\}$

* тогда вероятность каждого события $P(A) = P(B) = P(C) = \frac{1}{2}$

* тогда что означет наступление события двух событий вместе $A \cap B = B \cap C = C \cap A = \{1\}$ - эти события реализуемы, если выпадает еденица.

* тогда  вероятности таких пар событий $P(A \cap B) = P(B \cap C) = P(C \cap A) = \frac{1}{4}$


Что означает наступление трех событий сразу? Это означет что выпала еденица $P(A \cap B \cap C) = \frac{1}{4} \ne \frac{1}{8} = P(A)P(B)P(C)$.

Следовательно, тройка событий не является независимыми в совокупности, но является независимыми попарно.

Может показаться, что условий в определении независимости событий в совокупности слишком много и они изыбточные. Этот пример показывает, что это не так. Может оказаться, что любая пара событий независима, но в совокупности события будут зависимыми.

**Пример.** Кидаем два кубика. $i$ - количество очков, которое выпало на 1 кубике, $j$ - количество очков на втором кубике.

$\Omega = \{(i, j): i, j = 1, 2, \ldots, 6 \}$

$A = \{(i, j): j=1, 2, 5\}$, тогда $P(A) = \frac{1}{2}$

$B = \{(i, j): j=4, 5, 6\}$, тогда $P(B) = \frac{1}{2}$

$C = \{(i, j): i+j=9\}$, тогда $P(C) = \frac{1}{9}$

Что означет, что случились все три события? События $A, B$ могут произойти одновременно, если выпала $5$ на втором кубике. Если выпала $5$ на втором кубике, тогда на первом кубике может выпасть только $4$, так как для события $C$ сумма должна равняться $i + j = 9$.

$P(A \cap B \cap C) = P(\{4, 5\}) = \frac{1}{36} = P(A)P(B)P(C)$

Но для любой из пар, условия независимости не будут выполнятсья:

$P(A \cap B) = \frac{1}{6} \ne \frac{1}{4} = P(A)P(B)$

$P(A \cap C) =  P(\{4, 5\}) = \frac{1}{36} \ne \frac{1}{18} = P(A)P(C)$

$P(B \cap C) = \frac{1}{12} \ne \frac{1}{18} = P(B)P(C)$

Следовательно, нельзя отказыватсья от условий из определения независимости в совокупности.

<h3 style="color:#2E8B57">Задача № 3</h3>

Подбрасываются две правильные игральные кости. Событие $A_k$ при $k=2,3,\dots,12$ означает, что сумма очков на кубиках равна $k$, а событие $B_n$ при $n=1,2,\dots,6$ означает, что на первом кубике выпало $n$ очков. Для каких пар $(k, n)$ события $A_k$ и $B_n$ независимы?

<h4 align='center'>Решение:</h4>

Напишем вероятности событий $A_k$ и $B_n$: 

$P(B_n)=\frac{1}{6}$, $P(A_k)=\frac{k-1}{36}$ при $k=2, 3, \ldots, 7$ и $P(A_k)=\frac{13-k}{36}$ при $k=8, 9, \ldots, 12$. 

Вероятность события $A_k\cap B_n$ также легко находится, поскольку это событие означает, что на первом кубике выпало $n$, а в сумме выпало $k$. Если $k\leqslant n$, то это невозможное событие и его вероятность равна нулю. Тогда события $A_k$ и $B_n$ не могут быть независимыми, ибо $P(A_k)P(B_n)\gt0=P(A_k\cap B_n)$. Если же $k\gt n$, то это означает, что на втором кубике выпало $k-n$ и вероятность выпадения такой пары очков равна $\frac{1}{36}$. 

Установим когда $k\gt n$ и $P(A_k)P(B_n)=P(A_k\cap B_n)$. Рассмотрим случай $k\leqslant 7k⩽7, тогда \frac{1}{36}=\frac{k-1}{36}\cdot\frac{1}{6}$ и, значит, $k=7$ и $n$ — любое. Далее рассмотрим случай $k\geqslant 8$, тогда $\frac{1}{36}=\frac{13-k}{36}\cdot\frac{1}{6}$ и, значит, $13-k=6$, что для $k\geqslant 8$ невозможно. 

Отсюда находим ответ: $k=7$, $n$ — любое число от $1$ до $6$.

<h2 style="color:black" align="center">1.8 Схема Бернули</h2>

Имеется монетка с смещенным центром тяжести, где она выпадает орлом с вероятностью:

$p \in [0, 1] \quad q=1-p$

Выпадение орла соответсвует $1$, а выпадение решки $0$. Подбросим монету $n$ раз и выпишем последовательность полученных результатов:

$\omega = (x_1, \ldots, x_n)$, где $x_i \in \{0, 1\}$

Так как подбрасываение монеты независимо, посчитаем вероятность выпадения данной последовательности:

$\large P(\omega) = p^{\sum\limits_{i=1}^n x_i} (1-p)^{n - \sum\limits_{i=1}^n x_i}$

**Пространство элементарных исходов** Следовательно, пространство элементархы исходов состоит из всевозможных последовательностей нулей и едениц:

$\Omega = \{ \omega = (x_1, \ldots, x_n), x_i \in \{0, 1\}\}$

**Подсчет вероятности события наступления фиксированного количества успехов**

Рассмотрим события $A_k$, которое означает, что $k$ раз выпал орёл.

$A_k = \{\omega=(x_1, \ldots, x_n): x_1 + \ldots + x_n = k\}$ где $x_i \in \{0, 1\}$

Тогда посчитаем вероятность такого события. Для этого необходимо количество исходов события $A_k$ умноженное на вероятность элементарного исхода $p^k (1-p)^{n-k}$. Ну а количество исходов это просто количество способов раставить на $n$ позициях $k$ едениц. Получаем вероятность равна:

$P(A_k) = C_n^k p^k (1-p)^{n-k}$

$\sum\limits_{k=0}^n P(A_k) = \sum\limits_{k=0}^n C_n^k p^k (1-p)^{n-k} = (p + (1-p))^n = 1$

Такой набор вероятностей биномиальным распределением:

$(P(A_0), P(A_1), \ldots, P(A_n))$

### Формальное определение

Из наших определений совершенно неясно, что вообще могут существовать сто событий, независимых в совокупности. Формальное определение схемы Бернулли показывает, что могут. 

Рассмотрим те же самые элементарные события:

$p \in [0, 1] \quad q=1-p$

Выпадение орла соответсвует $1$, а выпадение решки $0$. Подбросим монету $n$ раз и выпишем последовательность полученных результатов:

$\omega = (x_1, \ldots, x_n)$, где $x_i \in \{0, 1\}$, $\quad \Omega = \{ \omega = (x_1, \ldots, x_n), x_i \in \{0, 1\}\}$




Так как подбрасываение монеты независимо, посчитаем вероятность выпадения данной последовательности:

$\large P(\omega) = p^{\sum\limits_{i=1}^n x_i} (1-p)^{n - \sum\limits_{i=1}^n x_i}$

Тогда сумма всех вероятностей по всем элементарым событиям это анлогично сложению всех событий $A_k$:

$\sum\limits_{\omega \in \Omega} P(\omega) = \sum\limits_{k=0}^n P(A_k) = 1$

Получаем, что все вероятности не отрицательны и их сумма равна $1$. Следовательно, так определить вероятности на элементарных событиях $\omega$ можем.

Рассмотрим событие $B_k$, которое означает, что на $k$ бросании выпала $1$:

$B_k = \{\omega: x_k = 1\}$

Можно проверить, что все $B_k$-ые будут независимым событием при разных $k$. Таким образом, события означающие что в разные моменты времени выпал орел они не зависят дург от друга. Следовательно бросание монеты тоже независимо. Делаем выводы, что данная конструкция возможна и бывает $n$ независимых подбрасываний монеты и никакого противоречия в конструкции нет.

**Небольшая сводка:**

мы определяем вероятность $P(\omega)P(ω) по формуле P(\omega)=p^{\sum x_k}(1-p)^{n-\sum x_k}$. Это определение корректно (сумма всех вероятностей единица). И при таком определении события $B_k$ оказываются независимыми. Это можно проверить формально, но довольно занудно. Либо нужно сначала проверить, что независимость событий $B_k$ равносильна тому, что $P(C_1\cap C_2\cap\dots\cap C_n)=P(C_1)P(C_2)\dots P(C_n)$, где $C_k=B_k$ или $C_k=\overline{B_k}$ (т.е. $2^n$ таким равенствам).
