<h2 style="color:black" align="center">1.2 Вероятностная модель эксперимента</h2>

### Конечно число исходов

Пространство элементарных исходов/событий $\Omega = \{\omega_1, \ldots, \omega_N\}$, где $\omega$ - элементарное событие или элементарный исход.

**Пример 1.** Однократное подбрасывание монеты

$\Omega = \{\text{Орёл, Решка}\}$, $|\Omega| = 2$

**Пример 2.** Однократное подбрасывание игральной кости

$\Omega = \{1, 2, 3, 4, 5, 6\}$, $|\Omega| = 6$

**Пример 3.** $n$-кратное подбрасывание монеты

$\Omega = \{\omega: \omega = (a_1, \ldots, a_n), \text{где } a_i = \text{О или Р}\}$, $|\Omega| = 2^n$

**Пример 4.** Есть 4 красных шара и 6 зеленых. Знаумируем наши шары, где от 1 до 4 - это красные шары, а числа от 5 до 10 это зеленые шары. Выберем 3 шара

$\Omega = \{\omega: \omega = (a_1, a_2, a_3), \text{где } a_i  \text{- различные числа от 1 до 10}\}$, $|\Omega| = C_{10}^3$

**Пример 5** Выбор с возвращением. В урне $m$ шаров, выбирают $n$

Если важен порядок, тогда пространством элементарных исходов будет все возможные последовательности из $n$ чисел:

$\Omega = \{\omega: \omega = (a_1, \ldots, a_n), a_i \in \{1,\ldots, m \}\}$, $|\Omega| = \bar{A}_m^n$

Если не важен порядок, тогда пространством элементарных исходов будет все возможные неупорядоченные наборы из $n$ чисел:

$\Omega = \{\omega: \omega = [a_1, \ldots, a_n], a_i \in \{1,\ldots, m \}\}$, $|\Omega| = \bar{С}_m^n$

### Событие

Если имеется пространство элементарынх исходов $\Omega = \{\omega_1, \ldots, \omega_N\}$, тогда событием называется произвольное подмножество пространства элементарынх исходов $A \subset \Omega$.

**Пример 1.** 3-кратное подбрасывание монеты

$A = \{\text{ООР}, \text{ОРО}, \text{РОО} \}$ - выпало два орла 

$B = \{\text{РРР}\}$ - ни разу не выпал орел

$A \cup B$ - Выпало четное число орлов, значит добавится событие, когда орлы вообще не выподают.

**Множества** для событиямй справедливы операции множествами, так как событие является множеством.

$A \cup B = \{\omega \in \Omega: \omega \in A \text{ или } \omega \in B\}$

$A \cap B = \{\omega \in \Omega: \omega \in A \text{ и } \omega \in B\}$

$\bar{A} = \{\omega \in \Omega: \omega \notin A\}$


**Примеры событий**

Рассмотрим "Пример 5. Выбор с возвращением. В урне $m$ шаров, выбирают $n$":

Пример события для упорядоченного выбора: Достаем шары с номерами идущими в возрастающем порядке. Для неупорядоченного выбора, данное событие не будет иметь смысла.

Пример события для двух упорядоченного и неупорядоченного выбора шаров: Все время достаем разные шары, тогда если $n > m$, то такое событие будет невозможным. Изначально выбор с возращением, но так как мы достаем каждый раз разные шары, тогда $n > m$ не может быть.

<h3 style="color:#2E8B57">Задача № 1</h3>

Из какого количества элементарных событий состоит пространство элементарных событий для следующих испытаний:

* производится выстрел по мишени, представляющей собой $10$ концентрических кругов, занумерованных числами от $1$ до $10$;
* три раза подбрасывается игральная кость;
* наудачу извлекается одна кость из полной игры домино.

<h4 align='center'>Решение:</h4>

1) Можно попасть в любой из десяти занумерованных кругов или не попасть по мишени, то есть $11$ исходов.

2) При каждом из трех бросаний можно получить по шесть вариантов, то есть $6^3=216$ исходов.

3) Полная игра домино содержит $28$ различных костей, то есть 28 исходов.

<h3 style="color:#2E8B57">Задача № 2</h3>

Из какого количества элементарных событий состоит пространство элементарных событий для следующего испытания: производится выстрел по мишени, представляющей собой $10$ концентрических кругов, занумерованных числами от $1$ до $10$, а затем столько раз кидается игральная кость, сколько очков выбито на мишени.

<h4 align='center'>Решение:</h4>

Если при выстреле промахнулись, то кость не бросается. В таком случае исход единственный. Если при выстреле попали в $1$, то кость бросается один раз, то есть исходов $6$. Если при выстреле попали в $k$, то кость бросается $k$ раз, то есть $6^k$ исходов. Вычислим общее количество исходов как сумму геометрической прогрессии:

$$\sum\limits_{k=0}^{10}6^k=\frac{6^{11}-1}{6-1}=72559411.$$

<h3 style="color:#2E8B57">Задача № 2</h3>

Из какого количества элементарных событий состоит каждое из следующих случайных событий:

* сумма двух наудачу выбранных однозначных чисел равна пятнадцати: (элементарное событие — появление пары однозначных чисел $(m,n)$);
* наудачу выбранная кость из полной игры домино оказалась дублем (элементарное событие — появление кости $m:n$);
* наудачу вырванный листок из нового календаря соответствует тридцатому числу (элементарное событие — появление одного из листков календаря).

<h4 align='center'>Решение:</h4>

1) Подходят лишь четыре пары чисел $(6;9), (7;8), (8;7)$ и $(9;6)$.

2) В полной игре домино содержится $7$ дублей: $0:0, 1:1,, \ldots, 6:6$.

3) Тридцатого числа не бывает только в феврале, значит в календаре есть $11$ листков с тридцатым числом.

<h2 style="color:black" align="center">1.3 Вероятностные пространства</h2>

### Определение веротяностного пространтсва

Пусть имеется пространство элементарных событий $\Omega = \{\omega_1, \ldots, \omega_N\}$, состоящее из $N$ исходов и каждому из возможных исходов припищем число от $0$ до $1$, обозначив через:

$p_1, \ldots, p_N \ge 0$ 

Добавим условие нормировки, указава, что сумма этих чисел должна равняться $1$:

$p_1 + \ldots + p_N = 1$

Тогда эти числа мы будем называть вероятностями элементарных событий, следовательно $p_i$ это вероятность события $\omega_i$.


**Вероятность события $A$**

Рассмотрим все возможные элементы $\omega_k$, которые содержатся в множестве $A$ и просуммируем соответствующие им $p_k$:

$\large P(A) = \sum\limits_{k: \omega_k \in A} p_k$ - вероятность события $A$

### Свойства вероятности

1. $P(\emptyset) = 0, \quad P(\Omega) = 1$

2. $0 \le P(A) \le 1$

3. Если $A \cap B = \emptyset$, то $P(A \cup B) = P(A) + P(B)$. Этот пунк следует из 4-го свойства, так как если $A \cap B = \emptyset$, тогда мы вычитаем ноль и получаем 3-е свойство

4. $P(A \cup B) = P(A) + P(B) - P(A \cap B)$

**Проверим 4 свойство**, записав все вероятности с помощью определения:

$\large \sum\limits_{k: \omega_k \in A \cup B} p_k = \sum\limits_{k: \omega_k \in A} p_k + \sum\limits_{k: \omega_k \in B} p_k + \sum\limits_{k: \omega_k \in A \cap B} p_k$

Если $\omega_k$ не лежит ни в множестве $A$ ни $B$, тогда соответсвующий ему $p_k$ не лежит ни в одной из сумм, следовательно для этих $p_k$ равенство выполняется, так как слева и справа ничего нет.

Если $\omega_k$ лежит только в множестве $A$, тогда $p_k$ есть в сумме слева и в первой сумме справа, но нет во всех остальных суммах, поэтму для таких $p_k$ выполняется равенство между суммами.

Аналогично, если $\omega_k$ лежит только в множестве $B$.

Если $\omega_k$ лежит в множестве $A$ и $B$, тогда соответсвующий ему $p_k$ учтено в сумме слева, в сумме для $\omega_k \in A$, $\omega_k \in B$ и учтено в сумме $\omega_k \in A \cap B$ с отрицательным знаком, следовательно, $p_k$ в правой часте учтено один раз как и в левой части, значит равенство выполняется.

И теперь, формула доказана.

5. $P(A \cup B) \le P(A) + P(B)$ Это свойство следует из 4-го, так как мы вчитаем $P(A \cap B)$

6. $P(\bar{A}) = 1 - P(A)$ Это свойство является следствием из 3-го свойства, так как если в 3-ем свойстве вместо $B$ подставить событие $\bar{A}$, тогда мы получим пустое пересечение $A \cap \bar{A} = \emptyset$, тогда $P(A \cup \bar{A}) = P(A) + P(\bar{A})$




7. $P\big(\bigcup\limits_{k=1}^n A_k\big) \le \sum\limits_{k=1}^n P(A_k)$

**Доказательство по индукции** База $n = 2$, тогда $P(A_1 \cup A_2) \le P(A_1) + P(A_2)$ - это неравенство было 5-свойством, поэтому оно уже проверено.

Остается доказать переход $n \to n + 1$. Заведем вспомогательное множество $B := \bigcup\limits_{k=1}^n A_k$ и тогда:

$P\big(\bigcup\limits_{k=1}^{n+1} A_k \big) = P(B \cup A_{n + 1}) \le$

По 5 свойству или по базе, получаем, что:

$\le P(B) + P(A_{n + 1}) \le$

Но $B$ является объединением множеств, тогда по индукционному предположению вероятность события $P\big(\bigcup\limits_{k=1}^n A_k\big)$ не превосходит суммы вероятностей:

$P\big(\bigcup\limits_{k=1}^n A_k\big) \le \sum\limits_{k=1}^n P(A_k)$, значит, вместо вероятности $B$ можем подставить следуюую сумму и от этого выражение станет только больше:

$\le \sum\limits_{k=1}^n P(A_k) + P(A_{n+1})$

Итого, мы получаем, что $P\big(\bigcup\limits_{k=1}^{n+1} A_k \big) \le \sum\limits_{k=1}^n P(A_k) + P(A_{n+1})$, что вместе дает сумму веротяностей событий $a_1, \ldots, a_{n+1}$

8. $P\big(\bigcup\limits_{k=1}^n A_k\big) = \sum\limits_{k=1}^n P(A_k) - \sum\limits_{j < k} P(A_j \cap A_k) + \sum\limits_{i < j < k} P(A_i \cap A_j \cap A_k) - \ldots + (-1)^{n-1} P(A_1 \cap \ldots \cap A_n)$

**Доказательство по индукции** База $n = 2$, тогда $P(A_1 \cup A_2) = P(A_1) + P(A_2) - P(A_1 \cap A_2)$, а это соответствует свойствую $4$, которое доказано.

Остается доказать переход $n \to n + 1$. Заведем вспомогательное множество $B := \bigcup\limits_{k=1}^n A_k$ и тогда:

$P\big(\bigcup\limits_{k=1}^{n+1} A_k \big) = P(B \cup A_{n + 1}) =$

По 4 свойству или по базе, получаем, что:

$= P(B) + P(A_{n + 1}) - P(B \cap A_{n + 1}) =$

---
А что значит событие $P(B \cap A_{n + 1})$? Подставим $B$ и получим объединение всевозможных пересечений множества $A_k$ с $A_{n+1}$

$P(B \cap A_{n + 1}) = A_{n+1} \bigcup\limits_{k=1}^n A_k = \bigcup\limits_{k=1}^n (A_k \cap A_{n + 1})$

Следовательно $P(B \cap A_{n + 1})$ - вероятность объеденения некоторых $n$ множеств, а для такой вероятности мы уже доказали формулу включений-исключений.

---

Расписываем $P(B)$ по формуле включений-исключений и для $P(B \cap A_{n + 1})$ тоже, только заместо $P(B \cap A_{n + 1})$ возьмем $\bigcup\limits_{k=1}^n (A_k \cap A_{n + 1})$:


$= P(A_{n + 1}) 
+ \big(
    \sum\limits_{k=1}^n P(A_k) 
        - \sum\limits_{j < k \le n} P(A_j \cap A_k) + \ldots 
  \big) 
- \big(
    \sum\limits_{k=1}^n P(A_k \cap A_{n+1}) 
        - \sum\limits_{j < k \le n} P(A_j \cap A_k \cap A_{n + 1}) + \ldots 
  \big)$
  
Имеется вероятность $P(A_{n + 1})$ множества $A_{n + 1}$. После идет сумма вероятностей остальных $\sum\limits_{k=1}^n P(A_k)$ множеств $a$. Именно то, что и нужно в формуле включений-исключений - это сумма вероятностей всех одиночных событий. Что вычитаем? Вычитаем вероятность всех попарных пересечений где $i, j$ меньше либо равны $n$.

Что представляет последнее выражение? Вычитается вероятность попарных пересечений $\sum\limits_{k=1}^n P(A_k \cap A_{n+1})$ и это выражение совместно с выражением $\sum\limits_{j < k \le n} P(A_j \cap A_k)$ является суммой всевозможных вероятностей для попарных пересечений множеств с индексами от $1$ до $n+1$. 

После выражения $\sum\limits_{j < k \le n} P(A_j \cap A_k)$ в первой скобе должны прибавить вероятность всех возможных пересечений по трое, в случае если индекс от $1$ до $n$. Так как перед второй скобкой стоит минус и перед выражением $\sum\limits_{j < k \le n} P(A_j \cap A_k \cap A_{n + 1})$ стоит минус, тогда мы должны прибавить пересечения по три, множеств. Следовательно, мы рассматриваем все возможные пересечений из первой скобки, когда ни один из индексов не равен $n + 1$, а во второй скобке когда один из индексов равняется $n + 1$. 

Значит мы рассматриваем все возможные пересечения по 3 и считаем их вероятности и складываем и так далее разбираемся с этими суммами и убеждаемся в том, что получается искомая нами формула.
