**Асимптотика** - подразумивает функцию, где аргумент функции стремится к бесконечности. Когда мы имеем функцию $f(x)$ и аргумент $x$ стремится к бесконечности, то можно сказать: как себя ведет $f(x)$ в асимптотике?

Правило создания чисел в математике называется **последовательностью**. Последовательность записывается в виде $\{a_n\}$, где $a$ - элемент последовательности, $n$ - номер элемента последовательности.

$\{a_n\}$ - некая формула, которая позволяет генерировать числа. Наша задача заключается в том, чтобы понять какое значение принимает последний (предельный) элемент последовательности. Иначе это можно сказать как "Чему равняется предел последовательности?"

$\large \lim\limits_{n \to \infty} a_n$ - предел

Почему "предел"? Представим, что $a_n$ - это некая формула:

$\large a_n = \frac{n + 1}{n + 3}$ - мы имеем правило получения каких-то чисел, как и для чего эти числа возникают мы пока не говорим. Мы только знаем, что есть правило по которому эти числа могут быть созданы. Мы хотим понять куда в бесконечности стремится элемент последовательности $a_n$. Для вычисления этого необходимо посчитать предел этой последовательности.

<h1 style="color:#008B8B">1. Определение предела по Коши</h1>

## Неформально

У нас есть последовательность $a_n$ - правило, которое создает числа. Мы хотим понять, существует ли предел у этой последовательности или он не существует. Если он суещствует, то мы полагаем что он равняется некоторому числу $L$:

$\large \lim\limits_{n \to \infty} a_n = L$

Мы хотим понять, что такое "предел"? Поэтому дадим определение предела:

Если взять числовую прямую и где-то поставить точку $L$, которая считается предельной точкой для этой последовательности, то $L$ будет является пределом елси: 

мы взяли последовательность $a_n$ и все значения элеметов последоватлеьности решили нарисовать на данной числовой прямой:

<img src='img/1_Introduction/1.png' width="700" height="700">

Если мы можем вокруг точки $L$ нарисовать интервал такой, чтобы его ширина была бесконечно маленькая по размеру, иначе бесконечно маленький по размеру интервал. Если начиная с какого-то элемента все остальные элементы попадают внутрь этого интервала, то только тогда $L$ называется пределом.

В нашем случае с $a_{999}$ начинается предел, который захватывает все элементы после него.

## Общее определение

$L$ будет является предельной точкой, если начиная с какого-то элемента $a_N$, все элементы с последующими индексами ($a_{N+1}\ldots$) попадают внутрь интервала вокруг точки $L$. При этом, мы хотим чтобы интервал был бесконечно маленьким по размеру. 

Тогда говорят, пускай есть число $\epsilon$, которое считается самым маленьким положительным числом $\epsilon > 0 \qquad \epsilon = 0.000\ldots0001$

Интервал бесконечно маленькой ширины вокруг $L$ - это интервал, который описывается от $(L - \epsilon; L + \epsilon)$ - Определение данного интервала.

<img src='img/1_Introduction/2.png' width="700" height="700">

Последовательность $a_n$ (все элементы этой последовательности) стремится в точку $L$, когда $n$ стремится к бесконечности:

$\large \lim\limits_{n \to \infty} a_n = L$ - Запишем что это означает с точки зрения Коши:

### Определение

$$\large \lim\limits_{n \to \infty} a_n = L
\iff
\forall \epsilon > 0 \quad \exists N \in \mathbb{N}: \forall n > N \Rightarrow |a_n - L| < \epsilon$$

Для любого бесконечно маленького числа $\epsilon$, которое может существовать ($\forall \epsilon > 0$). Всегда существует такой номер $N$ большое ($\exists N \in \mathbb{N}$) что (:), все элементы последовательности $a_n$, у которых номер $n$ больше порогового $N$ ($\forall n > N$) выполняется, что элементы лежат внутри бесконечно маленького интевала, вокруг точки $L$ ($L - \epsilon < a_n < L + \epsilon$). Если это определение выполняется, тогда $L$ называется пределом.

**Запись интервала:** Запись $L - \epsilon < a_n < L + \epsilon$ имеет неудобный вид, поэтому его записывают как: 

$\large -\epsilon < a_n - L < \epsilon$ и запишем по модулю:

$\large |a_n - L | < \epsilon$

<h3 style="color:#2E8B57">Задача № 1</h3>

Возьмем последовательность:

$\large a_n = \frac{100}{n^3 + 10}$ - Куда стремится последовательность, если $n$ стремится к бесконечности?

$\large \lim\limits_{n \to \infty} a_n = 0$ - Давайте проверим это по определению Коши, равняется ли данный предел нулю?

<h4 align='center'>Решение:</h4>

В запись $|a_n - L| < \epsilon$ подставим все что знаем:

$\large \bigg|\frac{100}{n^3 + 10} - 0\bigg| < \epsilon 
\Rightarrow
\bigg|\frac{100}{n^3 + 10}\bigg| < \epsilon  
\Rightarrow
\frac{100}{n^3 + 10} < \epsilon \Rightarrow$

Разделим все на $\epsilon$:

$\large \frac{100}{\epsilon} < n^3 + 10
\Rightarrow
\frac{100}{\epsilon} - 10 < n^3
\Rightarrow
n^3 > \frac{100}{\epsilon} - 10
\Rightarrow
n > \sqrt[3]{\frac{100}{\epsilon} + 10}$

Предаоложим что $\epsilon = 0.1$, тогда получаем:

$\large \sqrt[3]{\frac{100}{0.1} + 10} = \sqrt[3]{990} \approx 10$ - это $N$ большое.

Тем самым, начиная с 9-го элемента, все последующие будут входить в наш интервал. Значит, пороговый номер существует и он равен чуть меньше 10, так как 10 уже входит в интервал. Если шаг $n$ равен одному, то $N = 9$.

**Зависимость между нулем и бесконечностью:**

$\large \frac{1}{0} \to \infty \qquad \frac{1}{\infty} \to 0$

<h1 style="color:#008B8B">1. Непрерывность</h1>

У нас есть некоторая фукнция $f(x)$, эту функцию можно назвать непрерывной, если ее можно нарисовать не отрывая руки:

<img src='img/4_Function-analysis/1.png' width="700" height="700">

Функция называется непрерывной, если выполнятеся следующее условие:

* К функции можно подойти слева
* К функции можно подойти справа
* В любой точке, к которой предел стремится слева или справа, можно посчитать значение функции $f(a)$

## Определение

Функция $f(x)$ называется непрерывной, если предел в точке $a$ слева, равняется пределу в точке $a$ справа и равняется значению функции в точке $f(x)$:

$\large \lim\limits_{x \to a^{-}} f(x) = \lim\limits_{x \to a^{+}} f(x) = f(a)$

Предел слева равен пределу справа, если они существуют и равны, можно заменить на предел в точке $a$:

$\large \lim\limits_{x \to a^{-}} f(x) = f(a)$

**Обозначение:**

Когда мы говорим, что $x$ стремится в $a$ слева ($x \to a^{-}$) - аналог этой записи ($x \to a - 0$). Данные записи обозначают, что $x$ стремится в $a$ и $x$ меньше $a$:

$\large x \to a^{-} \quad (x \to a - 0) \iff x \to a \quad \cup \quad x < a$

$\large x \to a^{+} \quad (x \to a + 0) \iff x \to a \quad \cup \quad x > a$


<h3 style="color:#2E8B57">Задача № 1</h3>

Возьмем функцию 

$\large f(x) = x^4$

Является ли эта функция непрерывной в точке 2?

<h4 align='center'>Решение:</h4>

По определению необходимо посчитать предел в точке 2 слева, справа и значение функции в точке 2.

$\large \lim\limits_{x \to 2^{-}} f(x) = \lim\limits_{x \to 2^{-}} x^4 = 16$

$\large \lim\limits_{x \to 2^{+}} f(x) = \lim\limits_{x \to 2^{+}} x^4 = 16$

$\large f(2) = 2^4 = 16$

Все 3 значения равны, значит функция является непрерывной в точке два.

## Проверка на непрерывность

Проводить **проверку на непрерывность** для функции необходимо тогда, когда возникает две группы проблемных точек:

1. Точки, где возникает деление на 0.

2. Точки, где функция может менять свой функциональный вид, в случае системы.

Наше определение может не выполняться. По какой причине нарушается оределение, позволит нам оценить каким типом разрыва обладает функция.

### Виды разрывов:

**Разрыв типа 1 - устраняемый.** Оба предела слева и справа существуют и равны друг-другу. Но значение функции в точке $a$ не существует или не равно этим пределам.

**Пример 1.** $\large f(x) = \frac{\sin(x)}{x}$

Так как точке ноль возникает деление на ноль, необходимо проверить точку при $x=0$:

$\large \lim\limits_{x \to 0^{-}} f(x) = \lim\limits_{x \to 0^{-}} \frac{\sin(x)}{x} = 1$

$\large \lim\limits_{x \to 0^{+}} f(x) = \lim\limits_{x \to 0^{+}} \frac{\sin(x)}{x} = 1$

$\large f(0) = \not{\exists}$ - значение в 0 не существует

Данный тип разыва называется удаляемый или устраняемый. На графике это означает, что у нас есть функция и предел справа равен пределу слева, но в самой точке значение не вычилсяется.

В анализе данных на данный тип разрыва не обращают внимание, так как это всего одна точка. Данную функцию можено сделать непрерывной, можно доопределить ее: для всех $x \ne 0$, она остается неизменной, а когда $x = 0$ она будет равна 1, после чего функция станет непрерывной. Почему 1? При 1, функция совпадает со значением пределов и определение выполняется.

**Разрыв типа 1.** Предел в точке $a$ слева существует и не равняется значению предела справа, который тожже существует. Данный тип разрыва называется скачок, рывок.

**Пример 2.**

<img src='img/4_Function-analysis/2.png' width="500" height="500">

$\large f(x) = \frac{|x|}{x}$ - проблемная точка 0 - деление на ноль.

$\large \lim\limits_{x \to 0^{-}} f(x) = \lim\limits_{x \to 0^{-}} \frac{-x}{x} = -1 < \infty$ - При стремлении слква мы знаем, что $x$ меньше нуля - отрицательный. И расскрываем модуль с минусом $|x| = -x$.

$\large \lim\limits_{x \to 0^{+}} f(x) = \lim\limits_{x \to 0^{+}} \frac{x}{x} = 1 < \infty$

$\large f(0) = \not{\exists}$ - на ноль делить нельзя.

У данной функции проблема с интегрированием в точке 0. Если фукнция имеет данный тип разрыва и мы не планируем ее интегрировать, тогда с ней можно работать. Так как она везде существует, кроме одной точки. Необходимо просто разбить на два случая $x > 0$ и $x < 0$.

**Разрыв типа 2.** Если предел слева от фукнции уходит в бесконечность или предел справа уходит в бесконечность, либо оба предела уходят в бескончность.

**Пример 3.**

$\large f(x) = \frac{2}{x}$

<img src='img/4_Function-analysis/3.png' width="500" height="500">

$\large \lim\limits_{x \to 0^{-}} f(x) = \lim\limits_{x \to 0^{-}} \frac{2}{x} = \frac{2}{0} = - \infty$

Расчет одного предела уже говорит о разрывае данного типа. Точка, в которой произошёл разрыв типа 2, в нашем случае это ноль - вертикальная асимптота функции. Значение функции уходит в бесконечность, что невозможно применять для обработки данных.

<h3 style="color:#2E8B57">Задача № 2</h3>

$\large f(x) = \frac{(x-1)(x + 1)}{(x+3)(x-1)}$ - Необходимо проверить ее на непрерывность для всех значений $x$ ($\forall \in \mathbb{R}$). Если мы хотим гарнитровать, что функция будет непрерывна на всех $x$, значит она должна быть непрерывана в двух группах проблемных точек. В этом примере возникает первая группа проблемых точке - деление на ноль, проверим их.

<h4 align='center'>Решение:</h4>

$\large x=-3,\quad x=1$

$\large \lim\limits_{x \to -3^{-}} f(x) = \frac{-4 \cdot (-2)}{-0} = + \infty$ 

$\large \lim\limits_{x \to -3^{+}} f(x) = \frac{-4 \cdot (-2)}{+0} = - \infty$

$\large f(-3) = \not{\exists}$ - на ноль делить нельзя. Делаем выводы, что это разрыв типа 2.

Когда считаем предел в точке 1 слева и справа, тогда в обоих этих пределах возникает неопределенность 0 на 0. Эта проблема возникает из-за того, что в числителе и в знаминаетеле имеется (x - 1). Поэтому мы можем сократить на данную скобку. В обычной функции такое сокращение производить нельзя, так как изменится ОДЗ функции, а в пределах так можно.

$\large \lim\limits_{x \to 1^{-}} f(x) = \frac{0}{0} = \lim\limits_{x \to 1^{-}} \frac{x+1}{x+3} =  \frac{2}{4}$ 

$\large \lim\limits_{x \to 1^{+}} f(x) = \frac{0}{0} = \lim\limits_{x \to 1^{+}} \frac{x+1}{x+3} = \frac{2}{4}$

$\large f(1) = \not{\exists}$ - на ноль делить нельзя. Делаем выводы, что это устраняемый разрыв.
